## Why This Unit

Partial derivatives measure the slope in individual coordinate directions. But to understand how $f$ changes when we move in an *arbitrary* direction — not just along a coordinate axis — we need the total derivative. This is the object that justifies the gradient as the direction of steepest ascent (Module 03), and it is also what the multivariable chain rule (Unit 02-04) differentiates.

The tangent plane is the geometric realisation of the total derivative: the flat surface that best approximates the graph of $f$ near a point. Near the MLE, the log-likelihood surface is well approximated by its tangent plane — and the next term in the Taylor expansion, quadratic in the parameters, gives the Gaussian (Laplace) approximation to the posterior.

---

## 1. Total Differentiability

### 1.1 The Key Distinction

The existence of partial derivatives $\frac{\partial f}{\partial x}$ and $\frac{\partial f}{\partial y}$ at a point does **not** guarantee that $f$ is differentiable there in the full multivariable sense. Partial derivatives measure the rate of change along coordinate axes only; a function can have well-defined partial derivatives but behave badly in diagonal directions.

**Example (pathological).** The function
$$
f(x,y) = \begin{cases} \frac{xy}{x^2 + y^2} & (x,y) \neq (0,0) \\ 0 & (x,y) = (0,0) \end{cases}
$$
has $f_x(0,0) = 0$ and $f_y(0,0) = 0$, but is not continuous at the origin (the limit along $y = x$ is $1/2$, not $0$). Partial derivatives exist; the function is not even continuous — let alone differentiable.

For this course the pathological case never arises: all functions we encounter are compositions of polynomials, exponentials, and logarithms, and their partial derivatives are continuous. Continuous partial derivatives imply total differentiability (the theorem goes: if all first-order partial derivatives exist and are continuous in an open set, then $f$ is differentiable there).

### 1.2 The Total Derivative

**Definition.** $f: D \subseteq \mathbb{R}^n \to \mathbb{R}$ is **differentiable** at $\boldsymbol{a} \in D$ if there exists a linear map $L: \mathbb{R}^n \to \mathbb{R}$ such that

$$
\lim_{\boldsymbol{h} \to \boldsymbol{0}} \frac{f(\boldsymbol{a} + \boldsymbol{h}) - f(\boldsymbol{a}) - L(\boldsymbol{h})}{\|\boldsymbol{h}\|} = 0.
$$

The linear map $L$ is called the **total derivative** of $f$ at $\boldsymbol{a}$, written $Df(\boldsymbol{a})$. It is unique when it exists.

When $f$ is differentiable at $\boldsymbol{a}$, the total derivative is represented by the **gradient vector**:
$$
L(\boldsymbol{h}) = \nabla f(\boldsymbol{a}) \cdot \boldsymbol{h} = \sum_{i=1}^{n} \frac{\partial f}{\partial x_i}(\boldsymbol{a})\, h_i.
$$

This is the content of Module 03 stated in advance: the gradient is the vector representation of the total derivative.

## 2. Tangent Planes

### 2.1 Equation of the Tangent Plane

For $f: \mathbb{R}^2 \to \mathbb{R}$, the total derivative at $(a, b)$ defines a tangent plane to the surface $z = f(x,y)$. The plane passes through the point $(a, b, f(a,b))$ with the slopes in the $x$- and $y$-directions given by $f_x(a,b)$ and $f_y(a,b)$:

$$
z = f(a, b) + f_x(a, b)(x - a) + f_y(a, b)(y - b).
$$

This is the **linearisation** $L(x,y)$ of $f$ at $(a,b)$.

**Worked example.** Let $f(x,y) = x^2 + y^2$ and find the tangent plane at $(1, 2)$.

$f(1,2) = 5$, $f_x = 2x \Rightarrow f_x(1,2) = 2$, $f_y = 2y \Rightarrow f_y(1,2) = 4$.

Tangent plane: $z = 5 + 2(x-1) + 4(y-2) = 2x + 4y - 5$.

### 2.2 The Total Differential

The **total differential** of $f$ is the expression

$$
df = \frac{\partial f}{\partial x}\,dx + \frac{\partial f}{\partial y}\,dy
$$

where $dx$ and $dy$ are infinitesimal increments in $x$ and $y$. More concretely, for small finite changes $\Delta x$ and $\Delta y$:

$$
\Delta f \approx f_x(a,b)\,\Delta x + f_y(a,b)\,\Delta y.
$$

This is the linear approximation to $\Delta f = f(a + \Delta x, b + \Delta y) - f(a,b)$. The approximation improves as $\|\Delta\boldsymbol{x}\| \to 0$.

In $n$ dimensions:
$$
df = \sum_{i=1}^n \frac{\partial f}{\partial x_i}\,dx_i = \nabla f \cdot d\boldsymbol{x}.
$$

### 2.3 Error Propagation

The total differential has an immediate practical application: **error propagation**. If $f$ is a quantity computed from measurements $x_1, \ldots, x_n$ with small errors $\delta x_i$, then the resulting error in $f$ is approximately

$$
\delta f \approx \sum_{i=1}^n \left|\frac{\partial f}{\partial x_i}\right| |\delta x_i|.
$$

Variables with large partial derivatives contribute most to the total error — they are the sensitive inputs.

## 3. Second-Order Taylor Expansion

The tangent plane is the first-order (linear) approximation. The second-order approximation adds a curvature term:

$$
f(\boldsymbol{a} + \boldsymbol{h}) \approx f(\boldsymbol{a}) + \nabla f(\boldsymbol{a})^{\top}\boldsymbol{h} + \frac{1}{2}\boldsymbol{h}^{\top}\boldsymbol{H}f(\boldsymbol{a})\boldsymbol{h} + O(\|\boldsymbol{h}\|^3),
$$

where $\boldsymbol{H}f(\boldsymbol{a})$ is the Hessian matrix at $\boldsymbol{a}$.

At a critical point $\boldsymbol{a}$ where $\nabla f(\boldsymbol{a}) = \boldsymbol{0}$, the linear term vanishes and the quadratic term dominates:

$$
f(\boldsymbol{a} + \boldsymbol{h}) \approx f(\boldsymbol{a}) + \frac{1}{2}\boldsymbol{h}^{\top}\boldsymbol{H}f(\boldsymbol{a})\boldsymbol{h}.
$$

If $\boldsymbol{H}f(\boldsymbol{a})$ is negative definite, then $\frac{1}{2}\boldsymbol{h}^{\top}\boldsymbol{H}f\boldsymbol{h} < 0$ for all $\boldsymbol{h} \neq \boldsymbol{0}$, confirming $\boldsymbol{a}$ is a local maximum. This is the multivariable second-derivative test, developed in Module 04.

## 4. Statistical Bridge — The Laplace Approximation [MLE, BAY]

### 4.1 Quadratic Approximation of the Log-Likelihood

At the MLE $\hat{\boldsymbol{\theta}}$, the gradient of $\ell$ is zero. The second-order Taylor expansion around the MLE is:

$$
\ell(\hat{\boldsymbol{\theta}} + \boldsymbol{h}) \approx \ell(\hat{\boldsymbol{\theta}}) + \frac{1}{2}\boldsymbol{h}^{\top}\boldsymbol{H}\ell(\hat{\boldsymbol{\theta}})\boldsymbol{h}.
$$

Since $\boldsymbol{H}\ell(\hat{\boldsymbol{\theta}}) = -\hat{\boldsymbol{I}}$ (the negative observed Fisher information), this becomes:

$$
\ell(\hat{\boldsymbol{\theta}} + \boldsymbol{h}) \approx \ell(\hat{\boldsymbol{\theta}}) - \frac{1}{2}\boldsymbol{h}^{\top}\hat{\boldsymbol{I}}\,\boldsymbol{h}.
$$

Exponentiating gives the **likelihood approximation**:

$$
L(\hat{\boldsymbol{\theta}} + \boldsymbol{h}) \propto \exp\!\left(-\frac{1}{2}\boldsymbol{h}^{\top}\hat{\boldsymbol{I}}\,\boldsymbol{h}\right).
$$

This is a **Gaussian** in $\boldsymbol{h}$ with covariance matrix $\hat{\boldsymbol{I}}^{-1}$. Near the MLE, the likelihood surface is approximately a multivariate Gaussian.

### 4.2 The Laplace Approximation

In Bayesian inference, the posterior is $p(\boldsymbol{\theta} \mid \boldsymbol{x}) \propto L(\boldsymbol{\theta})\,p(\boldsymbol{\theta})$. With a flat prior, the posterior is proportional to the likelihood. The second-order Taylor expansion above gives:

$$
p(\boldsymbol{\theta} \mid \boldsymbol{x}) \approx \mathcal{N}\!\left(\hat{\boldsymbol{\theta}},\, \hat{\boldsymbol{I}}^{-1}\right),
$$

a multivariate Gaussian centred at the MLE with covariance $\hat{\boldsymbol{I}}^{-1}$. This is the **Laplace approximation** — one of the workhorses of approximate Bayesian inference.

The quality of the approximation depends on how well the quadratic term captures the shape of $\ell$. For large samples (where the posterior concentrates near the MLE), the Laplace approximation is excellent. For small samples or multi-modal posteriors, it can fail badly.

> **Why tangent planes matter here.** The tangent plane is the linear (first-order) approximation to the log-likelihood surface. Adding the quadratic term (the Hessian) produces the Laplace approximation. The entire sequence — from partial derivatives to the Hessian to the Laplace approximation — flows from the total derivative definition in Section 1.

## Section 1 — Tangent Planes

```python
# Surface: f(x,y) = sin(sqrt(x^2+y^2)) / sqrt(x^2+y^2)   (sinc-like surface)
# Partial derivatives computed analytically
def f_sinc(xv, yv):
    r = np.sqrt(xv**2 + yv**2)
    return np.where(r == 0, 1.0, np.sin(r) / r)

def df_dx_sinc(xv, yv):
    r = np.sqrt(xv**2 + yv**2)
    # d/dx [sin(r)/r] = (r*cos(r) - sin(r))/r^2 * x/r
    return np.where(r == 0, 0.0,
                    (r * np.cos(r) - np.sin(r)) / r**3 * xv)

def df_dy_sinc(xv, yv):
    r = np.sqrt(xv**2 + yv**2)
    return np.where(r == 0, 0.0,
                    (r * np.cos(r) - np.sin(r)) / r**3 * yv)

# Evaluation point
x0, y0 = 1.5, 1.0
f0      = f_sinc(x0, y0)
fx0     = df_dx_sinc(x0, y0)
fy0     = df_dy_sinc(x0, y0)

# Grids
lim = 3.0
xg  = np.linspace(-lim, lim, 200)
yg  = np.linspace(-lim, lim, 200)
XG, YG = np.meshgrid(xg, yg)
ZG  = f_sinc(XG, YG)

# Tangent plane: z = f0 + fx0*(x - x0) + fy0*(y - y0)
TP = f0 + fx0 * (XG - x0) + fy0 * (YG - y0)
# Clip to visible range for plotting
TP_clipped = np.clip(TP, ZG.min() - 0.3, ZG.max() + 0.3)

fig = plt.figure(figsize=(16, 5))

# Left: surface + tangent plane
ax1 = fig.add_subplot(131, projection='3d')
ax1.plot_surface(XG, YG, ZG,  cmap='coolwarm', alpha=0.55, linewidth=0)
ax1.plot_surface(XG, YG, TP_clipped, color=TS_BLUE, alpha=0.35, linewidth=0)
ax1.scatter([x0], [y0], [f0], color=TS_AMBER, s=80, zorder=6)
ax1.set_title('Surface + tangent plane', fontweight='bold', color=TS_GRAY)
ax1.set_xlabel('$x$'); ax1.set_ylabel('$y$'); ax1.set_zlabel('$f$')
ax1.tick_params(labelsize=8)

# Middle: approximation error |f - tangent plane|
ax2 = fig.add_subplot(132)
err = np.abs(ZG - TP)
cf = ax2.contourf(XG, YG, err, levels=20, cmap='Oranges')
plt.colorbar(cf, ax=ax2, label='$|f - L|$')
ax2.scatter([x0], [y0], color=TS_AMBER, s=80, zorder=5, label='Tangent point')
ax2.set_title('Linearisation error $|f - L|$\n(zero at tangent point)',
              fontweight='bold', color=TS_GRAY)
ax2.set_xlabel('$x$'); ax2.set_ylabel('$y$'); ax2.set_aspect('equal')
ax2.legend(fontsize=9)

# Right: cross-section comparison
ax3 = fig.add_subplot(133)
xslice = np.linspace(-lim, lim, 300)
f_slice = f_sinc(xslice, y0)
L_slice = f0 + fx0 * (xslice - x0)
ax3.plot(xslice, f_slice, color=TS_BLUE, lw=2, label=r'$f(x, y_0)$')
ax3.plot(xslice, L_slice, color=TS_AMBER, lw=1.8, ls='--',
         label=r'Tangent line $L(x, y_0)$')
ax3.scatter([x0], [f0], color=TS_GREEN, s=70, zorder=5)
ax3.set_xlabel('$x$'); ax3.set_ylabel('$f$')
ax3.set_title(r'Cross-section at $y = y_0$', fontweight='bold', color=TS_GRAY)
ax3.legend(fontsize=9)

plt.suptitle('Module 02, Unit 03 — Tangent Plane to a Surface',
             fontsize=13, fontweight='bold', color=TS_GRAY)
plt.tight_layout()
plt.savefig('module-02-unit-03-tangent.png', dpi=150, bbox_inches='tight')
plt.show()
print(f'At ({x0},{y0}): f={f0:.4f},  df/dx={fx0:.4f},  df/dy={fy0:.4f}')
```

### What Do You See?

1. The tangent plane (blue sheet) touches the surface exactly at the orange dot and closely approximates the surface for small distances from that point. The approximation error (middle panel) is smallest right at the tangent point and grows with distance — especially where the surface curves sharply.

2. The cross-section (right panel) shows the tangent line to the curve $f(x, y_0)$ — this is the partial derivative $f_x$ as a slope, exactly as in Unit 02-01. The tangent plane is the 2D version of this tangent line.

## Section 2 — First vs. Second-Order Approximation

We compare the linear (tangent plane) and quadratic (second-order Taylor) approximations to the Gaussian bell $f(x,y) = e^{-(x^2+y^2)/2}$ at the origin.

```python
# f(x,y) = exp(-(x^2+y^2)/2) — expand around (a,b) = (0.8, 0.6)
a, b = 0.8, 0.6

def f_bell(xv, yv):
    return np.exp(-(xv**2 + yv**2) / 2)

# Analytical derivatives
f0b   = f_bell(a, b)
fx0b  = -a * f_bell(a, b)
fy0b  = -b * f_bell(a, b)
fxx0b = (a**2 - 1) * f_bell(a, b)
fyy0b = (b**2 - 1) * f_bell(a, b)
fxy0b = a * b * f_bell(a, b)

xg2 = np.linspace(-2.5, 2.5, 300)
yg2 = np.linspace(-2.5, 2.5, 300)
X2, Y2 = np.meshgrid(xg2, yg2)

Z_true  = f_bell(X2, Y2)
dx, dy  = X2 - a, Y2 - b
Z_lin   = f0b + fx0b * dx + fy0b * dy
Z_quad  = Z_lin + 0.5 * (fxx0b * dx**2 + 2 * fxy0b * dx * dy + fyy0b * dy**2)

fig, axes = plt.subplots(1, 3, figsize=(17, 5))
cmaps = ['Greens', 'Blues', 'Purples']
titles = ['True surface\n$f = e^{-(x^2+y^2)/2}$',
          'Linear approx (tangent plane)',
          'Quadratic approx (2nd-order Taylor)']
Zs = [Z_true, Z_lin, Z_quad]

for ax, Z, cmap, title in zip(axes, Zs, cmaps, titles):
    cf = ax.contourf(X2, Y2, Z, levels=15, cmap=cmap)
    ax.contour(X2, Y2, Z, levels=15, colors='k', linewidths=0.3, alpha=0.4)
    plt.colorbar(cf, ax=ax)
    ax.scatter([a], [b], color=TS_AMBER, s=80, zorder=5,
               label=f'Expansion point ({a},{b})')
    ax.set_xlabel('$x$'); ax.set_ylabel('$y$')
    ax.set_title(title, fontweight='bold', color=TS_GRAY)
    ax.set_aspect('equal')
    ax.legend(fontsize=8)

plt.suptitle('Module 02, Unit 03 — Linear vs. Quadratic Taylor Approximation',
             fontsize=13, fontweight='bold', color=TS_GRAY)
plt.tight_layout()
plt.savefig('module-02-unit-03-taylor.png', dpi=150, bbox_inches='tight')
plt.show()
```

### What Do You See?

1. The linear approximation (middle) produces straight, parallel contours — it cannot capture the curved shape of the bell. It is accurate only very close to the expansion point.

2. The quadratic approximation (right) shows curved, ellipse-like contours — much closer to the true surface. How far from the expansion point do the quadratic contours match the true ones? Where do they diverge?

## Section 3 — Statistical Bridge: The Laplace Approximation [MLE, BAY]

```python
rng    = np.random.default_rng(seed=42)
n_data = 30
data   = rng.normal(loc=2.0, scale=1.0, size=n_data)
x_bar  = data.mean()
s2_hat = data.var()

# True log-likelihood
mu_vals = np.linspace(x_bar - 1.2, x_bar + 1.2, 400)
s2_vals = np.linspace(max(0.1, s2_hat - 0.9), s2_hat + 1.3, 400)
MU, S2  = np.meshgrid(mu_vals, s2_vals)

resid_sq = np.sum((data[:, None, None] - MU[None, :, :])**2, axis=0)
LOGLIK   = -n_data / 2 * np.log(2*np.pi*S2) - resid_sq / (2*S2)
ell_max  = LOGLIK.max()

# Laplace (quadratic) approximation around the MLE
# log q(theta) = ell_max - 0.5*(theta - theta_hat)^T I_hat (theta - theta_hat)
# For the Gaussian: I_hat is diagonal with (n/s2_hat, n/(2*s2_hat^2))
I_mu = n_data / s2_hat
I_s2 = n_data / (2 * s2_hat**2)

LAPLACE  = ell_max - 0.5 * I_mu * (MU - x_bar)**2 - 0.5 * I_s2 * (S2 - s2_hat)**2

# Plot: true log-likelihood vs Laplace approximation
levels = np.linspace(ell_max - 15, ell_max, 20)

fig, axes = plt.subplots(1, 2, figsize=(15, 6))

for ax, Z, title in zip(axes,
    [LOGLIK, LAPLACE],
    ['True log-likelihood $\\ell(\\mu,\\sigma^2)$',
     'Laplace (quadratic) approximation']):
    cf = ax.contourf(MU, S2, Z, levels=levels, cmap='Greens')
    plt.colorbar(cf, ax=ax, label='$\\ell$ or approximation')
    ax.contour(MU, S2, Z, levels=levels, colors='k', linewidths=0.4, alpha=0.5)
    ax.scatter([x_bar], [s2_hat], color=TS_AMBER, s=120, marker='*', zorder=6,
               label=f'MLE $({x_bar:.2f}, {s2_hat:.2f})$')
    ax.set_xlabel('$\\mu$'); ax.set_ylabel('$\\sigma^2$')
    ax.set_title(title, fontweight='bold', color=TS_GRAY)
    ax.legend(fontsize=9)

plt.suptitle('Module 02, Unit 03 — Laplace Approximation to the Posterior [MLE, BAY]',
             fontsize=13, fontweight='bold', color=TS_GRAY)
plt.tight_layout()
plt.savefig('module-02-unit-03-laplace.png', dpi=150, bbox_inches='tight')
plt.show()

# Numerical comparison at a few points
test_pts = [(x_bar + 0.2, s2_hat), (x_bar, s2_hat + 0.3), (x_bar + 0.4, s2_hat + 0.4)]
print('Point-wise comparison: true ell vs Laplace approximation')
print(f'{"Point":35s}  {"True ell":>12s}  {"Laplace":>12s}  {"Error":>10s}')
for (m, s) in test_pts:
    true_val   = (-n_data/2*np.log(2*np.pi*s)
                  - np.sum((data-m)**2)/(2*s))
    laplace_val = ell_max - 0.5*I_mu*(m-x_bar)**2 - 0.5*I_s2*(s-s2_hat)**2
    print(f'mu={m:.2f}, s2={s:.2f}  {true_val:12.4f}  {laplace_val:12.4f}  {abs(true_val-laplace_val):10.4f}')
```

### What Do You See?

1. The Laplace approximation (right panel) has elliptical contours — it is a perfect quadratic surface. The true log-likelihood (left) is not perfectly quadratic, but close to it near the MLE.

2. The numerical comparison shows that the Laplace error is small close to the MLE and grows as you move away. For $n = 30$, how large is the error at the $0.4$-unit-radius test point? What would you expect as $n$ increases?

3. The Laplace approximation contours are axis-aligned ellipses (no tilt) because the observed Fisher information matrix is diagonal for the Gaussian. For the Gamma distribution (extension challenge in Unit 02-02), the approximation ellipses would be tilted. What would cause the tilt?

## Summary

| Concept | Key result |
|---|---|
| Total differentiability | Stronger than existence of partial derivatives; requires linear approximation error $o(\|\boldsymbol{h}\|)$ |
| Continuous partials | Sufficient condition for total differentiability (holds for all course functions) |
| Tangent plane | $z = f(a,b) + f_x(a,b)(x-a) + f_y(a,b)(y-b)$ — best linear approximation at $(a,b)$ |
| Total differential | $df = \sum_i \frac{\partial f}{\partial x_i}\,dx_i = \nabla f \cdot d\boldsymbol{x}$ |
| 2nd-order Taylor | $f(\boldsymbol{a}+\boldsymbol{h}) \approx f(\boldsymbol{a}) + \nabla f^\top\boldsymbol{h} + \frac{1}{2}\boldsymbol{h}^\top\boldsymbol{H}f\,\boldsymbol{h}$ |
| At a critical point | Linear term vanishes; quadratic term $\frac{1}{2}\boldsymbol{h}^\top\boldsymbol{H}f\,\boldsymbol{h}$ determines local shape |
| Laplace approximation | $p(\boldsymbol{\theta}\mid\boldsymbol{x}) \approx \mathcal{N}(\hat{\boldsymbol{\theta}},\,\hat{\boldsymbol{I}}^{-1})$ — second-order Taylor of the log-posterior |
