## Why This Unit

First-order partials tell you the *slope* of a surface in each coordinate direction. Second-order partials tell you the *curvature* — how fast the slope is changing. Curvature determines whether a critical point is a minimum, maximum, or saddle (Module 04), and it determines the width of the likelihood surface around the MLE (confidence regions, Module 01-02).

The Hessian matrix — assembled from all second-order partials — is the multivariable analogue of the second derivative. It plays a central role in optimization, in the Laplace approximation to Bayesian posteriors, and in the Fisher information matrix that governs the precision of MLEs.

## 1. Second-Order Partial Derivatives

### 1.1 Definition and Notation

Given $f: \mathbb{R}^n \to \mathbb{R}$ with first-order partials $\frac{\partial f}{\partial x_i}$, the **second-order partial derivatives** are obtained by differentiating again:

$$
\frac{\partial^2 f}{\partial x_i^2} = \frac{\partial}{\partial x_i}\left(\frac{\partial f}{\partial x_i}\right) \qquad \text{(pure second partial)}
$$

$$
\frac{\partial^2 f}{\partial x_j \, \partial x_i} = \frac{\partial}{\partial x_j}\left(\frac{\partial f}{\partial x_i}\right) \qquad \text{(mixed partial, differentiate } x_i \text{ first, then } x_j\text{)}
$$

> **Order of differentiation warning.** In the notation $\frac{\partial^2 f}{\partial x_j \, \partial x_i}$, the rightmost variable is differentiated *first*. So $\frac{\partial^2 f}{\partial y\, \partial x}$ means: differentiate with respect to $x$ first, then with respect to $y$. This is also written $f_{xy}$ in subscript notation — subscripts are read left-to-right in the *same* order as differentiation. Some texts reverse one of these conventions; always check which convention is in use.

**Shorthand:** $f_{xx} = \frac{\partial^2 f}{\partial x^2}$, $\;f_{xy} = \frac{\partial^2 f}{\partial y\,\partial x}$, $\;f_{yx} = \frac{\partial^2 f}{\partial x\,\partial y}$.

For a function of $n$ variables, there are $n^2$ second-order partials: $n$ pure second partials and $n(n-1)$ mixed partials (of which $n(n-1)/2$ pairs may or may not be equal).

### 1.2 Worked Examples

**Example 1.** Let $f(x,y) = x^3 y^2 + e^{xy}$.

First-order partials:
$$
f_x = 3x^2 y^2 + y\,e^{xy}, \qquad f_y = 2x^3 y + x\,e^{xy}.
$$

Pure second partials:
$$
f_{xx} = 6x y^2 + y^2\,e^{xy}, \qquad f_{yy} = 2x^3 + x^2\,e^{xy}.
$$

Mixed partials (differentiate $f_x$ with respect to $y$, and $f_y$ with respect to $x$):
$$
f_{xy} = \frac{\partial}{\partial y}(3x^2 y^2 + y\,e^{xy}) = 6x^2 y + e^{xy} + xy\,e^{xy}.
$$
$$
f_{yx} = \frac{\partial}{\partial x}(2x^3 y + x\,e^{xy}) = 6x^2 y + e^{xy} + xy\,e^{xy}.
$$

The mixed partials are equal: $f_{xy} = f_{yx}$. This is not a coincidence.

**Example 2.** Let $f(x,y) = \log(x^2 + y^2)$ (defined away from the origin).

$$
f_x = \frac{2x}{x^2+y^2}, \qquad f_y = \frac{2y}{x^2+y^2}.
$$

$$
f_{xx} = \frac{2(x^2+y^2) - 2x \cdot 2x}{(x^2+y^2)^2} = \frac{2y^2 - 2x^2}{(x^2+y^2)^2}.
$$

$$
f_{yy} = \frac{2x^2 - 2y^2}{(x^2+y^2)^2}.
$$

Note that $f_{xx} + f_{yy} = 0$ everywhere. Functions satisfying this identity are called **harmonic**; $\log(x^2+y^2)$ is a classic harmonic function outside the origin.

## 2. Clairaut's Theorem (Symmetry of Mixed Partials)

**Theorem (Clairaut / Schwarz).** If $f: D \subseteq \mathbb{R}^n \to \mathbb{R}$ and the mixed partial derivatives $f_{x_i x_j}$ and $f_{x_j x_i}$ are both **continuous** on an open set containing $\boldsymbol{a}$, then

$$
\frac{\partial^2 f}{\partial x_j \, \partial x_i}(\boldsymbol{a}) = \frac{\partial^2 f}{\partial x_i \, \partial x_j}(\boldsymbol{a}).
$$

The order of differentiation does not matter, provided the mixed partials are continuous.

**Practical significance.** For virtually every function encountered in statistics and machine learning — compositions of polynomials, exponentials, and logarithms — the mixed partials are continuous everywhere in the domain. Clairaut's theorem therefore applies universally in this course, and we will use it without further comment.

**Why it matters for the Hessian.** The second-order partial derivatives of $f$ are assembled into the **Hessian matrix**:

$$
\boldsymbol{H}f(\boldsymbol{x}) = \begin{pmatrix}
\frac{\partial^2 f}{\partial x_1^2} & \frac{\partial^2 f}{\partial x_2\,\partial x_1} & \cdots & \frac{\partial^2 f}{\partial x_n\,\partial x_1} \\
\frac{\partial^2 f}{\partial x_1\,\partial x_2} & \frac{\partial^2 f}{\partial x_2^2} & \cdots & \frac{\partial^2 f}{\partial x_n\,\partial x_2} \\
\vdots & & \ddots & \vdots \\
\frac{\partial^2 f}{\partial x_1\,\partial x_n} & \cdots & \cdots & \frac{\partial^2 f}{\partial x_n^2}
\end{pmatrix}.
$$

Clairaut's theorem guarantees that $\boldsymbol{H}f$ is a **symmetric matrix**: $(\boldsymbol{H}f)_{ij} = (\boldsymbol{H}f)_{ji}$. This symmetry is not merely algebraic tidiness — it means the Hessian can be diagonalised by an orthonormal basis (spectral theorem), its eigenvalues are real, and the second-derivative test for extrema has a clean statement in terms of definiteness (Module 04).

## 3. Higher-Order Partials

Differentiation can be iterated to any order. Third-order partials arise in Taylor expansions (useful for error analysis), and the pattern $f_{x_i x_j x_k} = f_{x_j x_i x_k} = \cdots$ holds whenever all the relevant partials are continuous (an iterated application of Clairaut's theorem).

In this course, third- and higher-order partials appear only implicitly — inside the remainder of a second-order Taylor expansion (Unit 02-03). For optimization and inference, the Hessian (second order) is the key object.

**Example.** For $f(x,y) = x^4 + x^2 y^2 + y^4$:
$$
f_x = 4x^3 + 2xy^2, \quad f_{xx} = 12x^2 + 2y^2, \quad f_{xxx} = 24x.
$$
$$
f_{xy} = 4xy, \quad f_{xxy} = 4y, \quad f_{xyy} = 4x.
$$

Note $f_{xxy} \neq f_{xyy}$ in general (they equal different things here), but $f_{xxy} = f_{yxx} = f_{xyx}$ by Clairaut.

## 4. Statistical Bridge — Second-Order Partials of the Log-Likelihood & the Fisher Information Matrix [MLE]

### 4.1 Second-Order Partials of $\ell(\mu, \sigma^2)$

Continuing from Unit 02-01, we compute the second-order partials of the Gaussian log-likelihood. From the first-order results:

$$
\frac{\partial \ell}{\partial \mu} = \frac{n(\bar{x} - \mu)}{\sigma^2}, \qquad \frac{\partial \ell}{\partial \sigma^2} = -\frac{n}{2\sigma^2} + \frac{\sum_i(x_i-\mu)^2}{2\sigma^4}.
$$

**Pure second partial w.r.t. $\mu$:**
$$
\frac{\partial^2 \ell}{\partial \mu^2} = -\frac{n}{\sigma^2}.
$$
This is **negative** everywhere — the log-likelihood is strictly concave in $\mu$ for any fixed $\sigma^2 > 0$. This guarantees the MLE $\hat{\mu}$ is a global maximum.

**Pure second partial w.r.t. $\sigma^2$** (writing $s = \sigma^2$):
$$
\frac{\partial^2 \ell}{\partial s^2} = \frac{n}{2s^2} - \frac{\sum_i(x_i-\mu)^2}{s^3}.
$$
At the MLE $\hat{s} = \frac{1}{n}\sum_i(x_i - \bar{x})^2$, this simplifies to $-\frac{n}{2\hat{s}^2}$, which is also negative — confirming the MLE is a local maximum in the $\sigma^2$-direction as well.

**Mixed partial:**
$$
\frac{\partial^2 \ell}{\partial \sigma^2\,\partial \mu} = \frac{\partial}{\partial \sigma^2}\left(\frac{n(\bar{x}-\mu)}{\sigma^2}\right) = -\frac{n(\bar{x}-\mu)}{\sigma^4}.
$$
At the MLE $\mu = \bar{x}$, this is **zero**. This confirms the orthogonality of $\mu$ and $\sigma^2$ we noted in Unit 02-01: at the MLE, the two parameters are uncorrelated in the second-order sense.

### 4.2 The Observed Fisher Information Matrix

The **observed Fisher information matrix** is the negative Hessian of the log-likelihood evaluated at the MLE:

$$
\hat{\boldsymbol{I}} = -\boldsymbol{H}\ell(\hat{\boldsymbol{\theta}}) = -\begin{pmatrix}
\frac{\partial^2 \ell}{\partial \mu^2} & \frac{\partial^2 \ell}{\partial \sigma^2 \partial \mu} \\
\frac{\partial^2 \ell}{\partial \mu \partial \sigma^2} & \frac{\partial^2 \ell}{\partial (\sigma^2)^2}
\end{pmatrix}_{\boldsymbol{\theta} = \hat{\boldsymbol{\theta}}} = \begin{pmatrix}
\frac{n}{\hat{\sigma}^2} & 0 \\
0 & \frac{n}{2\hat{\sigma}^4}
\end{pmatrix}.
$$

The off-diagonal zeros confirm parameter orthogonality. The inverse of $\hat{\boldsymbol{I}}$ is the **asymptotic covariance matrix of the MLE**:

$$
\hat{\boldsymbol{I}}^{-1} = \begin{pmatrix}
\frac{\hat{\sigma}^2}{n} & 0 \\
0 & \frac{2\hat{\sigma}^4}{n}
\end{pmatrix}.
$$

The $(1,1)$ entry $\hat{\sigma}^2/n$ is the variance of the sample mean — a result familiar from introductory statistics, now derived from the Hessian of the log-likelihood.

Clairaut's theorem guarantees $\hat{\boldsymbol{I}}$ is symmetric, which in turn ensures the asymptotic covariance matrix is symmetric and positive semi-definite — a requirement for it to be interpretable as a covariance matrix.

> **Forward link.** In Module 04 we will use the Hessian to classify critical points via the second-derivative test. The condition for a local maximum is that $-\boldsymbol{H}\ell$ is positive definite — exactly what the observed Fisher information matrix is at the MLE for a well-specified model.

## Section 1 — Symbolic Second-Order Partials & Clairaut Verification

```python
x, y = sp.symbols('x y', real=True)
mu, s, n_sym = sp.symbols('mu s n', real=True, positive=True)

# --- Example 1: f = x^3*y^2 + exp(x*y) ---
f1 = x**3 * y**2 + sp.exp(x*y)
f1_xy = sp.diff(f1, x, y)   # differentiate x first, then y
f1_yx = sp.diff(f1, y, x)   # differentiate y first, then x
print('f1 = x^3*y^2 + exp(x*y)')
print('  f_xy =', sp.simplify(f1_xy))
print('  f_yx =', sp.simplify(f1_yx))
print('  Clairaut verified:', sp.simplify(f1_xy - f1_yx) == 0)

# --- Gaussian log-likelihood (symbolic, n observations) ---
# ell(mu, s) = -n/2*log(2*pi) - n/2*log(s) - 1/(2*s)*(ss + n*(mu-xbar)^2)
# For cleanliness, work with a general ss (sum of sq deviations from xbar)
ss, xbar = sp.symbols('ss xbar', real=True, positive=True)
ell = (-n_sym/2 * sp.log(2*sp.pi)
       - n_sym/2 * sp.log(s)
       - (ss + n_sym*(mu - xbar)**2) / (2*s))

# Second-order partials
ell_mumu  = sp.diff(ell, mu, mu)
ell_ss2   = sp.diff(ell, s, s)
ell_mus   = sp.diff(ell, mu, s)   # mixed: mu first, then s
ell_smu   = sp.diff(ell, s, mu)   # mixed: s first, then mu

print('\nGaussian log-likelihood second-order partials:')
print('  d2ell/dmu2  =', sp.simplify(ell_mumu), '  (expected: -n/s)')
print('  d2ell/ds2   =', sp.simplify(ell_ss2))
print('  d2ell/d(mu)d(s) =', sp.simplify(ell_mus))
print('  d2ell/d(s)d(mu) =', sp.simplify(ell_smu))
print('  Clairaut verified:', sp.simplify(ell_mus - ell_smu) == 0)
print('  Mixed partial at MLE (mu=xbar):', sp.simplify(ell_mus.subs(mu, xbar)))
```

### What Do You See?

1. SymPy confirms $f_{xy} = f_{yx}$ for both examples — Clairaut's theorem in action. The mixed partials of the Gaussian log-likelihood are also equal.
2. The mixed partial $\frac{\partial^2\ell}{\partial \sigma^2\,\partial \mu}$ vanishes when $\mu = \bar{x}$. This is the orthogonality result: at the MLE, the curvature of the log-likelihood surface is described by a *diagonal* Hessian. What does a diagonal Hessian imply about the shape of the log-likelihood contours near the MLE?

## Section 2 — Hessian Structure Visualised

We visualise the four entries of the Hessian of $\ell(\mu, \sigma^2)$ as functions over parameter space. Each entry is itself a scalar field. The off-diagonal entries (mixed partials) should be zero at the MLE.

```python
rng    = np.random.default_rng(seed=42)
n_data = 30
data   = rng.normal(loc=2.0, scale=1.0, size=n_data)
x_bar  = data.mean()
s2_hat = data.var()

# Parameter grids
mu_vals = np.linspace(x_bar - 1.0, x_bar + 1.0, 200)
s2_vals = np.linspace(max(0.1, s2_hat - 0.7), s2_hat + 1.0, 200)
MU, S2  = np.meshgrid(mu_vals, s2_vals)

# Hessian entries (derived analytically in Section 4)
SS_grid = np.sum((data[:, None, None] - MU[None, :, :])**2, axis=0)

H_mumu = -n_data / S2                                      # d2ell/dmu2
H_s2s2 = n_data / (2*S2**2) - SS_grid / S2**3              # d2ell/ds2^2
H_mus2 = -(n_data * (x_bar - MU)) / S2**2                  # mixed partial
# Note: H_s2mu = H_mus2 by Clairaut

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

entries = [
    (H_mumu, r'$\partial^2\ell / \partial\mu^2$',
     'Always negative (concave in $\\mu$)', 'RdBu_r'),
    (H_mus2, r'$\partial^2\ell / \partial\sigma^2\,\partial\mu$',
     'Mixed partial — zero at MLE', 'RdBu_r'),
    (H_mus2, r'$\partial^2\ell / \partial\mu\,\partial\sigma^2$',
     'Equal to entry above (Clairaut)', 'RdBu_r'),
    (H_s2s2, r'$\partial^2\ell / \partial(\sigma^2)^2$',
     'Negative at MLE (concave in $\\sigma^2$)', 'RdBu_r'),
]

for ax, (Z, title, note, cmap) in zip(axes.flat, entries):
    vmax = np.percentile(np.abs(Z), 97)
    cf = ax.contourf(MU, S2, Z, levels=20, cmap=cmap,
                     vmin=-vmax, vmax=vmax)
    plt.colorbar(cf, ax=ax)
    ax.scatter([x_bar], [s2_hat], color=TS_AMBER, s=100, marker='*',
               zorder=5, label='MLE')
    ax.set_xlabel('$\\mu$'); ax.set_ylabel('$\\sigma^2$')
    ax.set_title(title + '\n' + note, fontweight='bold', color=TS_GRAY)
    ax.legend(fontsize=9)

plt.suptitle('Module 02, Unit 02 — Hessian Entries of the Gaussian Log-Likelihood',
             fontsize=13, fontweight='bold', color=TS_GRAY)
plt.tight_layout()
plt.savefig('module-02-unit-02-hessian.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Section 3 — Statistical Bridge: Observed Fisher Information Matrix [MLE]

```python
# Observed Fisher information: I_hat = -H_ell evaluated at the MLE
I_mumu = n_data / s2_hat
I_s2s2 = n_data / (2 * s2_hat**2)
I_mus2 = 0.0  # orthogonal at MLE

I_hat = np.array([[I_mumu, I_mus2],
                  [I_mus2, I_s2s2]])

# Asymptotic covariance = I_hat^{-1}
cov_hat = np.linalg.inv(I_hat)

print('=== Observed Fisher Information Matrix ===')
print(f'  I_hat =\n    [[ {I_hat[0,0]:.4f}  {I_hat[0,1]:.4f} ]')
print(f'           [ {I_hat[1,0]:.4f}  {I_hat[1,1]:.4f} ]]')
print(f'\n  Diagonal? {np.allclose(I_hat, np.diag(np.diag(I_hat)))}')
print(f'  Symmetric? {np.allclose(I_hat, I_hat.T)}')
print(f'  Positive definite? eigenvalues = {np.linalg.eigvalsh(I_hat).round(4)}')

print('\n=== Asymptotic Covariance Matrix (I_hat^{-1}) ===')
print(f'  Var(mu_hat)    = sigma^2/n = {cov_hat[0,0]:.6f}')
print(f'  Var(s2_hat)    = 2*sigma^4/n = {cov_hat[1,1]:.6f}')
print(f'  Cov(mu, s2)    = {cov_hat[0,1]:.6f}  (zero: orthogonal params)')
print(f'\n  Theoretical Var(mu_hat) = sigma^2/n = {s2_hat/n_data:.6f}  [matches]')
print(f'  Theoretical Var(s2_hat) = 2*sigma^4/n = {2*s2_hat**2/n_data:.6f}  [matches]')
```

## Summary

| Concept | Key result |
|---|---|
| Second-order partial | $f_{x_i x_j} = \frac{\partial}{\partial x_j}\left(\frac{\partial f}{\partial x_i}\right)$ — curvature in the $(x_i, x_j)$ directions |
| Clairaut's theorem | $f_{x_i x_j} = f_{x_j x_i}$ whenever both mixed partials are continuous |
| Hessian matrix | Symmetric $n \times n$ matrix of all second-order partials; $(\boldsymbol{H}f)_{ij} = \frac{\partial^2 f}{\partial x_j\,\partial x_i}$ |
| Gaussian $\ell_{\mu\mu}$ | $-n/\sigma^2$ — always negative; log-likelihood is strictly concave in $\mu$ |
| Gaussian mixed partial at MLE | $\ell_{\mu,\sigma^2} = 0$ at $\mu = \bar{x}$ — parameters are orthogonal |
| Observed Fisher info | $\hat{\boldsymbol{I}} = -\boldsymbol{H}\ell(\hat{\boldsymbol{\theta}})$ — negative Hessian at MLE |
| Asymptotic covariance | $\hat{\boldsymbol{I}}^{-1}$ — inverse Fisher info; diagonal entry = asymptotic variance of each MLE |

**Up next**: Unit 02-03 — The Total Derivative & Tangent Planes

Partial derivatives measure the slope in individual coordinate directions. The **total derivative** packages all the partial information into a single linear map that best approximates $f$ near a point — a tangent plane for $f: \mathbb{R}^2 \to \mathbb{R}$, and a tangent hyperplane in higher dimensions. This linear approximation is the multivariable analogue of $f(a+h) \approx f(a) + f'(a)h$, and it is the foundation of the chain rule.
