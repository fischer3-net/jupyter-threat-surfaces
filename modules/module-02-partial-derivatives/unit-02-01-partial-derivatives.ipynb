## Why This Unit

Module 01 gave you the geometry: scalar fields as landscapes, level curves as contours, the gradient as the uphill-pointing perpendicular vector. Module 02 gives you the computation. The partial derivative is how we *calculate* the rate of change of a function in a specific coordinate direction — and from a collection of partial derivatives we will assemble the gradient in Module 03.

The central idea is simple and powerful: to differentiate $f(x_1, \ldots, x_n)$ with respect to $x_i$, hold all other variables constant and differentiate as if $f$ were a function of $x_i$ alone. Every rule from single-variable calculus (Module 00, Unit 03) applies without modification.

---

## 1. Definition

### 1.1 The Limit Definition

Let $f: D \subseteq \mathbb{R}^n \to \mathbb{R}$. The **partial derivative of $f$ with respect to $x_i$** at a point $\boldsymbol{a} = (a_1, \ldots, a_n)^{\top}$ is

$$
\frac{\partial f}{\partial x_i}(\boldsymbol{a}) = \lim_{h \to 0} \frac{f(a_1, \ldots, a_i + h, \ldots, a_n) - f(a_1, \ldots, a_i, \ldots, a_n)}{h},
$$

provided the limit exists. Only the $i$-th argument changes; all others are held at their values in $\boldsymbol{a}$.

This is the ordinary single-variable derivative of the function $g(t) = f(a_1, \ldots, t, \ldots, a_n)$ obtained by freezing all variables except $x_i$ and letting $x_i = t$ vary.

### 1.2 Notation

Four notations appear in the literature; all are standard:

| Notation | Read as |
|---|---|
| $\dfrac{\partial f}{\partial x}$ | "partial $f$ partial $x$" |
| $f_x$ | "$f$ sub $x$" |
| $\partial_x f$ | "del-$x$ $f$" |
| $D_x f$ | "$D$-sub-$x$ $f$" |

This course uses $\frac{\partial f}{\partial x_i}$ as the primary notation and $f_{x_i}$ as a compact alternative. The subscript notation is especially convenient for higher-order partials (Unit 02-02).

### 1.3 Geometric Interpretation

For $f: \mathbb{R}^2 \to \mathbb{R}$, the partial derivative $\frac{\partial f}{\partial x}(a, b)$ is the slope of the curve obtained by intersecting the surface $z = f(x,y)$ with the vertical plane $y = b$. This is a curve in the plane $y = b$, and $\frac{\partial f}{\partial x}(a,b)$ is its slope at $x = a$.

Similarly, $\frac{\partial f}{\partial y}(a, b)$ is the slope of the cross-section obtained by fixing $x = a$ and varying $y$.

The two partial derivatives together give the slope of the surface in the two coordinate directions. The gradient $\nabla f$ — assembled from the partials in Module 03 — gives the direction of steepest ascent.

## 2. Computing Partial Derivatives

The computational rule is: **treat all variables except the differentiation variable as constants, then apply single-variable rules.**

### 2.1 Polynomial and Rational Functions

**Example 1.** Let $f(x,y) = 3x^2 y + y^3 - 5x$.

$$
\frac{\partial f}{\partial x} = 6xy - 5 \qquad (y \text{ treated as constant})
$$
$$
\frac{\partial f}{\partial y} = 3x^2 + 3y^2 \qquad (x \text{ treated as constant})
$$

**Example 2.** Let $f(x,y) = \dfrac{x}{x^2 + y^2}$.

Treating $y$ as constant and applying the quotient rule in $x$:
$$
\frac{\partial f}{\partial x} = \frac{(x^2 + y^2) - x \cdot 2x}{(x^2 + y^2)^2} = \frac{y^2 - x^2}{(x^2 + y^2)^2}.
$$

Treating $x$ as constant and differentiating in $y$:
$$
\frac{\partial f}{\partial y} = \frac{-x \cdot 2y}{(x^2 + y^2)^2} = \frac{-2xy}{(x^2 + y^2)^2}.
$$

### 2.2 Exponential and Logarithmic Functions

**Example 3.** Let $f(x,y) = e^{x^2 + y^2}$.

Chain rule in $x$ (with $y$ constant): the outer function is $e^u$, the inner function is $u = x^2 + y^2$, and $\frac{\partial u}{\partial x} = 2x$:
$$
\frac{\partial f}{\partial x} = e^{x^2 + y^2} \cdot 2x = 2x\,e^{x^2 + y^2}.
$$

By symmetry: $\frac{\partial f}{\partial y} = 2y\,e^{x^2 + y^2}$.

**Example 4 — Gaussian density.** Let
$$
f(x, \mu, \sigma^2) = -\frac{(x - \mu)^2}{2\sigma^2}.
$$

This is the exponent of the Gaussian density. The partial with respect to $\mu$ (treating $x$ and $\sigma^2$ as constants) uses the chain rule on $(x - \mu)^2$:
$$
\frac{\partial f}{\partial \mu} = -\frac{2(x - \mu) \cdot (-1)}{2\sigma^2} = \frac{x - \mu}{\sigma^2}.
$$

The partial with respect to $\sigma^2$ (treating $x$ and $\mu$ as constants, writing $s = \sigma^2$):
$$
\frac{\partial f}{\partial \sigma^2} = \frac{(x - \mu)^2}{2(\sigma^2)^2} = \frac{(x-\mu)^2}{2\sigma^4}.
$$

These are the building blocks for the score of the Gaussian model.

### 2.3 Functions of $n$ Variables

The rule generalises directly. For $f(x_1, \ldots, x_n)$, to find $\frac{\partial f}{\partial x_k}$: treat all $x_j$ with $j \neq k$ as constants and differentiate with respect to $x_k$.

**Example 5.** Let $f(\boldsymbol{x}) = \boldsymbol{x}^{\top}\boldsymbol{A}\boldsymbol{x} = \sum_{i=1}^n \sum_{j=1}^n A_{ij} x_i x_j$ where $\boldsymbol{A}$ is a symmetric matrix.

The partial derivative with respect to $x_k$ picks out all terms containing $x_k$:
$$
\frac{\partial}{\partial x_k}(\boldsymbol{x}^{\top}\boldsymbol{A}\boldsymbol{x}) = 2\sum_{j=1}^n A_{kj} x_j = 2(\boldsymbol{A}\boldsymbol{x})_k.
$$

Stacking over all $k$: $\nabla_x (\boldsymbol{x}^{\top}\boldsymbol{A}\boldsymbol{x}) = 2\boldsymbol{A}\boldsymbol{x}$. This result is used constantly in regression and in the Hessian calculations of Module 04.

## 3. Partial Derivatives of the Gaussian Log-Likelihood [MLE]

We now compute the score components of the Gaussian model — the first partial derivatives of the log-likelihood — formally and carefully.

Given $n$ i.i.d. observations $x_1, \ldots, x_n \overset{\text{iid}}{\sim} \mathcal{N}(\mu, \sigma^2)$, the log-likelihood is

$$
\ell(\mu, \sigma^2) = -\frac{n}{2}\log(2\pi) - \frac{n}{2}\log(\sigma^2) - \frac{1}{2\sigma^2}\sum_{i=1}^{n}(x_i - \mu)^2.
$$

### 3.1 Partial with Respect to $\mu$

Only the last term depends on $\mu$. Apply the chain rule with $u_i = x_i - \mu$:

$$
\frac{\partial \ell}{\partial \mu} = -\frac{1}{2\sigma^2} \sum_{i=1}^{n} \frac{\partial}{ \partial \mu}(x_i - \mu)^2 = -\frac{1}{2\sigma^2} \sum_{i=1}^{n} 2(x_i - \mu)(-1) = \frac{1}{\sigma^2}\sum_{i=1}^{n}(x_i - \mu).
$$

Setting equal to zero: $\sum(x_i - \mu) = 0 \implies \hat{\mu}_{\text{MLE}} = \bar{x}$.

### 3.2 Partial with Respect to $\sigma^2$

Write $s = \sigma^2$ for clarity. Both the second and third terms depend on $s$:

$$
\frac{\partial \ell}{\partial s} = -\frac{n}{2} \cdot \frac{1}{s} + \frac{1}{2s^2}\sum_{i=1}^{n}(x_i - \mu)^2.
$$

Setting equal to zero at $\mu = \bar{x}$:

$$
\frac{n}{2s} = \frac{1}{2s^2}\sum_{i=1}^{n}(x_i - \bar{x})^2 \implies \hat{\sigma}^2_{\text{MLE}} = \frac{1}{n}\sum_{i=1}^{n}(x_i - \bar{x})^2.
$$

Both MLEs follow directly from setting each partial derivative to zero independently — because $\mu$ and $\sigma^2$ are **orthogonal parameters** for the Gaussian: the mixed partial $\frac{\partial^2 \ell}{\partial \mu\, \partial \sigma^2} = 0$. We will verify this in Unit 02-02.

> **Notation checkpoint.** We use $s = \sigma^2$ (not $\sigma$) as the second parameter because $\ell$ is a simpler rational function of $\sigma^2$ than of $\sigma$. This is the standard convention in statistics. Differentiating with respect to $\sigma$ instead introduces a factor of $2\sigma$ via the chain rule and gives the same MLE after simplification — but more algebra.

## Section 1 — Symbolic Partial Derivatives with SymPy

SymPy's `diff(f, var)` computes partial derivatives by treating all other symbols as constants — exactly what the definition requires.

```python
x, y, mu, s, n_sym = sp.symbols('x y mu s n', real=True)
# s represents sigma^2 throughout

# --- Example 1: f = 3x^2*y + y^3 - 5x ---
f1 = 3*x**2*y + y**3 - 5*x
df1_dx = sp.diff(f1, x)
df1_dy = sp.diff(f1, y)
print('f1 =', f1)
print('  df1/dx =', df1_dx, '  (expected: 6*x*y - 5)')
print('  df1/dy =', df1_dy, '  (expected: 3*x**2 + 3*y**2)')

# --- Example 3: f = exp(x^2 + y^2) ---
f3 = sp.exp(x**2 + y**2)
df3_dx = sp.diff(f3, x)
df3_dy = sp.diff(f3, y)
print('\nf3 =', f3)
print('  df3/dx =', df3_dx, '  (expected: 2*x*exp(x^2+y^2))')
print('  df3/dy =', df3_dy)

# --- Gaussian log-likelihood (one observation, symbolic) ---
# ell_1 = log-likelihood contribution from one observation x
ell_1 = -sp.Rational(1,2)*sp.log(2*sp.pi) - sp.Rational(1,2)*sp.log(s) - (x - mu)**2 / (2*s)
dell_dmu = sp.diff(ell_1, mu)
dell_ds  = sp.diff(ell_1, s)
print('\nSingle-obs log-likelihood contribution:')
print('  d(ell_1)/d(mu) =', sp.simplify(dell_dmu), '  (expected: (x-mu)/s)')
print('  d(ell_1)/d(s)  =', sp.simplify(dell_ds),  '  (expected: -(1/(2s)) + (x-mu)^2/(2s^2))')
```

```python
# --- Full log-likelihood: sum over n observations ---
# Represent the sufficient statistics as symbols
x_bar_sym, ss_sym = sp.symbols('xbar ss', real=True, positive=True)
# ss = sum((x_i - mu)^2) = sum(x_i^2) - n*x_bar^2 + n*(x_bar - mu)^2 ...
# For the MLE derivation, it is cleaner to write ell in terms of mu and s directly.

# ell(mu, s) = -n/2*log(2*pi) - n/2*log(s) - 1/(2s)*sum((x_i-mu)^2)
# After substituting sum((x_i-mu)^2) = n*s_hat + n*(mu - x_bar)^2
# (where s_hat = (1/n)*sum((x_i-x_bar)^2)), write a symbolic version:

ell_sym = (-n_sym/2 * sp.log(2*sp.pi)
           - n_sym/2 * sp.log(s)
           - 1/(2*s) * (ss_sym + n_sym*(mu - x_bar_sym)**2))

score_mu = sp.diff(ell_sym, mu)
score_s  = sp.diff(ell_sym, s)

print('Score components (full n-observation log-likelihood):')
print('  s_mu  =', sp.simplify(score_mu))
print('  s_s2  =', sp.simplify(score_s))

# Verify MLE: solve score = 0 for each parameter
mle_mu = sp.solve(score_mu, mu)
mle_s  = sp.solve(score_s,  s)
print('\nMLE solutions:')
print('  mu_hat  =', mle_mu, '  (expected: x_bar)')
print('  s_hat   =', mle_s,  '  (expected: ss/n)')
```

### What Do You See?

1. SymPy computes partial derivatives by holding all other symbols constant — exactly the limit definition. For the log-likelihood, the score components match what we derived in Section 3.
2. `sp.solve(score_mu, mu)` finds the root of the score equation $s_\mu = 0$. The result should be `x_bar` — the sample mean. Does `sp.solve(score_s, s)` return `ss/n` (i.e., $\frac{1}{n}\sum(x_i - \bar{x})^2$)?

## Section 2 — Geometric Interpretation: Cross-Sections

We visualise the surface $f(x,y) = \sin(x)\cos(y) + 0.5x$ and its two first-order partial derivatives as cross-section slopes.

```python
# Surface: f(x,y) = sin(x)*cos(y) + 0.5*x
def f_surf(xv, yv):
    return np.sin(xv) * np.cos(yv) + 0.5 * xv

def df_dx(xv, yv):
    return np.cos(xv) * np.cos(yv) + 0.5

def df_dy(xv, yv):
    return -np.sin(xv) * np.sin(yv)

# Evaluation point
x0, y0 = 1.0, 0.8

# Fine grid for surface
xg = np.linspace(-2, 3, 200)
yg = np.linspace(-2, 2, 200)
XG, YG = np.meshgrid(xg, yg)
ZG = f_surf(XG, YG)

fig = plt.figure(figsize=(18, 5))

# ---- Left: 3D surface with cross-section planes ----
ax1 = fig.add_subplot(131, projection='3d')
ax1.plot_surface(XG, YG, ZG, cmap='coolwarm', alpha=0.55, linewidth=0)

# x-cross-section at y = y0
xs = np.linspace(-2, 3, 200)
ax1.plot(xs, np.full_like(xs, y0), f_surf(xs, y0), color=TS_BLUE, lw=2,
         label=f'$y={y0}$ cross-section')
# y-cross-section at x = x0
ys = np.linspace(-2, 2, 200)
ax1.plot(np.full_like(ys, x0), ys, f_surf(x0, ys), color=TS_AMBER, lw=2,
         label=f'$x={x0}$ cross-section')
ax1.scatter([x0], [y0], [f_surf(x0, y0)], color=TS_GREEN, s=80, zorder=6)

ax1.set_xlabel('$x$'); ax1.set_ylabel('$y$'); ax1.set_zlabel('$f$')
ax1.set_title('Surface with cross-sections', fontweight='bold', color=TS_GRAY)
ax1.tick_params(labelsize=8)
ax1.legend(fontsize=8)

# ---- Middle: x-cross-section + tangent line ----
ax2 = fig.add_subplot(132)
ax2.plot(xs, f_surf(xs, y0), color=TS_BLUE, lw=2,
         label=r'$g(x) = f(x,' + f'{y0})$')
slope_x = df_dx(x0, y0)
tangent_x = f_surf(x0, y0) + slope_x * (xs - x0)
ax2.plot(xs, tangent_x, color=TS_AMBER, lw=1.5, ls='--',
         label=f'Tangent, slope = $\\partial f/\\partial x = {slope_x:.3f}$')
ax2.scatter([x0], [f_surf(x0, y0)], color=TS_GREEN, s=80, zorder=5)
ax2.set_xlim(-2, 3); ax2.set_ylim(-2, 2)
ax2.set_xlabel('$x$'); ax2.set_ylabel('$f$')
ax2.set_title(r'$x$-cross-section ($y$ fixed)' + '\n'
              + r'slope = $\partial f / \partial x$', fontweight='bold', color=TS_GRAY)
ax2.legend(fontsize=9)

# ---- Right: y-cross-section + tangent line ----
ax3 = fig.add_subplot(133)
ax3.plot(ys, f_surf(x0, ys), color=TS_AMBER, lw=2,
         label=r'$h(y) = f(' + f'{x0},y)$')
slope_y = df_dy(x0, y0)
tangent_y = f_surf(x0, y0) + slope_y * (ys - y0)
ax3.plot(ys, tangent_y, color=TS_BLUE, lw=1.5, ls='--',
         label=f'Tangent, slope = $\\partial f/\\partial y = {slope_y:.3f}$')
ax3.scatter([y0], [f_surf(x0, y0)], color=TS_GREEN, s=80, zorder=5)
ax3.set_xlim(-2, 2)
ax3.set_xlabel('$y$'); ax3.set_ylabel('$f$')
ax3.set_title(r'$y$-cross-section ($x$ fixed)' + '\n'
              + r'slope = $\partial f / \partial y$', fontweight='bold', color=TS_GRAY)
ax3.legend(fontsize=9)

plt.suptitle(f'Module 02, Unit 01 — Partial Derivatives as Cross-Section Slopes\n'
             f'Evaluation point: $({x0}, {y0})$',
             fontsize=13, fontweight='bold', color=TS_GRAY)
plt.tight_layout()
plt.savefig('module-02-unit-01-cross-sections.png', dpi=150, bbox_inches='tight')
plt.show()
print(f'df/dx at ({x0},{y0}) = {slope_x:.4f}')
print(f'df/dy at ({x0},{y0}) = {slope_y:.4f}')
```

### What Do You See?

1. The blue curve (left of middle panel) is the cross-section of the surface at $y = 0.8$ — a 1D curve in $x$. Its tangent line slope at $x_0$ is exactly $\frac{\partial f}{\partial x}(x_0, y_0)$. Does the sign of the slope match the direction the surface appears to be going at that point in the 3D plot?

2. The orange curve (right panel) is the cross-section at $x = 1.0$, varying in $y$. Its slope is $\frac{\partial f}{\partial y}(x_0, y_0)$. The two tangent slopes are independent measurements — one tells you about the $x$-direction, the other about the $y$-direction.

## Section 3 — Statistical Bridge: Score Components [MLE]

We now plot $\frac{\partial \ell}{\partial \mu}$ and $\frac{\partial \ell}{\partial \sigma^2}$ as functions of their respective parameters, holding the other fixed at its MLE value, and verify numerically that both are zero at the MLE.

```python
rng    = np.random.default_rng(seed=42)
n_data = 30
data   = rng.normal(loc=2.0, scale=1.0, size=n_data)
x_bar  = data.mean()
s2_hat = data.var()          # MLE for sigma^2
ss     = np.sum((data - x_bar)**2)  # sum of squared deviations

# Score components
def score_mu(mu_val, s2_val, obs):
    return np.sum(obs - mu_val) / s2_val

def score_s2(mu_val, s2_val, obs):
    n = len(obs)
    return -n / (2 * s2_val) + np.sum((obs - mu_val)**2) / (2 * s2_val**2)

mu_range  = np.linspace(x_bar - 1.5, x_bar + 1.5, 400)
s2_range  = np.linspace(max(0.05, s2_hat - 1.0), s2_hat + 1.5, 400)

s_mu_vals = score_mu(mu_range, s2_hat, data)
s_s2_vals = score_s2(x_bar, s2_range, data)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# ---- Left: score w.r.t. mu ----
ax = axes[0]
ax.plot(mu_range, s_mu_vals, color=TS_BLUE, lw=2)
ax.axhline(0, color=TS_GRAY, lw=1, ls='--')
ax.axvline(x_bar, color=TS_AMBER, lw=1.5, ls=':', label=f'MLE $\\bar{{x}}={x_bar:.3f}$')
ax.scatter([x_bar], [score_mu(x_bar, s2_hat, data)],
           color=TS_AMBER, s=80, zorder=5)
ax.fill_between(mu_range, s_mu_vals, 0,
                where=s_mu_vals > 0, alpha=0.12, color=TS_BLUE, label='Score > 0')
ax.fill_between(mu_range, s_mu_vals, 0,
                where=s_mu_vals < 0, alpha=0.12, color=TS_RED,  label='Score < 0')
ax.set_xlabel('$\\mu$')
ax.set_ylabel(r'$\partial \ell / \partial \mu$')
ax.set_title(r'Score w.r.t. $\mu$ — zero at MLE $\bar{x}$',
             fontweight='bold', color=TS_GRAY)
ax.legend(fontsize=9)

# ---- Right: score w.r.t. sigma^2 ----
ax = axes[1]
ax.plot(s2_range, s_s2_vals, color=TS_GREEN, lw=2)
ax.axhline(0, color=TS_GRAY, lw=1, ls='--')
ax.axvline(s2_hat, color=TS_AMBER, lw=1.5, ls=':',
           label=f'MLE $\\hat{{\\sigma}}^2={s2_hat:.3f}$')
ax.scatter([s2_hat], [score_s2(x_bar, s2_hat, data)],
           color=TS_AMBER, s=80, zorder=5)
ax.fill_between(s2_range, s_s2_vals, 0,
                where=s_s2_vals > 0, alpha=0.12, color=TS_GREEN, label='Score > 0')
ax.fill_between(s2_range, s_s2_vals, 0,
                where=s_s2_vals < 0, alpha=0.12, color=TS_RED,   label='Score < 0')
ax.set_xlabel('$\\sigma^2$')
ax.set_ylabel(r'$\partial \ell / \partial \sigma^2$')
ax.set_title(r'Score w.r.t. $\sigma^2$ — zero at MLE $\hat{\sigma}^2$',
             fontweight='bold', color=TS_GRAY)
ax.legend(fontsize=9)

plt.suptitle('Module 02, Unit 01 — Partial Derivatives of the Log-Likelihood [MLE]',
             fontsize=13, fontweight='bold', color=TS_GRAY)
plt.tight_layout()
plt.savefig('module-02-unit-01-score.png', dpi=150, bbox_inches='tight')
plt.show()

print(f'Score (mu)  at MLE: {score_mu(x_bar, s2_hat, data):.6e}  (expected ~0)')
print(f'Score (s2)  at MLE: {score_s2(x_bar, s2_hat, data):.6e}  (expected ~0)')
```

### What Do You See?

1. Both score components cross zero at exactly the MLE values (dotted lines). The score is positive to the left of the MLE (log-likelihood still increasing) and negative to the right (log-likelihood decreasing). This is the same sign-analysis from Unit 00-03, now applied to each partial independently.

2. Notice that both panels were produced by holding the *other* parameter at its MLE value. This is exactly what a partial derivative does: fix all other variables, vary one. We can only do this independently because $\mu$ and $\sigma^2$ are orthogonal — the two score equations decouple. We will quantify this decoupling in Unit 02-02.

## Summary

| Concept | Key result |
|---|---|
| Partial derivative definition | $\frac{\partial f}{\partial x_i}(\boldsymbol{a}) = \lim_{h\to 0}\frac{f(\boldsymbol{a} + h\boldsymbol{e}_i) - f(\boldsymbol{a})}{h}$ |
| Computation rule | Hold all other variables constant; apply single-variable rules to $x_i$ |
| Geometric meaning | Slope of the surface in the $x_i$-direction; slope of the $x_i$-cross-section |
| Notation | $\frac{\partial f}{\partial x}$, $f_x$, $\partial_x f$ — all equivalent |
| Gaussian score (mu) | $\frac{\partial \ell}{\partial \mu} = \frac{1}{\sigma^2}\sum_i(x_i - \mu)$; zero at $\hat{\mu} = \bar{x}$ |
