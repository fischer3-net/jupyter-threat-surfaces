## Why This Unit

In Module 02 you computed partial derivatives one at a time — $\frac{\partial f}{\partial x_1}$, $\frac{\partial f}{\partial x_2}$, and so on. Each partial derivative is a scalar that measures how $f$ changes in one coordinate direction while all others are held fixed.

The key insight of this module is that those scalars are not independent: **assembled into a single vector, the partial derivatives become the gradient**, and the gradient knows everything about how $f$ changes in every direction simultaneously. That vector is the central object in:

- Gradient-based optimization (Module 04)
- Maximum likelihood estimation (this unit)
- Neural network training via backpropagation (Unit 03)
- The multivariate normal density (Module 06)

This unit builds the object. Unit 02 will show how to use it to move through space.

## 1. From Partial Derivatives to the Gradient

Let $f : \mathbb{R}^n \to \mathbb{R}$ be a differentiable scalar-valued function. Recall from Module 02 that the partial derivative $\frac{\partial f}{\partial x_i}$ is computed by differentiating with respect to $x_i$ while treating all other variables as constants.

**Definition.** The **gradient** of $f$ at a point $\boldsymbol{x} \in \mathbb{R}^n$ is the column vector of all partial derivatives:

$$
\nabla f(\boldsymbol{x}) \coloneqq \begin{pmatrix} \dfrac{\partial f}{\partial x_1}(\boldsymbol{x}) \\[8pt] \dfrac{\partial f}{\partial x_2}(\boldsymbol{x}) \\[4pt] \vdots \\[4pt] \dfrac{\partial f}{\partial x_n}(\boldsymbol{x}) \end{pmatrix} \in \mathbb{R}^n
$$

The symbol $\nabla$ is read **nabla** or **del**. The gradient $\nabla f(\boldsymbol{x})$ is a **vector**, not a scalar. Its dimension equals the number of input variables.

> **Input vs output.** The function $f$ maps $\mathbb{R}^n \to \mathbb{R}$ — it takes a vector and returns a scalar. The gradient $\nabla f$ maps $\mathbb{R}^n \to \mathbb{R}^n$ — it takes a point and returns a vector. At each point in the domain, there is a gradient vector telling you which way $f$ is increasing most rapidly from that point.

**Notation.** You will see several equivalent forms:

$$
\nabla f(\boldsymbol{x}) = \nabla_{\boldsymbol{x}} f = \text{grad}\, f
$$

In statistics and machine learning, the subscript $\nabla_{\boldsymbol{\theta}} \ell(\boldsymbol{\theta})$ is used to make the variable of differentiation explicit — you will see this in the statistical bridge.

## 2. Computing the Gradient

Computing the gradient is mechanical: differentiate $f$ with respect to each variable in turn and stack the results into a column vector.

### 2.1 Functions of Two Variables

For $f : \mathbb{R}^2 \to \mathbb{R}$, the gradient is a 2-vector:

$$
\nabla f(x, y) = \begin{pmatrix} \dfrac{\partial f}{\partial x} \\[8pt] \dfrac{\partial f}{\partial y} \end{pmatrix}
$$

**Worked Example 2.1.** Let $f(x, y) = x^2 y + 3y^2$.

$$
\frac{\partial f}{\partial x} = 2xy \qquad \frac{\partial f}{\partial y} = x^2 + 6y
$$

$$
\nabla f(x, y) = \begin{pmatrix} 2xy \\ x^2 + 6y \end{pmatrix}
$$

At the point $(1, 2)$:

$$
\nabla f(1, 2) = \begin{pmatrix} 2(1)(2) \\ (1)^2 + 6(2) \end{pmatrix} = \begin{pmatrix} 4 \\ 13 \end{pmatrix}
$$

This tells you that at the point $(1, 2)$, $f$ is increasing at rate $4$ in the $x$-direction, at rate $13$ in the $y$-direction, and most rapidly in the direction of the vector $(4, 13)^{\top}$.

**Worked Example 2.2.** Let $f(x, y) = e^{xy} + \ln(x^2 + 1)$. (Assume $x > 0$.)

Using the chain rule on each partial:

$$
\frac{\partial f}{\partial x} = y\,e^{xy} + \frac{2x}{x^2 + 1} \qquad \frac{\partial f}{\partial y} = x\,e^{xy}
$$

$$
\nabla f(x, y) = \begin{pmatrix} y\,e^{xy} + \dfrac{2x}{x^2 + 1} \\[8pt] x\,e^{xy} \end{pmatrix}
$$

### 2.2 Functions of Three Variables

For $f : \mathbb{R}^3 \to \mathbb{R}$, simply add a third component.

**Worked Example 2.3.** Let $f(x, y, z) = x^2 z - y\sin(z)$.

$$
\frac{\partial f}{\partial x} = 2xz \qquad \frac{\partial f}{\partial y} = -\sin(z) \qquad \frac{\partial f}{\partial z} = x^2 - y\cos(z)
$$

$$
\nabla f(x, y, z) = \begin{pmatrix} 2xz \\ -\sin(z) \\ x^2 - y\cos(z) \end{pmatrix}
$$

### 2.3 The Gradient of a Quadratic Form

This result appears throughout statistics and machine learning whenever an objective function is quadratic in the parameters.

**Setup.** Let $\boldsymbol{A} \in \mathbb{R}^{n \times n}$ and define $f(\boldsymbol{x}) = \boldsymbol{x}^{\top}\boldsymbol{A}\boldsymbol{x}$ for $\boldsymbol{x} \in \mathbb{R}^n$.

**Result.**

$$
\nabla_{\boldsymbol{x}}\,\boldsymbol{x}^{\top}\boldsymbol{A}\boldsymbol{x} = (\boldsymbol{A} + \boldsymbol{A}^{\top})\boldsymbol{x}
$$

**Symmetric case.** If $\boldsymbol{A}$ is symmetric ($\boldsymbol{A} = \boldsymbol{A}^{\top}$), this simplifies to $2\boldsymbol{A}\boldsymbol{x}$. This is the form that appears in Ridge regression and the multivariate normal.

> **Two useful special cases to memorize:**
>
> $$\nabla_{\boldsymbol{x}}\,\boldsymbol{a}^{\top}\boldsymbol{x} = \boldsymbol{a} \qquad \nabla_{\boldsymbol{x}}\,\|\boldsymbol{x}\|^2 = \nabla_{\boldsymbol{x}}\,\boldsymbol{x}^{\top}\boldsymbol{x} = 2\boldsymbol{x}$$
>
> The first says the gradient of a linear function is its coefficient vector. The second says the gradient of the squared norm points radially outward.

## 3. Geometric Interpretation

The algebraic definition tells you how to compute the gradient. The geometric interpretation tells you what it *means*.

### 3.1 Steepest Ascent

**Theorem.** Among all unit vectors $\boldsymbol{u} \in \mathbb{R}^n$ with $\|\boldsymbol{u}\| = 1$, the directional derivative of $f$ is maximized in the direction of $\nabla f(\boldsymbol{x})$.

More precisely (we will prove this in Unit 02):

- The direction of **steepest ascent** at $\boldsymbol{x}$ is $\hat{\boldsymbol{u}} = \dfrac{\nabla f(\boldsymbol{x})}{\|\nabla f(\boldsymbol{x})\|}$.
- The direction of **steepest descent** is $-\dfrac{\nabla f(\boldsymbol{x})}{\|\nabla f(\boldsymbol{x})\|}$.
- The **maximum rate of increase** of $f$ from $\boldsymbol{x}$ is $\|\nabla f(\boldsymbol{x})\|$.

This is why gradient descent subtracts the gradient at each step: moving in the direction $-\nabla f$ is guaranteed to decrease $f$ (locally).

### 3.2 Perpendicularity to Level Sets

A **level set** (level curve in $\mathbb{R}^2$, level surface in $\mathbb{R}^3$) of $f$ is the set of all points where $f$ takes a constant value $c$:

$$
L_c = \{\boldsymbol{x} \in \mathbb{R}^n : f(\boldsymbol{x}) = c\}
$$

**Theorem.** The gradient $\nabla f(\boldsymbol{x})$ is **perpendicular to the level set** through $\boldsymbol{x}$.

**Why this is true.** Consider a smooth curve $\boldsymbol{\gamma}(t)$ that stays on the level set $L_c$, so $f(\boldsymbol{\gamma}(t)) = c$ for all $t$. Differentiating both sides with respect to $t$ using the chain rule:

$$
\frac{d}{dt}\,f(\boldsymbol{\gamma}(t)) = \nabla f(\boldsymbol{\gamma}(t)) \cdot \boldsymbol{\gamma}'(t) = 0
$$

Since $\boldsymbol{\gamma}'(t)$ is the tangent vector to the level set, and the dot product with $\nabla f$ is zero, the gradient is orthogonal to every tangent vector of the level set — that is, perpendicular to the level set itself.

**Consequence.** To move along a level set (keeping $f$ constant), you move perpendicular to the gradient. To change $f$ as rapidly as possible, you move parallel to the gradient.

> **Example.** For $f(x, y) = x^2 + y^2$ (a bowl), the level curves are circles. At any point $(x_0, y_0)$ on a circle, the gradient $\nabla f = (2x_0, 2y_0)^{\top}$ points radially outward — exactly perpendicular to the circle. This can be verified directly: the tangent to the circle $x^2 + y^2 = r^2$ at $(x_0, y_0)$ is proportional to $(-y_0, x_0)^{\top}$, and $(2x_0, 2y_0) \cdot (-y_0, x_0) = -2x_0 y_0 + 2y_0 x_0 = 0$. ✓

## 4. Properties of the Gradient

These properties mirror the rules from single-variable differentiation and are used constantly in statistical derivations.

### 4.1 Linearity

For differentiable $f, g : \mathbb{R}^n \to \mathbb{R}$ and constants $a, b \in \mathbb{R}$:

$$
\nabla(af + bg) = a\,\nabla f + b\,\nabla g
$$

This follows immediately from linearity of partial derivatives.

### 4.2 Product Rule

$$
\nabla(f \cdot g) = g\,\nabla f + f\,\nabla g
$$

Note that $g$ and $f$ here are scalars (the function values), while $\nabla f$ and $\nabla g$ are vectors.

### 4.3 Chain Rule for Gradients

If $h : \mathbb{R} \to \mathbb{R}$ is a differentiable scalar function and $f : \mathbb{R}^n \to \mathbb{R}$, then the composition $h \circ f : \mathbb{R}^n \to \mathbb{R}$ has gradient:

$$
\nabla[h(f(\boldsymbol{x}))] = h'(f(\boldsymbol{x})) \cdot \nabla f(\boldsymbol{x})
$$

Read: *scalar multiplier* $h'(f(\boldsymbol{x}))$ times *gradient vector* $\nabla f(\boldsymbol{x})$.

**Example.** Let $f(\boldsymbol{x}) = \|\boldsymbol{x} - \boldsymbol{\mu}\|^2$ and $h(t) = e^{-t/2}$. Then $h(f(\boldsymbol{x})) = e^{-\|\boldsymbol{x}-\boldsymbol{\mu}\|^2 / 2}$ (the unnormalized Gaussian kernel), and:

$$
\nabla_{\boldsymbol{x}}\,e^{-\|\boldsymbol{x} - \boldsymbol{\mu}\|^2 / 2} = e^{-\|\boldsymbol{x} - \boldsymbol{\mu}\|^2 / 2} \cdot (-1) \cdot (\boldsymbol{x} - \boldsymbol{\mu}) = -(\boldsymbol{x} - \boldsymbol{\mu})\,e^{-\|\boldsymbol{x} - \boldsymbol{\mu}\|^2 / 2}
$$

This gradient appears when differentiating the multivariate normal density in Module 06.

### 4.4 Gradient Reference Table

| $f(\boldsymbol{x})$ | $\nabla_{\boldsymbol{x}} f$ | Notes |
|---|---|---|
| $\boldsymbol{a}^{\top}\boldsymbol{x}$ | $\boldsymbol{a}$ | $\boldsymbol{a}$ constant |
| $\boldsymbol{x}^{\top}\boldsymbol{a}$ | $\boldsymbol{a}$ | same as above |
| $\boldsymbol{x}^{\top}\boldsymbol{x}$ | $2\boldsymbol{x}$ | squared norm |
| $\boldsymbol{x}^{\top}\boldsymbol{A}\boldsymbol{x}$ | $(\boldsymbol{A} + \boldsymbol{A}^{\top})\boldsymbol{x}$ | quadratic form |
| $\boldsymbol{x}^{\top}\boldsymbol{A}\boldsymbol{x}$, $\boldsymbol{A}$ symmetric | $2\boldsymbol{A}\boldsymbol{x}$ | symmetric quadratic form |
| $\|\boldsymbol{x} - \boldsymbol{a}\|^2$ | $2(\boldsymbol{x} - \boldsymbol{a})$ | squared distance |
| $\log(\boldsymbol{a}^{\top}\boldsymbol{x})$ | $\dfrac{\boldsymbol{a}}{\boldsymbol{a}^{\top}\boldsymbol{x}}$ | $\boldsymbol{a}$ constant, $\boldsymbol{a}^{\top}\boldsymbol{x} > 0$ |
| $h(\boldsymbol{a}^{\top}\boldsymbol{x})$ | $h'(\boldsymbol{a}^{\top}\boldsymbol{x})\,\boldsymbol{a}$ | chain rule, $h$ scalar |

## 5. Statistical Bridge — The Score Function `[MLE]`

Maximum likelihood estimation (MLE) finds the parameter vector $\boldsymbol{\theta}$ that maximizes the log-likelihood $\ell(\boldsymbol{\theta})$. The gradient of the log-likelihood with respect to the parameters has a name: the **score function**. Finding the MLE is exactly the problem of finding where the score is zero.

### 5.1 Setup: Gaussian Log-Likelihood

Suppose we observe $n$ independent draws from a normal distribution: $x_1, \ldots, x_n \overset{\text{iid}}{\sim} \mathcal{N}(\mu, \sigma^2)$. The log-likelihood is a function of two parameters $\boldsymbol{\theta} = (\mu, \sigma^2)^{\top}$:

$$
\ell(\mu, \sigma^2) = -\frac{n}{2}\log(2\pi\sigma^2) - \frac{1}{2\sigma^2}\sum_{i=1}^{n}(x_i - \mu)^2
$$

This is a scalar-valued function of the two-dimensional parameter vector $\boldsymbol{\theta} = (\mu, \sigma^2)^{\top}$. Its gradient is the score:

$$
\nabla_{\boldsymbol{\theta}}\,\ell(\boldsymbol{\theta}) = \begin{pmatrix} \dfrac{\partial \ell}{\partial \mu} \\[8pt] \dfrac{\partial \ell}{\partial \sigma^2} \end{pmatrix}
$$

### 5.2 Computing the Score

**Partial with respect to $\mu$:**

$$
\frac{\partial \ell}{\partial \mu} = -\frac{1}{2\sigma^2} \cdot \frac{\partial}{\partial \mu}\sum_{i=1}^{n}(x_i - \mu)^2 = -\frac{1}{2\sigma^2} \sum_{i=1}^{n} 2(x_i - \mu)(-1) = \frac{1}{\sigma^2}\sum_{i=1}^{n}(x_i - \mu)
$$

**Partial with respect to $\sigma^2$ (treating $\sigma^2$ as a single variable $v = \sigma^2$):**

$$
\frac{\partial \ell}{\partial \sigma^2} = -\frac{n}{2\sigma^2} + \frac{1}{2(\sigma^2)^2}\sum_{i=1}^{n}(x_i - \mu)^2
$$

### 5.3 The MLE Condition

Setting $\nabla_{\boldsymbol{\theta}}\,\ell = \boldsymbol{0}$ and solving:

**From** $\dfrac{\partial \ell}{\partial \mu} = 0$:

$$
\sum_{i=1}^{n}(x_i - \hat{\mu}) = 0 \implies \hat{\mu}_{\text{MLE}} = \frac{1}{n}\sum_{i=1}^{n} x_i = \bar{x}
$$

**From** $\dfrac{\partial \ell}{\partial \sigma^2} = 0$:

$$
\hat{\sigma}^2_{\text{MLE}} = \frac{1}{n}\sum_{i=1}^{n}(x_i - \bar{x})^2
$$

Both familiar estimators fall out directly from the gradient condition $\nabla_{\boldsymbol{\theta}}\,\ell(\boldsymbol{\theta}) = \boldsymbol{0}$.

### 5.4 The General Picture

For any model with log-likelihood $\ell(\boldsymbol{\theta})$:

$$
\hat{\boldsymbol{\theta}}_{\text{MLE}} = \arg\max_{\boldsymbol{\theta}}\,\ell(\boldsymbol{\theta}) \qquad \Longleftrightarrow \qquad \nabla_{\boldsymbol{\theta}}\,\ell(\hat{\boldsymbol{\theta}}_{\text{MLE}}) = \boldsymbol{0}
$$

When there is no closed-form solution to $\nabla_{\boldsymbol{\theta}}\,\ell = \boldsymbol{0}$, we use gradient-based methods (Module 04) to find $\hat{\boldsymbol{\theta}}_{\text{MLE}}$ numerically — following $+\nabla_{\boldsymbol{\theta}}\,\ell$ uphill until the gradient is small.

> **Notation reminder.** The score function is sometimes written $\boldsymbol{s}(\boldsymbol{\theta}) \coloneqq \nabla_{\boldsymbol{\theta}}\,\ell(\boldsymbol{\theta})$. The Fisher information matrix is $\mathcal{I}(\boldsymbol{\theta}) \coloneqq -\mathbb{E}[\nabla_{\boldsymbol{\theta}}^2\,\ell(\boldsymbol{\theta})]$ — the negative expected Hessian. We will return to this in Module 04 when we cover second-order conditions.

### Section 1 — Symbolic Gradient Verification

Use SymPy to verify the three worked examples from Section 2 exactly. If your hand answer disagrees with SymPy, trust SymPy and find the error.

```python
# ============================================================
# Section 1 — Symbolic gradient verification
# ============================================================
x, y, z = symbols('x y z', real=True)

# --- Example 2.1 ---
f1 = x**2 * y + 3*y**2
grad_f1 = Matrix([diff(f1, x), diff(f1, y)])
print('Example 2.1  f(x,y) = x²y + 3y²')
print('  ∇f =', grad_f1.T)
print('  ∇f(1,2) =', grad_f1.subs([(x, 1), (y, 2)]).T)

# --- Example 2.2 ---
f2 = exp(x*y) + log(x**2 + 1)
grad_f2 = Matrix([diff(f2, x), diff(f2, y)])
print('\nExample 2.2  f(x,y) = e^(xy) + ln(x² + 1)')
print('  ∂f/∂x =', simplify(diff(f2, x)))
print('  ∂f/∂y =', diff(f2, y))

# --- Example 2.3 ---
f3 = x**2 * z - y * sin(z)
grad_f3 = Matrix([diff(f3, x), diff(f3, y), diff(f3, z)])
print('\nExample 2.3  f(x,y,z) = x²z − y sin(z)')
print('  ∇f =', grad_f3.T)

# --- Quadratic form: ∇(xᵀAx) = (A + Aᵀ)x ---
a11, a12, a21, a22 = symbols('a11 a12 a21 a22', real=True)
xv = Matrix([x, y])
A  = Matrix([[a11, a12], [a21, a22]])
qf = (xv.T * A * xv)[0, 0]   # scalar
grad_qf = Matrix([diff(qf, x), diff(qf, y)])
expected = (A + A.T) * xv
print('\nQuadratic form  ∇(xᵀAx) = (A+Aᵀ)x ?', simplify(grad_qf - expected) == Matrix([0, 0]))
```

**What do you see?** Confirm each gradient matches your hand calculation. Note that the quadratic form check prints `True` — SymPy confirms the formula $(\boldsymbol{A} + \boldsymbol{A}^{\top})\boldsymbol{x}$ algebraically for any $2 \times 2$ matrix.

### Section 2 — Gradient Field Visualization

A **gradient field** plots the gradient vector $\nabla f(x, y)$ at a grid of points as arrows. Arrows pointing in the direction of steepest ascent at each location. Observe how the arrows relate to the level curves.

```python
# ============================================================
# Section 2 — Gradient field with level curves
# ============================================================
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# --- Left: f(x,y) = x² + y²  (bowl — radially symmetric) ---
ax = axes[0]
grid_vals = np.linspace(-2.5, 2.5, 300)
X, Y = np.meshgrid(grid_vals, grid_vals)
Z = X**2 + Y**2

# Level curves
cs = ax.contour(X, Y, Z, levels=np.arange(0.5, 6.0, 0.75),
                colors=[TS_GRAY], alpha=0.5, linewidths=1.0)
ax.clabel(cs, inline=True, fontsize=8, fmt='%.2g')

# Gradient field
xq = np.linspace(-2.2, 2.2, 10)
Xq, Yq = np.meshgrid(xq, xq)
Gx = 2 * Xq   # ∂f/∂x = 2x
Gy = 2 * Yq   # ∂f/∂y = 2y
norm = np.sqrt(Gx**2 + Gy**2) + 1e-8
ax.quiver(Xq, Yq, Gx/norm, Gy/norm,
          color=TS_BLUE, alpha=0.75, scale=18, width=0.004)

ax.set_title(r'$f(x,y) = x^2 + y^2$', fontsize=13)
ax.set_xlabel('x'); ax.set_ylabel('y')
ax.set_aspect('equal')

# --- Right: f(x,y) = x²y + 3y²  (Example 2.1) ---
ax = axes[1]
Z2 = X**2 * Y + 3 * Y**2

cs2 = ax.contour(X, Y, Z2, levels=np.linspace(-8, 8, 16),
                 colors=[TS_GRAY], alpha=0.5, linewidths=1.0)
ax.clabel(cs2, inline=True, fontsize=8, fmt='%d')

Gx2 = 2 * Xq * Yq          # ∂f/∂x = 2xy
Gy2 = Xq**2 + 6 * Yq       # ∂f/∂y = x² + 6y
norm2 = np.sqrt(Gx2**2 + Gy2**2) + 1e-8
ax.quiver(Xq, Yq, Gx2/norm2, Gy2/norm2,
          color=TS_AMBER, alpha=0.75, scale=18, width=0.004)

# Highlight the point (1, 2) from the worked example
px, py = 1.0, 2.0
gx_pt, gy_pt = 2*px*py, px**2 + 6*py
n_pt = np.sqrt(gx_pt**2 + gy_pt**2)
ax.quiver(px, py, gx_pt/n_pt, gy_pt/n_pt,
          color=TS_RED, scale=7, width=0.009, label=r'$\nabla f(1,2)$')
ax.plot(px, py, 'o', color=TS_RED, markersize=7)
ax.legend(fontsize=10)

ax.set_title(r'$f(x,y) = x^2y + 3y^2$', fontsize=13)
ax.set_xlabel('x'); ax.set_ylabel('y')
ax.set_aspect('equal')
ax.set_xlim(-2.5, 2.5); ax.set_ylim(-2.5, 2.5)

fig.suptitle('Gradient Fields — arrows point in direction of steepest ascent',
             fontsize=12, y=1.01)
plt.tight_layout()
plt.show()
```

**What do you see?**

- **Left**: For the bowl $f = x^2 + y^2$, every arrow points radially outward — perpendicular to the circular level curves. The gradient is largest far from the origin (the arrows are normalized here; in an unnormalized plot they would grow with distance).
- **Right**: For $f = x^2 y + 3y^2$, the level curves are not circles. Notice that the red arrow at $(1, 2)$ is still exactly perpendicular to the level curve passing through that point.

### Section 3 — Perpendicularity of the Gradient to Level Sets

We verify the perpendicularity claim explicitly: at a chosen point, compute the gradient, find the tangent direction of the level curve, and confirm their dot product is zero.

```python
# ============================================================
# Section 3 — Explicit perpendicularity verification
# ============================================================
# f(x,y) = x² + y²   at the point (1, √3) on the level set f = 4

# Gradient at the point
p = np.array([1.0, np.sqrt(3)])
grad_p = 2 * p          # ∇f = (2x, 2y)ᵀ

# Tangent to the circle x²+y²=4 at (1,√3)
# Parameterize: (2cosθ, 2sinθ), tangent = (-2sinθ, 2cosθ)
# At (1,√3): cosθ = 1/2, sinθ = √3/2  →  tangent ∝ (-√3, 1)
tangent_p = np.array([-np.sqrt(3), 1.0])

dot_product = np.dot(grad_p, tangent_p)
print(f'Point: {p}')
print(f'Gradient ∇f: {grad_p}')
print(f'Tangent to level curve: {tangent_p}')
print(f'Dot product ∇f · tangent = {dot_product:.6f}  (expect 0)')

# --- Visualization ---
fig, ax = plt.subplots(figsize=(6, 6))
theta_vals = np.linspace(0, 2*np.pi, 300)
ax.plot(2*np.cos(theta_vals), 2*np.sin(theta_vals),
        color=TS_GRAY, lw=1.5, label='Level set $f = 4$')

# Gradient arrow
scale = 0.35
ax.annotate('', xy=p + grad_p*scale, xytext=p,
            arrowprops=dict(arrowstyle='->', color=TS_BLUE, lw=2.5))
ax.text(p[0] + grad_p[0]*scale + 0.05, p[1] + grad_p[1]*scale,
        r'$\nabla f$', color=TS_BLUE, fontsize=13, fontweight='bold')

# Tangent arrow
tang_norm = tangent_p / np.linalg.norm(tangent_p)
ax.annotate('', xy=p + tang_norm*0.7, xytext=p - tang_norm*0.7,
            arrowprops=dict(arrowstyle='->', color=TS_AMBER, lw=2.5))
ax.text(p[0] - tang_norm[0]*0.75 - 0.35, p[1] - tang_norm[1]*0.75 - 0.15,
        'tangent', color=TS_AMBER, fontsize=11)

# Right-angle mark
s = 0.12
u = grad_p / np.linalg.norm(grad_p) * s
v = tang_norm * s
corner = np.array([p, p+u, p+u+v, p+v, p])
ax.plot(corner[:,0], corner[:,1], color=TS_GRAY, lw=1.0)

ax.plot(*p, 'o', color=TS_RED, markersize=8, zorder=5)
ax.text(p[0]+0.08, p[1]+0.08, f'({p[0]:.1f}, {p[1]:.2f})',
        color=TS_RED, fontsize=10)

ax.set_xlim(-2.8, 2.8); ax.set_ylim(-2.8, 2.8)
ax.set_aspect('equal')
ax.set_title(r'$\nabla f$ is perpendicular to the level set at each point', fontsize=12)
ax.set_xlabel('x'); ax.set_ylabel('y')
ax.legend(fontsize=10)
ax.axhline(0, color=TS_GRAY, lw=0.5, alpha=0.5)
ax.axvline(0, color=TS_GRAY, lw=0.5, alpha=0.5)
plt.tight_layout()
plt.show()
```

**What do you see?** The dot product prints as exactly $0$ (or numerical noise near $10^{-16}$), confirming perpendicularity analytically. The diagram shows the right-angle mark between $\nabla f$ (blue) and the tangent to the level curve (amber).

### Section 4 — Statistical Bridge: Gradient of the Log-Likelihood

Visualize the Gaussian log-likelihood $\ell(\mu)$ with $\sigma^2$ held fixed, its gradient (score), and the convergence of a gradient-ascent search to the MLE $\hat{\mu}_{\text{MLE}} = \bar{x}$.

```python
# ============================================================
# Section 4 — Log-likelihood gradient and gradient ascent
# ============================================================
rng = np.random.default_rng(42)
n   = 20
true_mu    = 3.5
true_sigma = 1.2
data = rng.normal(true_mu, true_sigma, size=n)
x_bar = data.mean()
sigma2 = true_sigma**2   # treat variance as known for this demonstration

def log_lik(mu, x, s2):
    """Gaussian log-likelihood as a function of mu (sigma^2 fixed)."""
    return -0.5 * np.sum((x - mu)**2) / s2  # constant terms dropped

def score_mu(mu, x, s2):
    """∂ℓ/∂μ = (1/σ²) Σ(xᵢ − μ)"""
    return np.sum(x - mu) / s2

mu_vals = np.linspace(0.5, 6.5, 400)
ll_vals = np.array([log_lik(m, data, sigma2) for m in mu_vals])
sc_vals = np.array([score_mu(m, data, sigma2) for m in mu_vals])

# --- Gradient ascent from a starting point ---
lr      = 0.08
mu_curr = 1.0
history = [mu_curr]
for _ in range(25):
    mu_curr = mu_curr + lr * score_mu(mu_curr, data, sigma2)
    history.append(mu_curr)
history = np.array(history)

# --- Plotting ---
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Left: log-likelihood + MLE
ax = axes[0]
ax.plot(mu_vals, ll_vals, color=TS_BLUE, lw=2, label=r'$\ell(\mu)$')
ax.axvline(x_bar, color=TS_AMBER, lw=1.5, ls='--',
           label=r'$\hat{\mu}_{\mathrm{MLE}} = \bar{x} = $' + f'{x_bar:.3f}')
ax.axvline(true_mu, color=TS_GREEN, lw=1.0, ls=':',
           label=f'True $\mu = {true_mu}$')

# Mark gradient ascent path on the curve
ll_hist = np.array([log_lik(m, data, sigma2) for m in history])
ax.plot(history, ll_hist, 'o--', color=TS_RED, markersize=4, lw=1.0,
        alpha=0.6, label='Gradient ascent steps')
ax.plot(history[0], ll_hist[0], 's', color=TS_RED, markersize=9, label='Start')

ax.set_xlabel(r'$\mu$'); ax.set_ylabel(r'$\ell(\mu)$ (up to constant)')
ax.set_title('Log-likelihood and gradient ascent')
ax.legend(fontsize=9)

# Right: score function (= gradient)
ax = axes[1]
ax.plot(mu_vals, sc_vals, color=TS_GREEN, lw=2,
        label=r'Score: $\partial\ell/\partial\mu$')
ax.axhline(0, color=TS_GRAY, lw=1.0, ls='-', alpha=0.8)
ax.axvline(x_bar, color=TS_AMBER, lw=1.5, ls='--',
           label=r'$\hat{\mu}_{\mathrm{MLE}}$ (score = 0)')
ax.fill_between(mu_vals, sc_vals, 0, where=(sc_vals > 0),
                alpha=0.12, color=TS_GREEN, label='Score > 0 (ascend)')
ax.fill_between(mu_vals, sc_vals, 0, where=(sc_vals < 0),
                alpha=0.12, color=TS_RED, label='Score < 0 (descend)')
ax.set_xlabel(r'$\mu$'); ax.set_ylabel(r'$\partial\ell / \partial\mu$')
ax.set_title('Score function — zero at the MLE')
ax.legend(fontsize=9)

fig.suptitle(r'Gaussian log-likelihood: $n=20$, $\sigma^2$ known',
             fontsize=12, y=1.01)
plt.tight_layout()
plt.show()

print(f'\nGradient ascent final value: μ = {history[-1]:.5f}')
print(f'Sample mean x̄              : μ = {x_bar:.5f}')
print(f'Absolute error             :   {abs(history[-1] - x_bar):.2e}')
```

**What do you see?**

- **Left**: The log-likelihood is a concave curve in $\mu$ with its maximum at $\bar{x}$. The gradient-ascent path (red dots) climbs the curve from the starting point $\mu = 1.0$ and converges to $\hat{\mu}_{\text{MLE}}$.
- **Right**: The score function is a straight line that crosses zero exactly at $\bar{x}$. Left of $\bar{x}$ the score is positive (gradient says go right). Right of $\bar{x}$ the score is negative (gradient says go left). Gradient ascent zeros in on the MLE because it follows the sign of the score.

**Extend**: Increase the learning rate to $0.5$. What happens? Decrease it to $0.005$. How many steps are needed to converge? This is the learning rate sensitivity that dominates practical ML training.

## Summary

| Concept | Statement |
|---|---|
| Gradient definition | $\nabla f(\boldsymbol{x}) = \left(\frac{\partial f}{\partial x_1}, \ldots, \frac{\partial f}{\partial x_n}\right)^{\top} \in \mathbb{R}^n$ |
| Geometric meaning | Points in direction of steepest ascent; perpendicular to level sets |
| Magnitude | $\|\nabla f(\boldsymbol{x})\|$ = maximum rate of change of $f$ at $\boldsymbol{x}$ |
| Linearity | $\nabla(af + bg) = a\,\nabla f + b\,\nabla g$ |
| Chain rule | $\nabla[h(f(\boldsymbol{x}))] = h'(f(\boldsymbol{x}))\,\nabla f(\boldsymbol{x})$ |
| Linear function | $\nabla(\boldsymbol{a}^{\top}\boldsymbol{x}) = \boldsymbol{a}$ |
| Squared norm | $\nabla\,\|\boldsymbol{x}\|^2 = 2\boldsymbol{x}$ |
| Symmetric quadratic form | $\nabla(\boldsymbol{x}^{\top}\boldsymbol{A}\boldsymbol{x}) = 2\boldsymbol{A}\boldsymbol{x}$ for $\boldsymbol{A} = \boldsymbol{A}^{\top}$ |
| MLE condition | $\hat{\boldsymbol{\theta}}_{\text{MLE}}$ solves $\nabla_{\boldsymbol{\theta}}\,\ell(\boldsymbol{\theta}) = \boldsymbol{0}$ |
