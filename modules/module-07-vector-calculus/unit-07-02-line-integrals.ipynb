## Why This Unit

In Module 05 we integrated functions over regions of $\mathbb{R}^n$. Here we integrate along **curves** — one-dimensional paths through $\mathbb{R}^n$. The two key types are the scalar line integral (arc length weighted by $f$) and the vector line integral (work done by a force field along a path).

The payoff is the fundamental theorem of line integrals: for gradient fields, the line integral depends only on the endpoints, not the path. Gradient descent trajectories in machine learning are paths in exactly this kind of field — the gradient of the loss function. Understanding why the loss decreases monotonically along such paths, and why the total decrease depends only on the start and end points, is the content of this unit.

## 1. Curves and Parameterization

A **curve** $C$ in $\mathbb{R}^n$ is described by a parameterization $\boldsymbol{r}(t) = (x_1(t), \ldots, x_n(t))$ for $t \in [a, b]$. The curve starts at $\boldsymbol{r}(a)$ and ends at $\boldsymbol{r}(b)$.

The **tangent vector** at parameter $t$ is $\boldsymbol{r}'(t) = (x_1'(t), \ldots, x_n'(t))$, and the **arc length element** is:

$$
ds = \|\boldsymbol{r}'(t)\|\,dt = \sqrt{(x_1'(t))^2 + \cdots + (x_n'(t))^2}\,dt
$$

**Common parameterizations:**

| Curve | Parameterization | Domain |
|---|---|---|
| Line segment from $\boldsymbol{A}$ to $\boldsymbol{B}$ | $\boldsymbol{r}(t) = (1-t)\boldsymbol{A} + t\boldsymbol{B}$ | $t \in [0,1]$ |
| Circle of radius $r$ in $xy$-plane | $\boldsymbol{r}(t) = (r\cos t, r\sin t, 0)$ | $t \in [0, 2\pi]$ |
| Helix | $\boldsymbol{r}(t) = (\cos t, \sin t, t)$ | $t \in [0, 2\pi k]$ |

## 2. Line Integrals of Scalar Fields

**Definition.** The **line integral** of a scalar field $f : \mathbb{R}^n \to \mathbb{R}$ over a curve $C$ parameterized by $\boldsymbol{r}(t)$, $t \in [a,b]$, is:

$$
\int_C f\,ds = \int_a^b f(\boldsymbol{r}(t))\,\|\boldsymbol{r}'(t)\|\,dt
$$

**Geometric meaning:** integrate the values of $f$ along the curve, weighting each piece by its arc length. If $f = 1$, the result is the total arc length of $C$.

**Worked Example 2.1.** Let $f(x,y) = x + y$ and $C$ be the line segment from $(0,0)$ to $(1,2)$.

Parameterize: $\boldsymbol{r}(t) = (t, 2t)$, $t \in [0,1]$. Then $\boldsymbol{r}'(t) = (1, 2)$ and $\|\boldsymbol{r}'(t)\| = \sqrt{1^2 + 2^2} = \sqrt{5}$.

$$
\int_C f\,ds = \int_0^1 (t + 2t)\cdot\sqrt{5}\,dt = \sqrt{5}\int_0^1 3t\,dt = \sqrt{5}\cdot\frac{3}{2} = \frac{3\sqrt{5}}{2}
$$

**Worked Example 2.2.** Arc length of the helix $\boldsymbol{r}(t) = (\cos t, \sin t, t)$ for $t \in [0, 2\pi]$.

$\boldsymbol{r}'(t) = (-\sin t, \cos t, 1)$, $\|\boldsymbol{r}'(t)\| = \sqrt{\sin^2 t + \cos^2 t + 1} = \sqrt{2}$.

$$
\text{Arc length} = \int_0^{2\pi} \sqrt{2}\,dt = 2\pi\sqrt{2}
$$

## 3. Line Integrals of Vector Fields

### 3.1 Definition

**Definition.** The **line integral of a vector field** $\boldsymbol{F} : \mathbb{R}^n \to \mathbb{R}^n$ over an oriented curve $C$ is:

$$
\int_C \boldsymbol{F} \cdot d\boldsymbol{r} = \int_a^b \boldsymbol{F}(\boldsymbol{r}(t)) \cdot \boldsymbol{r}'(t)\,dt
$$

In $\mathbb{R}^3$ with $\boldsymbol{F} = (P, Q, R)$ and $d\boldsymbol{r} = (dx, dy, dz)$, this is often written $\int_C P\,dx + Q\,dy + R\,dz$.

**Physical meaning:** the work done by force field $\boldsymbol{F}$ on a particle moving along $C$. The dot product $\boldsymbol{F} \cdot \boldsymbol{r}'(t)$ picks out the component of the force in the direction of motion — components perpendicular to the path do no work.

**Orientation matters.** Reversing the direction of $C$ (from $\boldsymbol{r}(b)$ to $\boldsymbol{r}(a)$) changes the sign of the integral: $\int_{-C} \boldsymbol{F} \cdot d\boldsymbol{r} = -\int_C \boldsymbol{F} \cdot d\boldsymbol{r}$. This is unlike the scalar line integral, which is independent of orientation.

### 3.2 Worked Examples

**Worked Example 3.1.** Let $\boldsymbol{F}(x,y) = (y, -x)$ and $C$ be the upper semicircle from $(1,0)$ to $(-1,0)$.

Parameterize: $\boldsymbol{r}(t) = (\cos t, \sin t)$, $t \in [0, \pi]$. Then $\boldsymbol{r}'(t) = (-\sin t, \cos t)$.

$\boldsymbol{F}(\boldsymbol{r}(t)) = (\sin t, -\cos t)$.

$$
\int_C \boldsymbol{F} \cdot d\boldsymbol{r} = \int_0^{\pi} (\sin t, -\cos t) \cdot (-\sin t, \cos t)\,dt = \int_0^{\pi} (-\sin^2 t - \cos^2 t)\,dt = \int_0^{\pi} (-1)\,dt = -\pi
$$

**Worked Example 3.2.** Let $\boldsymbol{F}(x,y,z) = (yz, xz, xy)$ and $C$ be the line segment from $(0,0,0)$ to $(1,2,3)$.

Parameterize: $\boldsymbol{r}(t) = (t, 2t, 3t)$, $t \in [0,1]$. Then $\boldsymbol{r}'(t) = (1, 2, 3)$.

$\boldsymbol{F}(\boldsymbol{r}(t)) = ((2t)(3t),\, (t)(3t),\, (t)(2t)) = (6t^2, 3t^2, 2t^2)$.

$$
\int_C \boldsymbol{F} \cdot d\boldsymbol{r} = \int_0^1 (6t^2, 3t^2, 2t^2) \cdot (1, 2, 3)\,dt = \int_0^1 (6t^2 + 6t^2 + 6t^2)\,dt = 18\int_0^1 t^2\,dt = 6
$$

We will see in Section 4 that this field is conservative — so the answer depends only on the endpoints.

## 4. Conservative Fields and the Fundamental Theorem

### 4.1 The Fundamental Theorem of Line Integrals

**Theorem.** Let $f : \mathbb{R}^n \to \mathbb{R}$ be a differentiable scalar field and let $C$ be a smooth curve from $\boldsymbol{A}$ to $\boldsymbol{B}$. Then:

$$
\int_C \nabla f \cdot d\boldsymbol{r} = f(\boldsymbol{B}) - f(\boldsymbol{A})
$$

**Proof.** By the chain rule (Module 03):

$$
\int_C \nabla f \cdot d\boldsymbol{r} = \int_a^b \nabla f(\boldsymbol{r}(t)) \cdot \boldsymbol{r}'(t)\,dt = \int_a^b \frac{d}{dt}[f(\boldsymbol{r}(t))]\,dt = f(\boldsymbol{r}(b)) - f(\boldsymbol{r}(a)) = f(\boldsymbol{B}) - f(\boldsymbol{A})
$$

where the second equality is the chain rule for $f \circ \boldsymbol{r}$, and the third is the fundamental theorem of calculus from Module 00 Unit 03. $\square$

**Consequence.** For any gradient field $\boldsymbol{F} = \nabla f$, the line integral depends only on the values of $f$ at the start and end points — not on the path taken between them.

### 4.2 Conservative Fields

**Definition.** A vector field $\boldsymbol{F}$ on a simply connected domain is **conservative** if any of the following equivalent conditions hold:

1. $\boldsymbol{F} = \nabla f$ for some scalar field $f$ (called the **potential function** of $\boldsymbol{F}$)
2. $\int_C \boldsymbol{F} \cdot d\boldsymbol{r}$ is **path-independent** — depends only on the endpoints
3. $\oint_C \boldsymbol{F} \cdot d\boldsymbol{r} = 0$ for every closed curve $C$
4. $\nabla \times \boldsymbol{F} = \boldsymbol{0}$ everywhere (**irrotational**)

Conditions 1–3 are equivalent in any domain. Condition 4 implies 1–3 on simply connected domains (no holes); on non-simply connected domains, irrotational does not guarantee conservative.

### 4.3 Finding the Potential Function

Given a conservative field $\boldsymbol{F} = (P, Q, R)$, find $f$ such that $\nabla f = \boldsymbol{F}$.

**Method:** integrate $P$ with respect to $x$, then differentiate with respect to $y$ and match to $Q$, then match to $R$.

**Worked Example 4.1.** Show that $\boldsymbol{F}(x,y,z) = (yz, xz, xy)$ is conservative and find its potential.

We already showed $\nabla \times \boldsymbol{F} = \boldsymbol{0}$ in Example 4.3 of Unit 01. So $\boldsymbol{F}$ is conservative.

Find $f$ such that $\nabla f = (yz, xz, xy)$:

**Step 1.** $\frac{\partial f}{\partial x} = yz \implies f = xyz + g(y,z)$ for some function $g$ that does not depend on $x$.

**Step 2.** $\frac{\partial f}{\partial y} = xz + \frac{\partial g}{\partial y} = xz \implies \frac{\partial g}{\partial y} = 0 \implies g = h(z)$.

**Step 3.** $\frac{\partial f}{\partial z} = xy + h'(z) = xy \implies h'(z) = 0 \implies h = C$ (constant).

Therefore $f(x,y,z) = xyz + C$. Verify: $\nabla(xyz) = (yz, xz, xy) = \boldsymbol{F}$. ✓

**Immediate verification of Example 3.2** from the fundamental theorem:

$$
\int_C (yz, xz, xy) \cdot d\boldsymbol{r} = f(1,2,3) - f(0,0,0) = (1)(2)(3) - 0 = 6 \checkmark
$$

## 5. Green's Theorem

Green's theorem relates a line integral over a closed plane curve to a double integral over the region it encloses.

**Theorem (Green's Theorem).** Let $D$ be a bounded region in $\mathbb{R}^2$ with positively oriented (counterclockwise) boundary $\partial D$. Let $P$ and $Q$ have continuous partial derivatives on $D$. Then:

$$
\oint_{\partial D} P\,dx + Q\,dy = \iint_D \left(\frac{\partial Q}{\partial x} - \frac{\partial P}{\partial y}\right)\,dA
$$

The integrand on the right is the $z$-component of $\nabla \times \boldsymbol{F}$ (the curl$_z$), so Green's theorem states:

$$
\oint_{\partial D} \boldsymbol{F} \cdot d\boldsymbol{r} = \iint_D (\nabla \times \boldsymbol{F})_z\,dA
$$

**Why it works.** The boundary integral measures the net circulation of $\boldsymbol{F}$ around $D$. The curl measures the local rotation rate at each interior point. Green's theorem says that summing up all the local rotations (the double integral) equals the net boundary circulation — because interior contributions cancel out between adjacent sub-regions.

**Conservative fields and Green's theorem.** If $\boldsymbol{F}$ is conservative ($\nabla \times \boldsymbol{F} = \boldsymbol{0}$), then the double integral is zero and so is the closed line integral — consistent with condition 3 in Section 4.2.

**Worked Example 5.1.** Evaluate $\oint_C x^2 y\,dx + x\,dy$ where $C$ is the triangle with vertices $(0,0)$, $(1,0)$, $(1,1)$, oriented counterclockwise.

Here $P = x^2 y$, $Q = x$. The region $D$ is the triangle $\{0 \leq x \leq 1,\, 0 \leq y \leq x\}$.

$$
\frac{\partial Q}{\partial x} - \frac{\partial P}{\partial y} = 1 - x^2
$$

$$
\iint_D (1 - x^2)\,dA = \int_0^1 \int_0^x (1 - x^2)\,dy\,dx = \int_0^1 (1-x^2)\cdot x\,dx = \int_0^1 (x - x^3)\,dx = \frac{1}{2} - \frac{1}{4} = \frac{1}{4}
$$

## 6. Statistical Bridge — Gradient Flow as a Conservative Field `[MLE]` · `[GLM]`

### 6.1 Gradient Descent as Path Integration

In gradient descent, the parameter trajectory $\boldsymbol{\theta}(t)$ satisfies the continuous-time gradient flow ODE:

$$
\frac{d\boldsymbol{\theta}}{dt} = -\nabla_{\boldsymbol{\theta}}\,J(\boldsymbol{\theta})
$$

The velocity vector at each point is $-\nabla J$ — the negative gradient. The total change in loss along this trajectory from $\boldsymbol{\theta}(0) = \boldsymbol{\theta}_0$ to $\boldsymbol{\theta}(T) = \boldsymbol{\theta}_T$ is:

$$
J(\boldsymbol{\theta}_T) - J(\boldsymbol{\theta}_0) = \int_C \nabla J \cdot d\boldsymbol{\theta}
$$

where $C$ is the trajectory curve. By the fundamental theorem of line integrals:

$$
J(\boldsymbol{\theta}_T) - J(\boldsymbol{\theta}_0) = J(\boldsymbol{\theta}_T) - J(\boldsymbol{\theta}_0)
$$

This is a tautology — but the content is that **the total loss change is path-independent**. No matter how complicated the gradient descent trajectory is, the total change in loss between two points in parameter space is the same as if you had taken the straight-line path or any other path. Only the start and end points matter.

### 6.2 The Loss as a Potential Function

The gradient field $\boldsymbol{F}(\boldsymbol{\theta}) = \nabla J(\boldsymbol{\theta})$ is conservative (trivially — it is a gradient). The potential function is $J$ itself. In physics, the potential energy is defined with a sign convention: $\boldsymbol{F} = -\nabla V$ (force = negative gradient of potential). Gradient descent follows this convention: the "force" is $-\nabla J$, and $J$ is the potential.

**Consequences:**
- $\nabla \times (\nabla J) = \boldsymbol{0}$: the gradient field is irrotational. Gradient descent trajectories never "spiral" — they always flow downhill without circling back
- The loss decreases monotonically along the gradient flow: $\frac{d}{dt}J(\boldsymbol{\theta}(t)) = \nabla J \cdot \dot{\boldsymbol{\theta}} = -\|\nabla J\|^2 \leq 0$
- Equilibria ($\dot{\boldsymbol{\theta}} = \boldsymbol{0}$) occur exactly at critical points of $J$ — where $\nabla J = \boldsymbol{0}$

### 6.3 Path Independence and Optimizer Choice

Since the total loss change between two parameter points is path-independent, what distinguishes different optimizers (SGD, Adam, RMSProp, L-BFGS) is not how much total loss they traverse, but **how quickly** they reach a given endpoint and **which endpoint** they converge to. The choice of optimizer determines the path; the path determines neither the total loss change (fixed by endpoints) nor whether a minimum is reached (determined by the geometry of the loss landscape).

### Section 1 — Symbolic Line Integral Verification

Verify the worked examples symbolically using SymPy, and confirm path independence for the conservative field by computing the integral along two different paths between the same endpoints.

```python
# ============================================================
# Section 1 — Symbolic line integral verification
# ============================================================
t = sp.Symbol('t', real=True)
x, y, z = sp.symbols('x y z', real=True)

# --- Example 2.1: f(x,y)=x+y along segment (0,0)→(1,2) ---
r1 = sp.Matrix([t, 2*t])
r1p = sp.diff(r1, t)
ds1 = sp.sqrt(r1p.dot(r1p))
f1 = r1[0] + r1[1]   # x + y evaluated on curve
result_21 = sp.integrate(f1 * ds1, (t, 0, 1))
print('Example 2.1  ∫_C (x+y) ds:')
print('  Result:', result_21, '= 3√5/2?', sp.simplify(result_21 - 3*sp.sqrt(5)/2) == 0)

# --- Example 3.1: F=(y,-x) around upper semicircle ---
r31 = sp.Matrix([cos(t), sin(t)])
r31p = sp.diff(r31, t)
F31 = sp.Matrix([r31[1], -r31[0]])   # F = (y, -x) on curve
integrand_31 = F31.dot(r31p)
result_31 = sp.integrate(integrand_31, (t, 0, pi))
print('\nExample 3.1  ∫_C (y,-x)·dr over upper semicircle:')
print('  Result:', result_31, '= -π?', sp.simplify(result_31 + pi) == 0)

# --- Example 3.2: F=(yz,xz,xy) along (0,0,0)→(1,2,3) ---
r32 = sp.Matrix([t, 2*t, 3*t])
r32p = sp.diff(r32, t)
F32 = sp.Matrix([r32[1]*r32[2], r32[0]*r32[2], r32[0]*r32[1]])  # (yz,xz,xy) on curve
result_32 = sp.integrate(F32.dot(r32p), (t, 0, 1))
print('\nExample 3.2  ∫_C (yz,xz,xy)·dr, straight line:')
print('  Result:', result_32, '= 6?', result_32 == 6)

# --- Path independence: different path, same endpoints ---
# Curved path: r(t) = (t, 2t², 3t³), t ∈ [0,1]
r32b = sp.Matrix([t, 2*t**2, 3*t**3])
r32bp = sp.diff(r32b, t)
F32b = sp.Matrix([r32b[1]*r32b[2], r32b[0]*r32b[2], r32b[0]*r32b[1]])
result_32b = sp.integrate(F32b.dot(r32bp), (t, 0, 1))
print('\nSame field, curved path r=(t, 2t², 3t³):')
print('  Result:', result_32b, '= 6?', result_32b == 6)
print('  Path independence confirmed:', result_32 == result_32b)

# --- Green's theorem Example 5.1 ---
result_green = sp.integrate(sp.integrate(1 - x**2, (y, 0, x)), (x, 0, 1))
print('\nGreen\'s theorem Example 5.1  ∫∫(1-x²)dA over triangle:')
print('  Result:', result_green, '= 1/4?', result_green == sp.Rational(1,4))
```

**What do you see?** All five results match the hand calculations. The path independence test is the key result: the integral of the conservative field $(yz, xz, xy)$ from $(0,0,0)$ to $(1,2,3)$ is $6$ along both the straight line and the curved path $(t, 2t^2, 3t^3)$. The fundamental theorem of line integrals guarantees this — both equal $f(1,2,3) - f(0,0,0) = (1)(2)(3) - 0 = 6$.

### Section 2 — Path Independence Visualization

Show three different paths between the same two points in a conservative field and confirm all three line integrals give the same value. Overlay the potential function contours to show the result depends only on the values at the endpoints.

```python
# ============================================================
# Section 2 — Path independence visualization
# ============================================================
# Conservative field: F = ∇f where f(x,y) = x²y
# F = (2xy, x²)

def F_field(x, y):
    return np.array([2*x*y, x**2])

def potential(x, y):
    return x**2 * y

def line_integral_F(path_x, path_y):
    """Numerical line integral ∫ F·dr via trapezoidal rule."""
    dx = np.diff(path_x)
    dy = np.diff(path_y)
    x_mid = 0.5*(path_x[:-1] + path_x[1:])
    y_mid = 0.5*(path_y[:-1] + path_y[1:])
    Fx = 2 * x_mid * y_mid
    Fy = x_mid**2
    return np.sum(Fx*dx + Fy*dy)

# Endpoints
A = np.array([0.5, 0.5])
B = np.array([2.5, 2.0])
exact = potential(*B) - potential(*A)

# Three different paths
N = 1000
t_arr = np.linspace(0, 1, N)

# Path 1: straight line
px1 = A[0] + t_arr*(B[0]-A[0])
py1 = A[1] + t_arr*(B[1]-A[1])

# Path 2: arc through top
mid_top = np.array([1.5, 3.2])
px2 = (1-t_arr)**2*A[0] + 2*t_arr*(1-t_arr)*mid_top[0] + t_arr**2*B[0]
py2 = (1-t_arr)**2*A[1] + 2*t_arr*(1-t_arr)*mid_top[1] + t_arr**2*B[1]

# Path 3: arc through bottom
mid_bot = np.array([1.5, 0.2])
px3 = (1-t_arr)**2*A[0] + 2*t_arr*(1-t_arr)*mid_bot[0] + t_arr**2*B[0]
py3 = (1-t_arr)**2*A[1] + 2*t_arr*(1-t_arr)*mid_bot[1] + t_arr**2*B[1]

I1 = line_integral_F(px1, py1)
I2 = line_integral_F(px2, py2)
I3 = line_integral_F(px3, py3)

print('Exact  f(B) - f(A) =', exact)
print(f'Path 1 (straight): {I1:.6f}')
print(f'Path 2 (arc high): {I2:.6f}')
print(f'Path 3 (arc low):  {I3:.6f}')

# Visualization
xg = np.linspace(0, 3, 200)
yg = np.linspace(0, 3, 200)
Xg, Yg = np.meshgrid(xg, yg)
Fg = potential(Xg, Yg)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Left: potential contours + three paths
ax = axes[0]
cf = ax.contourf(Xg, Yg, Fg, levels=20, cmap='Blues', alpha=0.55)
cs = ax.contour(Xg, Yg, Fg, levels=20, colors=[TS_GRAY], alpha=0.4, linewidths=0.7)
plt.colorbar(cf, ax=ax, label='$f(x,y) = x^2 y$ (potential)')

paths = [(px1,py1,TS_BLUE,'Path 1 (straight)'),
         (px2,py2,TS_AMBER,'Path 2 (arc high)'),
         (px3,py3,TS_GREEN,'Path 3 (arc low)')]
for px, py, col, lab in paths:
    ax.plot(px, py, color=col, lw=2.2, label=f'{lab}  I={line_integral_F(px,py):.4f}')
ax.plot(*A, 'o', color=TS_RED, markersize=11, zorder=6)
ax.plot(*B, 's', color=TS_RED, markersize=11, zorder=6)
ax.text(A[0]-0.2, A[1]-0.15, 'A', fontsize=12, color=TS_RED, fontweight='bold')
ax.text(B[0]+0.05, B[1]+0.1, 'B', fontsize=12, color=TS_RED, fontweight='bold')
ax.set_xlabel('x'); ax.set_ylabel('y')
ax.set_title('Path independence: all three integrals = $f(B)-f(A)$')
ax.legend(fontsize=8)
ax.set_aspect('equal')

# Right: gradient field
ax2 = axes[1]
xq = np.linspace(0.1, 3, 14)
yq = np.linspace(0.1, 3, 14)
Xq, Yq = np.meshgrid(xq, yq)
Gx = 2*Xq*Yq; Gy = Xq**2
norm = np.sqrt(Gx**2 + Gy**2) + 1e-8
ax2.contourf(Xg, Yg, Fg, levels=20, cmap='Blues', alpha=0.45)
ax2.quiver(Xq, Yq, Gx/norm, Gy/norm, color=TS_AMBER, alpha=0.7,
           scale=22, width=0.004)
ax2.set_xlabel('x'); ax2.set_ylabel('y')
ax2.set_title(r'Gradient field $\nabla f = (2xy, x^2)$ — conservative, irrotational')
ax2.set_aspect('equal')

plt.suptitle('Conservative field $\\mathbf{F} = \\nabla(x^2 y)$ — path-independent line integrals',
             fontsize=11, y=1.01)
plt.tight_layout()
plt.show()
```

**What do you see?** All three paths give the same line integral to six decimal places — equal to $f(B) - f(A) = (2.5)^2(2.0) - (0.5)^2(0.5) = 12.5 - 0.125 = 12.375$. The potential contours show why: the line integral measures the total climb in the potential function from $A$ to $B$, regardless of the route taken. The gradient field (right panel) points in the direction of steepest ascent of $f$ at every point — this is exactly the conservative field structure.

### Section 3 — Statistical Bridge: Gradient Flow Trajectories

Simulate continuous-time gradient flow on a 2D loss surface, show that the total loss change along each trajectory equals $J(\boldsymbol{\theta}_T) - J(\boldsymbol{\theta}_0)$ regardless of the learning rate used, and visualize trajectories as paths in the conservative field.

```python
# ============================================================
# Section 3 — Gradient flow as conservative field
# ============================================================
# Loss: J(θ₁,θ₂) = (θ₁-1)² + 2(θ₂-0.5)² + 0.5·θ₁·θ₂
# ∇J = (2(θ₁-1) + 0.5θ₂, 4(θ₂-0.5) + 0.5θ₁)

def J_loss(th1, th2):
    return (th1-1)**2 + 2*(th2-0.5)**2 + 0.5*th1*th2

def grad_J(th):
    th1, th2 = th
    return np.array([2*(th1-1) + 0.5*th2, 4*(th2-0.5) + 0.5*th1])

def gradient_flow(th0, lr, n_steps):
    path = [np.array(th0, dtype=float)]
    th = np.array(th0, dtype=float)
    for _ in range(n_steps):
        th = th - lr * grad_J(th)
        path.append(th.copy())
    return np.array(path)

th0_A = np.array([-1.5, 2.0])
th0_B = np.array([3.0, -0.8])

# Three different learning rates — different paths, different endpoints
configs = [
    (th0_A, 0.05, 80,  TS_BLUE,  'Start A, α=0.05'),
    (th0_A, 0.15, 80,  TS_AMBER, 'Start A, α=0.15'),
    (th0_B, 0.08, 80,  TS_GREEN, 'Start B, α=0.08'),
]

# Loss surface grid
t1g = np.linspace(-2.5, 3.5, 200)
t2g = np.linspace(-1.5, 2.5, 200)
T1, T2 = np.meshgrid(t1g, t2g)
ZJ = J_loss(T1, T2)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

ax = axes[0]
ax.contourf(T1, T2, ZJ, levels=np.linspace(0, 15, 25), cmap='Blues', alpha=0.45)
ax.contour(T1, T2, ZJ, levels=np.linspace(0, 15, 25), colors=[TS_GRAY],
           alpha=0.4, linewidths=0.7)

for th0, lr, n_steps, col, label in configs:
    path = gradient_flow(th0, lr, n_steps)
    dJ = J_loss(*path[-1]) - J_loss(*th0)
    line_int = sum(
        np.dot(0.5*(grad_J(path[i]) + grad_J(path[i+1])),
               path[i+1] - path[i])
        for i in range(len(path)-1)
    )
    ax.plot(path[:,0], path[:,1], 'o-', color=col, markersize=2,
            lw=1.8, alpha=0.85,
            label=f'{label}\n  ΔJ={dJ:.3f}, ∫∇J·dθ={line_int:.3f}')
    ax.plot(*th0, 's', color=col, markersize=9, zorder=6)
    ax.plot(*path[-1], '*', color=col, markersize=11, zorder=6)

ax.set_xlabel(r'$\theta_1$'); ax.set_ylabel(r'$\theta_2$')
ax.set_title('Gradient flow trajectories on loss surface')
ax.legend(fontsize=7, loc='upper right')

# Right: loss vs iteration for each trajectory
ax2 = axes[1]
for th0, lr, n_steps, col, label in configs:
    path = gradient_flow(th0, lr, n_steps)
    loss_vals = [J_loss(*th) for th in path]
    ax2.plot(loss_vals, color=col, lw=2, label=label.split(',')[0] + f', α={lr}')
    ax2.axhline(loss_vals[-1], color=col, ls=':', lw=1.0, alpha=0.6)

ax2.set_xlabel('Iteration'); ax2.set_ylabel('$J(\\theta)$')
ax2.set_title('Loss is monotonically decreasing:\n'
              r'$\frac{d}{dt}J(\theta(t)) = -\|\nabla J\|^2 \leq 0$')
ax2.legend(fontsize=9)

plt.suptitle('Gradient descent = path in a conservative field  ($\\nabla J$ is the field)',
             fontsize=11, y=1.01)
plt.tight_layout()
plt.show()
```

**What do you see?**

- **Left**: Three gradient descent trajectories from two different starting points. Each follows a different path, but the legend confirms that the line integral $\int \nabla J \cdot d\boldsymbol{\theta}$ equals $\Delta J$ numerically — the fundamental theorem at work.
- **Right**: Loss is monotonically non-increasing on every trajectory. This is guaranteed by $\frac{d}{dt}J = \nabla J \cdot \dot{\boldsymbol{\theta}} = -\|\nabla J\|^2 \leq 0$. The loss can only decrease or stay flat — it cannot increase along gradient flow.
- Different learning rates change the path, but not the key property: the change in loss between start and end is path-independent.

## Summary

| Concept | Statement |
|---|---|
| Scalar line integral | $\int_C f\,ds = \int_a^b f(\boldsymbol{r}(t))\|\boldsymbol{r}'(t)\|\,dt$ |
| Vector line integral | $\int_C \boldsymbol{F}\cdot d\boldsymbol{r} = \int_a^b \boldsymbol{F}(\boldsymbol{r}(t))\cdot\boldsymbol{r}'(t)\,dt$ |
| FT of line integrals | $\int_C \nabla f\cdot d\boldsymbol{r} = f(\boldsymbol{B}) - f(\boldsymbol{A})$ |
| Conservative field | $\boldsymbol{F} = \nabla f$ iff path-independent iff $\oint = 0$ iff $\nabla\times\boldsymbol{F}=\boldsymbol{0}$ |
| Potential recovery | Integrate component by component, matching cross-partials |
| Green's theorem | $\oint_{\partial D} P\,dx + Q\,dy = \iint_D(\partial_x Q - \partial_y P)\,dA$ |
| Gradient flow | $\dot{\boldsymbol{\theta}} = -\nabla J$ → $\frac{d}{dt}J = -\|\nabla J\|^2 \leq 0$ — loss is monotone |
| Loss as potential | $J = $ potential; total loss change = path-independent |
