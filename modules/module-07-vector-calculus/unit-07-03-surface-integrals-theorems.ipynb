## Why This Unit

Units 01 and 02 built the differential operators (divergence, curl) and the fundamental theorem of line integrals. This unit assembles them into the two major theorems of vector calculus: **Stokes' theorem** (relating a surface integral to a line integral around its boundary) and the **divergence theorem** (relating a volume integral to a surface integral over the boundary).

Both theorems are instances of a single idea: integration over a domain can be traded for integration over its boundary. This idea — the generalized Stokes' theorem — is one of the deepest in mathematics, and its analogue in probability theory underlies the evidence lower bound (ELBO) that makes variational inference tractable.

## 1. Surfaces and Surface Area Elements

### 1.1 Parameterized Surfaces

A **surface** $S$ in $\mathbb{R}^3$ is parameterized by $\boldsymbol{r}(u, v) = (x(u,v), y(u,v), z(u,v))$ for $(u,v) \in D \subset \mathbb{R}^2$.

**Common surfaces:**

| Surface | Parameterization |
|---|---|
| Sphere radius $a$ | $\boldsymbol{r}(\phi, \theta) = (a\sin\phi\cos\theta, a\sin\phi\sin\theta, a\cos\phi)$, $(\phi,\theta) \in [0,\pi]\times[0,2\pi]$ |
| Cylinder radius $a$, height $h$ | $\boldsymbol{r}(\theta, z) = (a\cos\theta, a\sin\theta, z)$, $(\theta,z) \in [0,2\pi]\times[0,h]$ |
| Graph $z = f(x,y)$ | $\boldsymbol{r}(x,y) = (x, y, f(x,y))$ |

### 1.2 The Surface Normal and Area Element

At each point of the surface, the two tangent vectors are:

$$
\boldsymbol{r}_u = \frac{\partial \boldsymbol{r}}{\partial u} \qquad \boldsymbol{r}_v = \frac{\partial \boldsymbol{r}}{\partial v}
$$

Their cross product gives a vector **normal** to the surface:

$$
\boldsymbol{r}_u \times \boldsymbol{r}_v
$$

The **scalar surface area element** is $dS = \|\boldsymbol{r}_u \times \boldsymbol{r}_v\|\,du\,dv$, and the **vector surface element** (with outward normal encoded) is:

$$
d\boldsymbol{S} = (\boldsymbol{r}_u \times \boldsymbol{r}_v)\,du\,dv
$$

**Surface area** of $S$: $\iint_D \|\boldsymbol{r}_u \times \boldsymbol{r}_v\|\,du\,dv$.

**For a graph** $z = f(x,y)$: $\boldsymbol{r}(x,y) = (x, y, f(x,y))$, so $\boldsymbol{r}_x = (1, 0, f_x)$ and $\boldsymbol{r}_y = (0, 1, f_y)$, giving:

$$
\boldsymbol{r}_x \times \boldsymbol{r}_y = (-f_x, -f_y, 1) \qquad \|\boldsymbol{r}_x \times \boldsymbol{r}_y\| = \sqrt{1 + f_x^2 + f_y^2}
$$

**Worked Example 1.1.** Surface area of the sphere of radius $a$.

Using spherical parameterization $\boldsymbol{r}(\phi, \theta)$ with $\phi \in [0,\pi]$, $\theta \in [0, 2\pi]$:

$$
\|\boldsymbol{r}_\phi \times \boldsymbol{r}_\theta\| = a^2\sin\phi
$$

(This is the spherical Jacobian factor $\rho^2\sin\phi$ from Module 05 Unit 02, evaluated at $\rho = a$.)

$$
\text{Surface area} = \int_0^{2\pi}\int_0^{\pi} a^2\sin\phi\,d\phi\,d\theta = 2\pi a^2 \cdot [-\cos\phi]_0^{\pi} = 2\pi a^2 \cdot 2 = 4\pi a^2
$$

## 2. Flux Integrals

**Definition.** The **flux** of a vector field $\boldsymbol{F}$ through an oriented surface $S$ is:

$$
\iint_S \boldsymbol{F} \cdot d\boldsymbol{S} = \iint_D \boldsymbol{F}(\boldsymbol{r}(u,v)) \cdot (\boldsymbol{r}_u \times \boldsymbol{r}_v)\,du\,dv
$$

**Physical meaning:** flux measures how much of the field flows through the surface per unit time. The dot product $\boldsymbol{F} \cdot \boldsymbol{n}$ picks out the component of $\boldsymbol{F}$ perpendicular to the surface — components parallel to the surface contribute zero flux.

**Worked Example 2.1.** Compute the flux of $\boldsymbol{F}(x,y,z) = (x, y, z)$ (radial field) outward through the sphere $x^2 + y^2 + z^2 = a^2$.

On the sphere, $\boldsymbol{r} = a\hat{\boldsymbol{r}}$ and the outward unit normal is $\hat{\boldsymbol{n}} = \hat{\boldsymbol{r}} = (x,y,z)/a$. So $\boldsymbol{F} \cdot \hat{\boldsymbol{n}} = a$.

$$
\iint_S \boldsymbol{F} \cdot d\boldsymbol{S} = a \cdot \text{(surface area)} = a \cdot 4\pi a^2 = 4\pi a^3
$$

We will verify this using the divergence theorem in Section 3.

## 3. The Divergence Theorem

**Theorem (Divergence Theorem / Gauss's Theorem).** Let $V$ be a bounded solid region in $\mathbb{R}^3$ with outward-oriented boundary surface $\partial V$. Let $\boldsymbol{F}$ be a differentiable vector field on $V$. Then:

$$
\boxed{\oiint_{\partial V} \boldsymbol{F} \cdot d\boldsymbol{S} = \iiint_V \nabla \cdot \boldsymbol{F}\,dV}
$$

**Reading the theorem:** the total outward flux through the boundary surface equals the total divergence (net source) inside the volume.

**Intuition:** divergence at a point measures the local rate of outward flow per unit volume. Summing this over the entire volume (the right side) gives the total net outflow — which must equal the net flux across the boundary (the left side).

**Verification of Example 2.1.** For $\boldsymbol{F}(x,y,z) = (x, y, z)$: $\nabla \cdot \boldsymbol{F} = 1 + 1 + 1 = 3$.

$$
\iiint_V \nabla \cdot \boldsymbol{F}\,dV = 3 \cdot \text{Vol}(\text{ball}) = 3 \cdot \frac{4}{3}\pi a^3 = 4\pi a^3 \checkmark
$$

**Worked Example 3.1.** Compute the flux of $\boldsymbol{F}(x,y,z) = (x^3, y^3, z^3)$ outward through the unit sphere.

$\nabla \cdot \boldsymbol{F} = 3x^2 + 3y^2 + 3z^2 = 3r^2$ in spherical coordinates.

$$
\iiint_V 3r^2\,dV = 3\int_0^{2\pi}\int_0^{\pi}\int_0^1 r^2 \cdot r^2\sin\phi\,dr\,d\phi\,d\theta = 3 \cdot 2\pi \cdot 2 \cdot \frac{1}{5} = \frac{12\pi}{5}
$$

(Much easier than evaluating the surface integral directly.)

## 4. Stokes' Theorem

**Theorem (Stokes' Theorem).** Let $S$ be an oriented surface in $\mathbb{R}^3$ with positively oriented boundary curve $\partial S$. Let $\boldsymbol{F}$ be a differentiable vector field on $S$. Then:

$$
\boxed{\oint_{\partial S} \boldsymbol{F} \cdot d\boldsymbol{r} = \iint_S (\nabla \times \boldsymbol{F}) \cdot d\boldsymbol{S}}
$$

**Reading the theorem:** the line integral of $\boldsymbol{F}$ around the boundary of $S$ equals the flux of the curl of $\boldsymbol{F}$ through $S$.

**Stokes' theorem generalizes Green's theorem.** For a flat surface $S$ in the $xy$-plane with boundary $C$, the normal is $\hat{\boldsymbol{k}} = (0,0,1)$ and $(\nabla \times \boldsymbol{F}) \cdot \hat{\boldsymbol{k}} = \partial_x Q - \partial_y P$. The theorem reduces to:

$$
\oint_C \boldsymbol{F} \cdot d\boldsymbol{r} = \iint_S (\partial_x Q - \partial_y P)\,dA
$$

which is exactly Green's theorem from Unit 02.

**Conservative fields and Stokes' theorem.** If $\boldsymbol{F} = \nabla f$ (conservative), then $\nabla \times \boldsymbol{F} = \boldsymbol{0}$ (from Unit 01), so $\iint_S (\nabla \times \boldsymbol{F}) \cdot d\boldsymbol{S} = 0$, and therefore $\oint_{\partial S} \boldsymbol{F} \cdot d\boldsymbol{r} = 0$ for any closed curve. This is consistent with path independence.

**Worked Example 4.1.** Verify Stokes' theorem for $\boldsymbol{F} = (y, z, x)$ and $S$ the upper hemisphere $x^2+y^2+z^2=1$, $z \geq 0$, with boundary $C$ the unit circle in the $xy$-plane.

**Line integral over $C$:** parameterize $\boldsymbol{r}(t) = (\cos t, \sin t, 0)$, $t \in [0, 2\pi]$.

$\boldsymbol{F}(\boldsymbol{r}(t)) = (\sin t, 0, \cos t)$, $\boldsymbol{r}'(t) = (-\sin t, \cos t, 0)$.

$$
\oint_C \boldsymbol{F} \cdot d\boldsymbol{r} = \int_0^{2\pi} (\sin t, 0, \cos t) \cdot (-\sin t, \cos t, 0)\,dt = \int_0^{2\pi} (-\sin^2 t)\,dt = -\pi
$$

**Curl of $\boldsymbol{F}$:** with $P=y$, $Q=z$, $R=x$:

$$
\nabla \times \boldsymbol{F} = \begin{pmatrix} \partial_y x - \partial_z z \\ \partial_z y - \partial_x x \\ \partial_x z - \partial_y y \end{pmatrix} = \begin{pmatrix} 0-1 \\ 0-1 \\ 0-1 \end{pmatrix} = (-1, -1, -1)
$$

**Flux of curl through $S$:** the outward (upward) unit normal on the hemisphere has $\boldsymbol{n}\cdot(-1,-1,-1) = -x - y - z$. By symmetry, $\iint_S x\,dS = \iint_S y\,dS = 0$ and $\iint_S z\,dS > 0$. On the unit sphere $\iint_S z\,dS = \pi$ (standard result). Therefore:

$$
\iint_S (\nabla \times \boldsymbol{F}) \cdot d\boldsymbol{S} = -(\underbrace{\iint_S x\,dS}_{0} + \underbrace{\iint_S y\,dS}_{0} + \underbrace{\iint_S z\,dS}_{\pi}) = -\pi \checkmark
$$

## 5. The Generalized Stokes' Theorem

All four major theorems of vector calculus are instances of a single abstract statement:

$$
\int_{\partial M} \omega = \int_M d\omega
$$

where $M$ is an oriented manifold with boundary $\partial M$, $\omega$ is a differential form, and $d\omega$ is its exterior derivative.

| Theorem | $M$ | $\partial M$ | Specialization |
|---|---|---|---|
| FT of Calculus | Interval $[a,b]$ | Points $\{a,b\}$ | $\int_a^b f'\,dx = f(b) - f(a)$ |
| FT of Line Integrals | Curve $C$ | Endpoints $\{A, B\}$ | $\int_C \nabla f \cdot d\boldsymbol{r} = f(B)-f(A)$ |
| Green's Theorem | Region $D$ in $\mathbb{R}^2$ | Boundary curve $\partial D$ | $\iint_D (\partial_x Q - \partial_y P)\,dA = \oint_{\partial D}$ |
| Stokes' Theorem | Surface $S$ | Boundary curve $\partial S$ | $\iint_S (\nabla\times\boldsymbol{F})\cdot d\boldsymbol{S} = \oint_{\partial S}$ |
| Divergence Theorem | Volume $V$ | Boundary surface $\partial V$ | $\iiint_V \nabla\cdot\boldsymbol{F}\,dV = \oiint_{\partial V}$ |

The common thread: **integration over the interior of a domain equals integration of a derivative over its boundary**. This principle extends far beyond $\mathbb{R}^3$ — it underlies differential geometry, topology, and, as we will see, probability theory.

## 6. Statistical Bridge — Divergence Theorem, KL Divergence, and the ELBO `[BAY]`

### 6.1 KL Divergence as Flux

The **Kullback-Leibler divergence** from $q$ to $p$ is:

$$
D_{\text{KL}}(q \| p) = \int q(\boldsymbol{x})\log\frac{q(\boldsymbol{x})}{p(\boldsymbol{x})}\,d\boldsymbol{x} = \mathbb{E}_q\!\left[\log q - \log p\right]
$$

It measures how much probability mass $q$ places in regions where $q$ and $p$ disagree. When $q = p$, $D_{\text{KL}} = 0$; otherwise $D_{\text{KL}} > 0$ (Gibbs' inequality).

### 6.2 The Evidence Lower Bound

In variational inference, we approximate a posterior $p(\boldsymbol{z} \mid \boldsymbol{x})$ by a tractable family $q(\boldsymbol{z})$ (e.g., a Gaussian). We want $q$ close to $p$ in KL divergence. Starting from the log marginal likelihood (the **evidence**):

$$
\log p(\boldsymbol{x}) = \log \int p(\boldsymbol{x}, \boldsymbol{z})\,d\boldsymbol{z}
$$

Multiply and divide by $q(\boldsymbol{z})$ and use Jensen's inequality:

$$
\log p(\boldsymbol{x}) = \log \mathbb{E}_q\!\left[\frac{p(\boldsymbol{x}, \boldsymbol{z})}{q(\boldsymbol{z})}\right] \geq \mathbb{E}_q\!\left[\log\frac{p(\boldsymbol{x}, \boldsymbol{z})}{q(\boldsymbol{z})}\right] \eqqcolon \text{ELBO}(q)
$$

The **ELBO** (Evidence Lower BOund) is a lower bound on the log evidence. Rearranging:

$$
\log p(\boldsymbol{x}) = \text{ELBO}(q) + D_{\text{KL}}(q \| p(\boldsymbol{z} \mid \boldsymbol{x}))
$$

Since $\log p(\boldsymbol{x})$ is fixed (independent of $q$), maximizing the ELBO is equivalent to minimizing $D_{\text{KL}}(q \| p)$. This is what variational autoencoders (VAEs) and other variational models optimize.

### 6.3 The Divergence Theorem Analogy

The divergence theorem says: **net outward flux through a boundary = total divergence inside**.

The KL decomposition says: **log evidence = ELBO + KL gap**.

Both are exact accounting identities. In the divergence theorem, the interior divergence (sources/sinks) accounts for the boundary flux. In variational inference, the KL gap accounts for the difference between the lower bound (ELBO) and the true quantity (log evidence).

More precisely, the score function $\nabla_{\boldsymbol{z}}\log p(\boldsymbol{z})$ is a vector field, and its divergence is related to the Fisher information and the Laplacian of the log-density. The divergence theorem applied to this field over probability regions gives integral identities that underlie score-based generative models and the Stein discrepancy — directly connecting vector calculus to the frontier of probabilistic machine learning.

### 6.4 ELBO Components

Writing the ELBO in terms of its components:

$$
\text{ELBO}(q) = \underbrace{\mathbb{E}_q[\log p(\boldsymbol{x} \mid \boldsymbol{z})]}_{{\text{expected log-likelihood}}} - \underbrace{D_{\text{KL}}(q(\boldsymbol{z}) \| p(\boldsymbol{z}))}_{{\text{KL from prior}}}
$$

- The **first term** encourages $q$ to place mass where the data is explained well
- The **second term** is a regularizer pulling $q$ back toward the prior $p(\boldsymbol{z})$

When $p(\boldsymbol{z}) = \mathcal{N}(\boldsymbol{0}, \boldsymbol{I})$ and $q(\boldsymbol{z}) = \mathcal{N}(\boldsymbol{\mu}, \boldsymbol{\Sigma})$ (the VAE choice), the KL term has a closed form (derivable from the MVN results of Module 06):

$$
D_{\text{KL}}(\mathcal{N}(\boldsymbol{\mu}, \boldsymbol{\Sigma}) \| \mathcal{N}(\boldsymbol{0}, \boldsymbol{I})) = \frac{1}{2}\left(\text{tr}(\boldsymbol{\Sigma}) + \boldsymbol{\mu}^{\top}\boldsymbol{\mu} - n - \log|\boldsymbol{\Sigma}|\right)
$$

### Section 1 — Symbolic Verification: Divergence and Stokes' Theorems

Verify Example 3.1 (divergence theorem) and Example 4.1 (Stokes' theorem) symbolically using SymPy.

```python
# ============================================================
# Section 1 — Symbolic divergence and Stokes' theorem verification
# ============================================================
x, y, z, r, phi, theta, t = sp.symbols('x y z r phi theta t', real=True)
a = sp.Symbol('a', positive=True)

# --- Divergence theorem Example 3.1: F = (x³, y³, z³) ---
# ∇·F = 3x² + 3y² + 3z² = 3r² in spherical
# ∫∫∫ 3r² dV over unit ball
print('=== Divergence theorem Example 3.1 ===')
print('F = (x³, y³, z³),  ∇·F = 3x² + 3y² + 3z²')
integrand_vol = 3 * r**2 * r**2 * sin(phi)   # 3r² × Jacobian r²sinφ
vol_integral = sp.integrate(integrand_vol, (r,0,1))
vol_integral = sp.integrate(vol_integral, (phi,0,pi))
vol_integral = sp.integrate(vol_integral, (theta,0,2*pi))
print(f'∫∫∫ ∇·F dV = {vol_integral}  = 12π/5?', vol_integral == 12*pi/5)

# --- Surface area of sphere of radius a ---
print('\n=== Sphere surface area ===')
# |r_phi × r_theta| = a²sinφ
area_elem = a**2 * sin(phi)
sphere_area = sp.integrate(sp.integrate(area_elem, (phi, 0, pi)), (theta, 0, 2*pi))
print(f'Surface area of sphere radius a = {sphere_area}  = 4πa²?', sp.simplify(sphere_area - 4*pi*a**2) == 0)

# --- Stokes' theorem Example 4.1: F = (y, z, x) ---
print('\n=== Stokes\' theorem Example 4.1 ===')
print('F = (y, z, x),  curl F = (-1, -1, -1)')
# Line integral: ∮_C F·dr over unit circle in xy-plane
r_circ = sp.Matrix([cos(t), sin(t), sp.Integer(0)])
r_circ_prime = sp.diff(r_circ, t)
F_on_C = sp.Matrix([r_circ[1], r_circ[2], r_circ[0]])  # (y, z, x) on circle
line_int = sp.integrate(F_on_C.dot(r_circ_prime), (t, 0, 2*pi))
print(f'Line integral ∮ F·dr = {line_int}  = -π?', line_int == -pi)

# Flux integral: ∫∫_S curl(F)·dS where curl F = (-1,-1,-1)
# On upper hemisphere with outward normal n = (sinφcosθ, sinφsinθ, cosφ)
# curl(F)·n = -sinφcosθ - sinφsinθ - cosφ
nhat = sp.Matrix([sin(phi)*cos(theta), sin(phi)*sin(theta), cos(phi)])
curl_F = sp.Matrix([-1, -1, -1])
integrand_flux = curl_F.dot(nhat) * sin(phi)  # × area element a²sinφ with a=1
flux_int = sp.integrate(sp.integrate(integrand_flux, (phi, 0, pi/2)), (theta, 0, 2*pi))
print(f'Flux integral ∫∫ curl(F)·dS = {flux_int}  = -π?', flux_int == -pi)
print(f'Stokes verified: {line_int == flux_int}')
```

**What do you see?** Both theorems verify exactly. The divergence theorem computation gives $12\pi/5$ for $\boldsymbol{F} = (x^3, y^3, z^3)$. Stokes' theorem gives $-\pi$ both from the line integral and from the surface flux of the curl — the two sides of the identity agree.

### Section 2 — Divergence Theorem: Flux vs Interior Divergence

Numerically verify the divergence theorem for a test field by computing the surface flux integral and the volume divergence integral, and show they match.

```python
# ============================================================
# Section 2 — Numerical divergence theorem verification
# ============================================================
# F(x,y,z) = (x, y, z) over ball radius R
# Divergence theorem: ∮ F·dS = ∫∫∫ div(F) dV = 3·(4/3)πR³ = 4πR³

R = 2.0
exact_flux = 4 * np.pi * R**3
print(f'Exact flux (divergence theorem): 4πR³ = {exact_flux:.6f}')

# Monte Carlo estimate of surface flux ∮ F·n dS
# On sphere radius R: F = (x,y,z), n = (x,y,z)/R, F·n = R
# So flux = R × 4πR² = 4πR³  (trivial for this field)
# Demonstrate numerically for F = (x³/3, y³/3, z³/3)
# div(F) = x² + y² + z²

def F2(x, y, z):
    return np.array([x**3/3, y**3/3, z**3/3])

def div_F2(x, y, z):
    return x**2 + y**2 + z**2

# Monte Carlo volume integral: ∫∫∫_{ball} div(F) dV
rng = np.random.default_rng(42)
N = 500_000
# Sample uniformly in ball via rejection from cube [-R,R]³
pts = rng.uniform(-R, R, (N*4, 3))
in_ball = np.sum(pts**2, axis=1) <= R**2
pts_ball = pts[in_ball][:N]
cube_vol = (2*R)**3
ball_vol = (4/3)*np.pi*R**3
vol_int_mc = np.mean(div_F2(*pts_ball.T)) * ball_vol

# Exact volume integral: ∫∫∫ r² dV = (4π/5)R⁵
exact_vol = (4*np.pi/5) * R**5

# Monte Carlo surface integral via uniform sphere sampling
N_surf = 200_000
# Uniform on sphere: use Gaussian normalization
pts_s = rng.standard_normal((N_surf, 3))
pts_s = R * pts_s / np.linalg.norm(pts_s, axis=1, keepdims=True)
# Outward normal n = pts_s / R
n_hat = pts_s / R
F_at_surf = np.stack([F2(*pts_s[i]) for i in range(N_surf)])
# F·n at each point
Fn = np.sum(F_at_surf * n_hat, axis=1)
surf_area = 4 * np.pi * R**2
surf_int_mc = np.mean(Fn) * surf_area

print(f'\nFor F = (x³/3, y³/3, z³/3) over ball radius R={R}:')
print(f'  Exact ∫∫∫ div(F) dV = (4π/5)R⁵ = {exact_vol:.6f}')
print(f'  MC    ∫∫∫ div(F) dV           = {vol_int_mc:.6f}')
print(f'  MC    ∮ F·n dS                = {surf_int_mc:.6f}')
print(f'  Volume ≈ surface: {np.isclose(vol_int_mc, surf_int_mc, rtol=0.01)}')

# --- Visualization: flux density on sphere surface ---
fig = plt.figure(figsize=(13, 5))

ax = fig.add_subplot(121, projection='3d')
phi_vals = np.linspace(0, np.pi, 40)
theta_vals = np.linspace(0, 2*np.pi, 80)
PHI, THETA = np.meshgrid(phi_vals, theta_vals)
Xs = R * np.sin(PHI)*np.cos(THETA)
Ys = R * np.sin(PHI)*np.sin(THETA)
Zs = R * np.cos(PHI)
# F·n on sphere for F=(x,y,z): F·n = R everywhere
Fn_surf = np.sqrt(Xs**2 + Ys**2 + Zs**2)  # = R everywhere here; use x²+y²+z² for F2
Fn_F2 = (Xs**3/3 + Ys**3/3 + Zs**3/3) / R  # = (x³+y³+z³)/(3R) for F2, dot with n=r/R
# Actually F2·n = (x³/3, y³/3, z³/3) · (x,y,z)/R = (x⁴+y⁴+z⁴)/(3R)
Fn_F2_correct = (Xs**4 + Ys**4 + Zs**4) / (3*R)
surf = ax.plot_surface(Xs, Ys, Zs, facecolors=plt.cm.RdBu_r(
    (Fn_F2_correct - Fn_F2_correct.min()) / (Fn_F2_correct.max() - Fn_F2_correct.min())
), alpha=0.85, linewidth=0)
ax.set_title(r'$\mathbf{F}\cdot\hat{n}$ on sphere ($=F_2\cdot n$)' + '\nred=outward, blue=inward')
ax.set_xlabel('x'); ax.set_ylabel('y'); ax.set_zlabel('z')

# Right: divergence inside the ball
ax2 = fig.add_subplot(122)
r_vals = np.linspace(0, R, 200)
div_radial = r_vals**2   # div(F2) = x²+y²+z² = r² on radial cross-section
cumulative_div = np.array([
    4*np.pi * np.trapz(r_vals[:i+1]**2 * r_vals[:i+1]**2, r_vals[:i+1])
    for i in range(len(r_vals))
])
ax2.plot(r_vals, cumulative_div, color=TS_BLUE, lw=2.5,
         label=r'$\int_0^r (\nabla\cdot F)\cdot 4\pi r^2\,dr$')
ax2.axhline(exact_vol, color=TS_AMBER, lw=2, ls='--',
            label=f'Total = {exact_vol:.3f}')
ax2.set_xlabel('Radius $r$'); ax2.set_ylabel('Cumulative divergence integral')
ax2.set_title('Divergence accumulates with radius')
ax2.legend(fontsize=9)

plt.suptitle('Divergence theorem: surface flux = interior divergence', fontsize=11, y=1.01)
plt.tight_layout()
plt.show()
```

**What do you see?** The Monte Carlo estimates of the volume divergence integral and the surface flux integral agree to within 1% — the divergence theorem confirmed numerically. The surface plot shows the flux density is highest at the poles ($|z|$ large) and lowest at the equator — reflecting the $x^4 + y^4 + z^4$ structure of the field. The right panel shows cumulative divergence accumulating as radius grows, reaching the exact total at $r = R$.

### Section 3 — Statistical Bridge: KL Divergence and the ELBO

Compute the closed-form KL divergence between two Gaussian distributions, verify the ELBO decomposition, and visualize how the ELBO tightens as $q \to p$.

```python
# ============================================================
# Section 3 — KL divergence and ELBO
# ============================================================

def kl_gaussians(mu_q, sigma2_q, mu_p, sigma2_p):
    """KL(N(mu_q,sigma2_q) || N(mu_p,sigma2_p)) — scalar 1D case."""
    return 0.5 * (np.log(sigma2_p/sigma2_q)
                  + sigma2_q/sigma2_p
                  + (mu_q - mu_p)**2/sigma2_p
                  - 1)

def kl_mvn_to_standard(mu, Sigma):
    """KL(N(mu,Sigma) || N(0,I)) — the VAE regularization term."""
    n = len(mu)
    sign, logdet = np.linalg.slogdet(Sigma)
    return 0.5 * (np.trace(Sigma) + mu @ mu - n - logdet)

# --- 1D visualization: KL as q approaches p ---
mu_p_1d = 0.0;  sigma2_p_1d = 1.0  # target p
mu_q_range = np.linspace(-3, 3, 200)  # vary mean of q
sigma2_q_range = np.logspace(-2, 1, 200)  # vary variance of q

# KL vs mean shift (sigma2_q = sigma2_p = 1)
kl_mu_shift = kl_gaussians(mu_q_range, 1.0, mu_p_1d, sigma2_p_1d)
# KL vs variance ratio (mu_q = mu_p = 0)
kl_var_shift = kl_gaussians(0.0, sigma2_q_range, mu_p_1d, sigma2_p_1d)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

ax = axes[0]
ax.plot(mu_q_range, kl_mu_shift, color=TS_BLUE, lw=2.5)
ax.axvline(mu_p_1d, color=TS_AMBER, lw=1.5, ls='--', label=f'$\\mu_p = {mu_p_1d}$')
ax.axhline(0, color=TS_GRAY, lw=0.8)
ax.set_xlabel(r'$\mu_q$'); ax.set_ylabel(r'$D_{\mathrm{KL}}(q \| p)$')
ax.set_title('KL vs mean shift\n(both $\\sigma^2=1$)')
ax.legend(fontsize=9)

ax2 = axes[1]
ax2.semilogx(sigma2_q_range, kl_var_shift, color=TS_GREEN, lw=2.5)
ax2.axvline(sigma2_p_1d, color=TS_AMBER, lw=1.5, ls='--', label=f'$\\sigma^2_p = {sigma2_p_1d}$')
ax2.axhline(0, color=TS_GRAY, lw=0.8)
ax2.set_xlabel(r'$\sigma^2_q$ (log scale)')
ax2.set_ylabel(r'$D_{\mathrm{KL}}(q \| p)$')
ax2.set_title('KL vs variance ratio\n(both $\\mu=0$)')
ax2.legend(fontsize=9)

# Right: ELBO decomposition — toy example
# True posterior: p(z|x) = N(1.5, 0.4)
# Variational: q(z) = N(mu_q, 0.5)
# Evidence: log p(x) = constant (treat as known)
mu_true_post = 1.5;  sigma2_true_post = 0.4
sigma2_q_fixed = 0.5
mu_q_arr = np.linspace(-1, 4, 200)

# ELBO = log p(x) - KL(q||p_posterior)
# Treat log p(x) = 0 for visualization (it's a constant)
kl_to_post = kl_gaussians(mu_q_arr, sigma2_q_fixed, mu_true_post, sigma2_true_post)
elbo_arr = -kl_to_post  # ELBO = const - KL

ax3 = axes[2]
ax3.plot(mu_q_arr, elbo_arr, color=TS_BLUE, lw=2.5, label='ELBO (up to constant)')
ax3.plot(mu_q_arr, -kl_to_post, color=TS_RED, lw=1.5, ls='--',
         alpha=0.6, label=r'$-D_{\mathrm{KL}}(q||p)$')
ax3.axvline(mu_true_post, color=TS_AMBER, lw=2, ls='--',
            label=fr'Posterior mean $\mu_p={mu_true_post}$')
ax3.axhline(0, color=TS_GRAY, lw=0.8)
ax3.set_xlabel(r'$\mu_q$ (variational mean)')
ax3.set_ylabel('ELBO')
ax3.set_title('Maximizing ELBO\n≡ minimizing KL$(q\\|p_{\\mathrm{post}})$')
ax3.legend(fontsize=8)

plt.suptitle('KL divergence and the ELBO in variational inference', fontsize=11, y=1.01)
plt.tight_layout()
plt.show()

# MVN KL verification
print('=== KL(N(μ,Σ) || N(0,I)) — VAE regularization term ===')
mu_test = np.array([0.5, -0.3, 1.0])
Sigma_test = np.array([[0.8, 0.2, 0.1],
                       [0.2, 1.2, 0.0],
                       [0.1, 0.0, 0.6]])
kl_val = kl_mvn_to_standard(mu_test, Sigma_test)
print(f'KL = (1/2)[tr(Σ) + μᵀμ - n - log|Σ|]')
print(f'   = (1/2)[{np.trace(Sigma_test):.3f} + {mu_test@mu_test:.3f} - 3 - {np.linalg.slogdet(Sigma_test)[1]:.3f}]')
print(f'   = {kl_val:.6f}')
print(f'KL ≥ 0: {kl_val >= 0}')
```

**What do you see?**

- **Left**: KL divergence as a function of the mean shift. It is zero when $\mu_q = \mu_p$ and grows as a quadratic in $|\mu_q - \mu_p|$ — a bowl shape. Maximizing the ELBO drives $\mu_q$ toward $\mu_p$.
- **Middle**: KL as a function of the variance ratio on a log scale. It is zero when $\sigma^2_q = \sigma^2_p$ and grows asymmetrically — underestimating the variance ($\sigma^2_q < \sigma^2_p$) costs more than overestimating it by the same factor.
- **Right**: The ELBO as a function of the variational mean. Its maximum is at the true posterior mean — maximizing the ELBO is exactly minimizing the KL divergence to the posterior. The VAE training loop does this stochastically for each data point.
- The MVN KL computation confirms the formula is non-negative for the test covariance matrix.

**Connection to the divergence theorem**: The KL divergence $D_{\text{KL}}(q \| p) = \int q \log(q/p)\,d\boldsymbol{x}$ is a volume integral over probability space. The ELBO is a lower bound obtained by bounding this integral from below. The divergence theorem tells us that volume integrals can be expressed as boundary integrals — in this analogy, the ELBO is the "boundary term" that captures the accessible part of the volume integral.
