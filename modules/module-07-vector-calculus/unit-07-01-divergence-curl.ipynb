## Why This Unit

Modules 02–06 focused on scalar-valued functions $f : \mathbb{R}^n \to \mathbb{R}$. The gradient $\nabla f$ was a vector that described how $f$ changes, but $f$ itself was always a scalar. This module turns to **vector fields** — functions that assign a vector to every point in space. They arise wherever a quantity has both magnitude and direction at every location: velocity of a fluid, force on a particle, flow of probability mass.

Two new differential operators — **divergence** and **curl** — measure different aspects of a vector field's behavior. Divergence measures how much the field spreads outward from a point; curl measures how much it rotates. Together with the gradient, they form the three fundamental first-order differential operators of vector calculus, all expressible using the single del operator $\nabla$.

---

## 1. Vector Fields

**Definition.** A **vector field** on $\mathbb{R}^n$ is a function $\boldsymbol{F} : \mathbb{R}^n \to \mathbb{R}^n$ that assigns to each point $\boldsymbol{x} \in \mathbb{R}^n$ a vector $\boldsymbol{F}(\boldsymbol{x}) \in \mathbb{R}^n$.

In $\mathbb{R}^2$ we write $\boldsymbol{F}(x, y) = (P(x,y),\, Q(x,y))$ with component functions $P, Q : \mathbb{R}^2 \to \mathbb{R}$.

In $\mathbb{R}^3$ we write $\boldsymbol{F}(x, y, z) = (P(x,y,z),\, Q(x,y,z),\, R(x,y,z))$.

> **Comparison with Module 03.** In Module 03 Unit 03 we defined the Jacobian of a vector-valued function $\boldsymbol{F} : \mathbb{R}^n \to \mathbb{R}^m$. A vector field is the square case $m = n$: the input and output spaces have the same dimension. The Jacobian $\boldsymbol{J}_{\boldsymbol{F}} \in \mathbb{R}^{n \times n}$ of a vector field is a square matrix at every point.

**Common examples in machine learning:**
- The gradient field $\boldsymbol{F}(\boldsymbol{\theta}) = \nabla_{\boldsymbol{\theta}}\,J(\boldsymbol{\theta})$ — a vector field on parameter space
- The score function $\boldsymbol{s}(\boldsymbol{x}) = \nabla_{\boldsymbol{x}}\log p(\boldsymbol{x})$ — a vector field on data space
- A neural network's hidden-layer activations as a function of input — a vector field on $\mathbb{R}^n$

**Worked Example 1.1.** Consider $\boldsymbol{F}(x,y) = (-y,\, x)$. At every point $(x,y)$, the vector $(-y, x)$ is perpendicular to $(x,y)$ (their dot product is $-xy + xy = 0$) and has the same magnitude $\sqrt{x^2 + y^2}$. This is the counterclockwise rotation field — it is tangent to every circle centered at the origin, pointing in the direction of counterclockwise rotation.

## 2. The Del Operator

The **del operator** (also called **nabla**) in $\mathbb{R}^3$ is the symbolic vector of partial derivative operators:

$$
\nabla = \left(\frac{\partial}{\partial x},\, \frac{\partial}{\partial y},\, \frac{\partial}{\partial z}\right)
$$

Treating $\nabla$ as a vector and applying it to scalar and vector fields in different ways generates the three fundamental differential operators:

| Operation | Notation | Input | Output |
|---|---|---|---|
| Gradient | $\nabla f$ | Scalar field | Vector field |
| Divergence | $\nabla \cdot \boldsymbol{F}$ | Vector field | Scalar field |
| Curl | $\nabla \times \boldsymbol{F}$ | Vector field (in $\mathbb{R}^3$) | Vector field |

Composing these gives the **Laplacian**: $\nabla^2 f = \nabla \cdot (\nabla f)$ — a scalar field from a scalar field.

The del operator is not a vector in the usual sense — it does not have numerical components. It is an **operator** that must act on something to produce a value. But the notation $\nabla \cdot$ and $\nabla \times$ is constructed to mimic dot and cross products, making the formulas intuitive to remember.

## 3. Divergence

### 3.1 Definition

For a differentiable vector field $\boldsymbol{F} = (P, Q, R) : \mathbb{R}^3 \to \mathbb{R}^3$, the **divergence** is:

$$
\nabla \cdot \boldsymbol{F} = \frac{\partial P}{\partial x} + \frac{\partial Q}{\partial y} + \frac{\partial R}{\partial z}
$$

The divergence takes a vector field and returns a scalar field. At each point, it measures the net rate of outward flow — how much the field is "spreading out" from that point.

In $\mathbb{R}^2$ with $\boldsymbol{F} = (P, Q)$: $\;\nabla \cdot \boldsymbol{F} = \frac{\partial P}{\partial x} + \frac{\partial Q}{\partial y}$.

In $\mathbb{R}^n$: $\;\nabla \cdot \boldsymbol{F} = \sum_{i=1}^n \frac{\partial F_i}{\partial x_i}$ — the **trace of the Jacobian** $\boldsymbol{J}_{\boldsymbol{F}}$.

### 3.2 Geometric Interpretation

- $\nabla \cdot \boldsymbol{F}(\boldsymbol{x}) > 0$: the field is **spreading outward** at $\boldsymbol{x}$ — $\boldsymbol{x}$ is a **source**
- $\nabla \cdot \boldsymbol{F}(\boldsymbol{x}) < 0$: the field is **converging inward** at $\boldsymbol{x}$ — $\boldsymbol{x}$ is a **sink**
- $\nabla \cdot \boldsymbol{F}(\boldsymbol{x}) = 0$: the field is **incompressible** at $\boldsymbol{x}$ — as much flows in as flows out

An **incompressible** field satisfies $\nabla \cdot \boldsymbol{F} = 0$ everywhere. Velocity fields of incompressible fluids (e.g., water) satisfy this.

### 3.3 Worked Examples

**Example 3.1.** $\boldsymbol{F}(x,y,z) = (x,\, y,\, z)$ (radial field — points directly away from the origin).

$$
\nabla \cdot \boldsymbol{F} = \frac{\partial x}{\partial x} + \frac{\partial y}{\partial y} + \frac{\partial z}{\partial z} = 1 + 1 + 1 = 3
$$

Constant positive divergence everywhere — every point is a source. The field spreads outward uniformly.

**Example 3.2.** $\boldsymbol{F}(x,y) = (-y,\, x)$ (counterclockwise rotation from Example 1.1).

$$
\nabla \cdot \boldsymbol{F} = \frac{\partial(-y)}{\partial x} + \frac{\partial x}{\partial y} = 0 + 0 = 0
$$

Zero divergence — the rotation field is incompressible. Fluid rotating in circles neither spreads nor contracts.

**Example 3.3.** $\boldsymbol{F}(x,y,z) = (xy,\, e^z,\, x^2 + z)$.

$$
\nabla \cdot \boldsymbol{F} = \frac{\partial(xy)}{\partial x} + \frac{\partial(e^z)}{\partial y} + \frac{\partial(x^2 + z)}{\partial z} = y + 0 + 1 = y + 1
$$

The divergence depends on position — positive where $y > -1$, zero on the line $y = -1$, negative where $y < -1$.

## 4. Curl

### 4.1 Definition

For a differentiable vector field $\boldsymbol{F} = (P, Q, R) : \mathbb{R}^3 \to \mathbb{R}^3$, the **curl** is:

$$
\nabla \times \boldsymbol{F} = \det\begin{pmatrix} \boldsymbol{e}_1 & \boldsymbol{e}_2 & \boldsymbol{e}_3 \\ \frac{\partial}{\partial x} & \frac{\partial}{\partial y} & \frac{\partial}{\partial z} \\ P & Q & R \end{pmatrix} = \begin{pmatrix} \frac{\partial R}{\partial y} - \frac{\partial Q}{\partial z} \\[6pt] \frac{\partial P}{\partial z} - \frac{\partial R}{\partial x} \\[6pt] \frac{\partial Q}{\partial x} - \frac{\partial P}{\partial y} \end{pmatrix}
$$

The curl takes a vector field in $\mathbb{R}^3$ and returns a vector field in $\mathbb{R}^3$. At each point, it measures the local **rotation** of the field — how much a small paddle wheel placed at that point would spin if the field were a fluid velocity.

> **Notation reminder.** The three components follow the cyclic pattern $(y,z)$, $(z,x)$, $(x,y)$ — each is a difference of cross-partial derivatives. If you forget the order, the determinant expansion recovers it unambiguously.

### 4.2 Geometric Interpretation

- $\|\nabla \times \boldsymbol{F}(\boldsymbol{x})\|$: the rate of rotation of the field at $\boldsymbol{x}$
- Direction of $\nabla \times \boldsymbol{F}(\boldsymbol{x})$: the axis of rotation (right-hand rule)
- $\nabla \times \boldsymbol{F} = \boldsymbol{0}$ everywhere: the field is **irrotational** — no local spinning

An **irrotational** field has zero curl everywhere. Conservative fields (Unit 02) are always irrotational.

### 4.3 Worked Examples

**Example 4.1.** $\boldsymbol{F}(x,y,z) = (x,\, y,\, z)$ (radial field from Example 3.1).

$$
\nabla \times \boldsymbol{F} = \begin{pmatrix} \frac{\partial z}{\partial y} - \frac{\partial y}{\partial z} \\[4pt] \frac{\partial x}{\partial z} - \frac{\partial z}{\partial x} \\[4pt] \frac{\partial y}{\partial x} - \frac{\partial x}{\partial y} \end{pmatrix} = \begin{pmatrix} 0 - 0 \\ 0 - 0 \\ 0 - 0 \end{pmatrix} = \boldsymbol{0}
$$

The radial field has zero curl — it spreads outward but does not rotate.

**Example 4.2.** $\boldsymbol{F}(x,y,z) = (-y,\, x,\, 0)$ (counterclockwise rotation in the $xy$-plane, extended to $\mathbb{R}^3$).

$$
\nabla \times \boldsymbol{F} = \begin{pmatrix} \frac{\partial(0)}{\partial y} - \frac{\partial x}{\partial z} \\[4pt] \frac{\partial(-y)}{\partial z} - \frac{\partial(0)}{\partial x} \\[4pt] \frac{\partial x}{\partial x} - \frac{\partial(-y)}{\partial y} \end{pmatrix} = \begin{pmatrix} 0 - 0 \\ 0 - 0 \\ 1 - (-1) \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \\ 2 \end{pmatrix}
$$

The curl points in the $+z$ direction (out of the page) with magnitude 2 — counterclockwise rotation about the $z$-axis, as expected. The curl is constant, confirming uniform rotation everywhere.

**Example 4.3.** $\boldsymbol{F}(x,y,z) = (yz,\, xz,\, xy)$.

$$
\nabla \times \boldsymbol{F} = \begin{pmatrix} \frac{\partial(xy)}{\partial y} - \frac{\partial(xz)}{\partial z} \\[4pt] \frac{\partial(yz)}{\partial z} - \frac{\partial(xy)}{\partial x} \\[4pt] \frac{\partial(xz)}{\partial x} - \frac{\partial(yz)}{\partial y} \end{pmatrix} = \begin{pmatrix} x - x \\ y - y \\ z - z \end{pmatrix} = \boldsymbol{0}
$$

Zero curl — this field is irrotational. We will see in Unit 02 that it is also conservative with potential function $f(x,y,z) = xyz$.

## 5. The Laplacian

### 5.1 Definition

The **Laplacian** of a scalar field $f : \mathbb{R}^n \to \mathbb{R}$ is the divergence of its gradient:

$$
\nabla^2 f = \nabla \cdot (\nabla f) = \sum_{i=1}^n \frac{\partial^2 f}{\partial x_i^2}
$$

In $\mathbb{R}^3$: $\;\nabla^2 f = \frac{\partial^2 f}{\partial x^2} + \frac{\partial^2 f}{\partial y^2} + \frac{\partial^2 f}{\partial z^2}$.

The Laplacian takes a scalar field and returns a scalar field. It measures the **average curvature** of $f$ — how much $f$ at a point differs from the average of $f$ in a small ball around that point.

### 5.2 Harmonic Functions

A function satisfying $\nabla^2 f = 0$ everywhere is called **harmonic**. Harmonic functions are the solutions to **Laplace's equation** — among the most important equations in mathematical physics (governing electrostatics, steady-state heat flow, and potential theory).

**Example 5.1.** $f(x,y) = x^2 - y^2$.

$$
\frac{\partial^2 f}{\partial x^2} = 2 \qquad \frac{\partial^2 f}{\partial y^2} = -2 \qquad \nabla^2 f = 2 + (-2) = 0 \checkmark
$$

**Example 5.2.** $f(x,y) = e^x \cos y$.

$$
\frac{\partial^2 f}{\partial x^2} = e^x \cos y \qquad \frac{\partial^2 f}{\partial y^2} = -e^x \cos y \qquad \nabla^2 f = 0 \checkmark
$$

### 5.3 Key Identities

These vector calculus identities hold for any sufficiently differentiable scalar field $f$ and vector field $\boldsymbol{F}$:

$$
\nabla \times (\nabla f) = \boldsymbol{0} \qquad \text{(curl of a gradient is always zero)}
$$

$$
\nabla \cdot (\nabla \times \boldsymbol{F}) = 0 \qquad \text{(divergence of a curl is always zero)}
$$

Both follow from the equality of mixed partial derivatives (Schwarz's theorem, Module 02). These identities are the reason conservative fields are irrotational — we will use the first one in Unit 02.

## 6. Statistical Bridge — Score Function as a Vector Field & Stein's Lemma `[MLE]`

### 6.1 The Score Function as a Vector Field

For a probability density $p(\boldsymbol{x}; \boldsymbol{\theta})$ on $\mathbb{R}^n$, the **score function** with respect to the data $\boldsymbol{x}$ is:

$$
\boldsymbol{s}(\boldsymbol{x}) = \nabla_{\boldsymbol{x}}\log p(\boldsymbol{x}) = \frac{\nabla_{\boldsymbol{x}}\, p(\boldsymbol{x})}{p(\boldsymbol{x})}
$$

This is a vector field on $\mathbb{R}^n$ — at each data point $\boldsymbol{x}$, it assigns the gradient of the log-density. It points in the direction that most increases the log-probability of $\boldsymbol{x}$ — toward regions of higher density.

> **Distinction from Module 03.** In Module 03 Unit 01, the score function $\nabla_{\boldsymbol{\theta}}\,\ell(\boldsymbol{\theta})$ was the gradient with respect to the **parameters** $\boldsymbol{\theta}$ — a vector in parameter space. Here, $\boldsymbol{s}(\boldsymbol{x}) = \nabla_{\boldsymbol{x}}\log p(\boldsymbol{x})$ is the gradient with respect to the **data** $\boldsymbol{x}$ — a vector field on data space. Both are score functions; the subscript clarifies which variable we differentiate with respect to.

### 6.2 Stein's Lemma

**Theorem (Stein's Lemma).** Let $p(\boldsymbol{x})$ be a differentiable probability density on $\mathbb{R}^n$ that vanishes at infinity (i.e., $p(\boldsymbol{x}) \to 0$ as $\|\boldsymbol{x}\| \to \infty$). Then for any differentiable function $g : \mathbb{R}^n \to \mathbb{R}$ with $\mathbb{E}_{p}[|\nabla g|] < \infty$:

$$
\mathbb{E}_{p}\!\left[g(\boldsymbol{X})\,\nabla_{\boldsymbol{x}}\log p(\boldsymbol{X})\right] = -\mathbb{E}_{p}\!\left[\nabla_{\boldsymbol{x}}\,g(\boldsymbol{X})\right]
$$

**Special case $g = 1$:** setting $g(\boldsymbol{x}) = 1$ (so $\nabla g = \boldsymbol{0}$):

$$
\mathbb{E}_{p}\!\left[\nabla_{\boldsymbol{x}}\log p(\boldsymbol{X})\right] = \boldsymbol{0}
$$

**The expected score (w.r.t. data) is zero.** The score vector field averages to zero under the density it characterizes.

### 6.3 Proof of the Special Case

Using $\nabla_{\boldsymbol{x}}\log p = \nabla_{\boldsymbol{x}}\, p / p$:

$$
\mathbb{E}_{p}\!\left[\nabla_{\boldsymbol{x}}\log p(\boldsymbol{X})\right]
= \int_{\mathbb{R}^n} \nabla_{\boldsymbol{x}}\log p(\boldsymbol{x})\cdot p(\boldsymbol{x})\,d\boldsymbol{x}
= \int_{\mathbb{R}^n} \nabla_{\boldsymbol{x}}\, p(\boldsymbol{x})\,d\boldsymbol{x}
= \nabla_{\boldsymbol{x}}\int_{\mathbb{R}^n} p(\boldsymbol{x})\,d\boldsymbol{x}
= \nabla_{\boldsymbol{x}}\, 1 = \boldsymbol{0}
$$

The third equality swaps differentiation and integration (justified by dominated convergence); the fourth uses the fact that $p$ normalizes to 1.

### 6.4 Why Stein's Lemma Matters

Stein's lemma underlies several modern techniques:

- **Score-based generative models** (diffusion models): the score $\nabla_{\boldsymbol{x}}\log p$ is the object learned by the neural network. Sampling is done by following the score field
- **Stein variational gradient descent**: approximates a target distribution by moving particles along the score field
- **Score matching**: an alternative to MLE for fitting energy-based models, using only the score field rather than the normalizing constant
- **Fisher information** (Module 06 Unit 03): the variance of the score $\mathbb{E}[(\nabla_{\boldsymbol{\theta}}\log p)^2]$ — which equals the Fisher information by Stein's lemma applied to parameters

### Section 1 — Symbolic Divergence and Curl Verification

Use SymPy to verify the worked examples from Sections 3 and 4, confirm the two key identities ($\nabla \times \nabla f = \boldsymbol{0}$ and $\nabla \cdot (\nabla \times \boldsymbol{F}) = 0$), and compute the Laplacian of the harmonic functions from Section 5.

```python
# ============================================================
# Section 1 — Symbolic divergence, curl, and Laplacian
# ============================================================
x, y, z = symbols('x y z', real=True)

def divergence_3d(P, Q, R):
    return diff(P, x) + diff(Q, y) + diff(R, z)

def curl_3d(P, Q, R):
    """Returns (curl_x, curl_y, curl_z)."""
    curl_x = diff(R, y) - diff(Q, z)
    curl_y = diff(P, z) - diff(R, x)
    curl_z = diff(Q, x) - diff(P, y)
    return Matrix([curl_x, curl_y, curl_z])

def laplacian_scalar(f):
    return diff(f, x, 2) + diff(f, y, 2) + diff(f, z, 2)

# --- Section 3 examples ---
print('=== Divergence examples ===')
# Example 3.1: F = (x, y, z)
div_31 = divergence_3d(x, y, z)
print(f'Ex 3.1  F=(x,y,z):      div F = {div_31}  (expect 3)')
# Example 3.2: F = (-y, x, 0)
div_32 = divergence_3d(-y, x, sp.Integer(0))
print(f'Ex 3.2  F=(-y,x,0):     div F = {div_32}  (expect 0)')
# Example 3.3: F = (xy, e^z, x^2+z)
div_33 = divergence_3d(x*y, exp(z), x**2 + z)
print(f'Ex 3.3  F=(xy,e^z,x²+z): div F = {div_33}  (expect y+1)')

print('\n=== Curl examples ===')
# Example 4.1: F = (x, y, z)
curl_41 = curl_3d(x, y, z)
print(f'Ex 4.1  F=(x,y,z):      curl F = {curl_41.T}  (expect [0,0,0])')
# Example 4.2: F = (-y, x, 0)
curl_42 = curl_3d(-y, x, sp.Integer(0))
print(f'Ex 4.2  F=(-y,x,0):     curl F = {curl_42.T}  (expect [0,0,2])')
# Example 4.3: F = (yz, xz, xy)
curl_43 = curl_3d(y*z, x*z, x*y)
print(f'Ex 4.3  F=(yz,xz,xy):   curl F = {curl_43.T}  (expect [0,0,0])')

print('\n=== Key identities ===')
# curl(grad f) = 0 for f = x²y + z³
f_test = x**2*y + z**3
gf = Matrix([diff(f_test, x), diff(f_test, y), diff(f_test, z)])
curl_grad_f = curl_3d(gf[0], gf[1], gf[2])
print(f'curl(∇f) = {curl_grad_f.T}  (expect [0,0,0]): {curl_grad_f == Matrix([0,0,0])}')

# div(curl F) = 0 for F = (yz, xz, xy)
div_curl = divergence_3d(curl_43[0], curl_43[1], curl_43[2])
print(f'div(curl F) = {div_curl}  (expect 0): {div_curl == 0}')

print('\n=== Laplacian (harmonic functions) ===')
print(f'∇²(x²-y²) = {laplacian_scalar(x**2 - y**2)}  (expect 0)')
print(f'∇²(e^x cos y) = {simplify(laplacian_scalar(exp(x)*cos(y)))}  (expect 0)')
```

**What do you see?** Every worked example is verified symbolically. The two key identities — $\nabla \times (\nabla f) = \boldsymbol{0}$ and $\nabla \cdot (\nabla \times \boldsymbol{F}) = 0$ — print `True`. Both harmonic functions have zero Laplacian. These results are purely algebraic consequences of Schwarz's theorem (equality of mixed partials), confirmed here by symbolic computation.

### Section 2 — Vector Field Visualization: Divergence and Curl

Plot four vector fields side by side — source, sink, incompressible rotation, and a mixed field — with divergence and curl overlaid as color maps.

```python
# ============================================================
# Section 2 — Vector field visualization
# ============================================================
xg = np.linspace(-2.5, 2.5, 18)
yg = np.linspace(-2.5, 2.5, 18)
Xq, Yq = np.meshgrid(xg, yg)

# Dense grid for divergence/curl background
xd = np.linspace(-2.5, 2.5, 200)
yd = np.linspace(-2.5, 2.5, 200)
Xd, Yd = np.meshgrid(xd, yd)

fields = [
    {
        'name': 'Source: $\\mathbf{F}=(x,y)$',
        'P': lambda X, Y: X,  'Q': lambda X, Y: Y,
        'div': lambda X, Y: np.ones_like(X)*2,
        'curl_z': lambda X, Y: np.zeros_like(X),
        'cmap': 'RdBu_r', 'show': 'div',
    },
    {
        'name': 'Sink: $\\mathbf{F}=(-x,-y)$',
        'P': lambda X, Y: -X,  'Q': lambda X, Y: -Y,
        'div': lambda X, Y: np.ones_like(X)*(-2),
        'curl_z': lambda X, Y: np.zeros_like(X),
        'cmap': 'RdBu_r', 'show': 'div',
    },
    {
        'name': 'Rotation: $\\mathbf{F}=(-y,x)$',
        'P': lambda X, Y: -Y,  'Q': lambda X, Y: X,
        'div': lambda X, Y: np.zeros_like(X),
        'curl_z': lambda X, Y: np.ones_like(X)*2,
        'cmap': 'PRGn', 'show': 'curl',
    },
    {
        'name': 'Mixed: $\\mathbf{F}=(y^2, x)$',
        'P': lambda X, Y: Y**2,  'Q': lambda X, Y: X,
        'div': lambda X, Y: np.zeros_like(X),
        'curl_z': lambda X, Y: 1 - 2*Y,
        'cmap': 'PRGn', 'show': 'curl',
    },
]

fig, axes = plt.subplots(2, 2, figsize=(13, 11))

for ax, field in zip(axes.flat, fields):
    P_q = field['P'](Xq, Yq)
    Q_q = field['Q'](Xq, Yq)
    norm = np.sqrt(P_q**2 + Q_q**2) + 1e-8

    if field['show'] == 'div':
        bg = field['div'](Xd, Yd)
        clabel = r'div $\mathbf{F}$'
    else:
        bg = field['curl_z'](Xd, Yd)
        clabel = r'curl$_z$ $\mathbf{F}$'

    vext = max(abs(bg.min()), abs(bg.max())) + 0.1
    cf = ax.contourf(Xd, Yd, bg, levels=20,
                     cmap=field['cmap'], alpha=0.55,
                     vmin=-vext, vmax=vext)
    plt.colorbar(cf, ax=ax, label=clabel, shrink=0.85)
    ax.quiver(Xq, Yq, P_q/norm, Q_q/norm,
              color='black', alpha=0.65, scale=22, width=0.004)
    ax.set_title(field['name'], fontsize=11)
    ax.set_xlabel('x'); ax.set_ylabel('y')
    ax.set_aspect('equal')
    ax.set_xlim(-2.6, 2.6); ax.set_ylim(-2.6, 2.6)

plt.suptitle('Vector fields colored by divergence (top) and curl$_z$ (bottom)',
             fontsize=12, y=1.01)
plt.tight_layout()
plt.show()
```

**What do you see?**

- **Source** (top left): arrows point radially outward, divergence is uniformly $+2$ (red everywhere). Zero curl — pure spreading, no rotation.
- **Sink** (top right): arrows point inward, divergence is $-2$ (blue everywhere). The field converges at the origin.
- **Rotation** (bottom left): arrows circle counterclockwise, divergence is zero, curl$_z = +2$ (uniform green). No spreading, pure rotation.
- **Mixed** (bottom right): $\boldsymbol{F} = (y^2, x)$ has zero divergence but non-constant curl$_z = 1 - 2y$. The curl changes sign at $y = 1/2$ — the field rotates counterclockwise below this line and clockwise above it.

Divergence and curl measure orthogonal aspects of a field's behavior. A field can have large divergence and zero curl, zero divergence and large curl, or both nonzero simultaneously.

### Section 3 — Statistical Bridge: Score Field and Stein's Lemma Verification

Visualize the score vector field $\nabla_{\boldsymbol{x}}\log p(\boldsymbol{x})$ for a bivariate normal density, and verify Stein's lemma numerically: $\mathbb{E}[\nabla_{\boldsymbol{x}}\log p(\boldsymbol{X})] \approx \boldsymbol{0}$.

```python
# ============================================================
# Section 3 — Score vector field and Stein's lemma
# ============================================================
import numpy as np

# Bivariate normal parameters
mu_bvn  = np.array([1.0, -0.5])
Sigma_bvn = np.array([[2.0, 0.9], [0.9, 1.0]])
Sigma_inv = np.linalg.inv(Sigma_bvn)

# Score function: ∇_x log p(x) = -Σ⁻¹(x - μ)  for MVN
# Derivation: log p(x) = const - (1/2)(x-μ)ᵀΣ⁻¹(x-μ)
# ∇_x log p(x) = -Σ⁻¹(x - μ)  (gradient of quadratic form, Module 03)
def score_mvn(x, mu, Sigma_inv):
    """Score ∇_x log p(x) for N(mu, Sigma)."""
    return -(Sigma_inv @ (x - mu).T).T

# Grid for quiver
xq = np.linspace(-2.5, 4.5, 16)
yq = np.linspace(-3.5, 2.5, 16)
Xq2, Yq2 = np.meshgrid(xq, yq)
pts_q = np.stack([Xq2.ravel(), Yq2.ravel()], axis=1)
scores_q = score_mvn(pts_q, mu_bvn, Sigma_inv).reshape(*Xq2.shape, 2)
S1, S2 = scores_q[..., 0], scores_q[..., 1]
norm_s = np.sqrt(S1**2 + S2**2) + 1e-8

# Dense grid for density contours
xd = np.linspace(-3, 5, 200)
yd = np.linspace(-4, 3, 200)
Xd2, Yd2 = np.meshgrid(xd, yd)
pts_d = np.stack([Xd2.ravel(), Yd2.ravel()], axis=1)
from scipy import stats as st
rv = st.multivariate_normal(mu_bvn, Sigma_bvn)
Zd = rv.pdf(pts_d).reshape(200, 200)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Left: density contours + score field
ax = axes[0]
ax.contourf(Xd2, Yd2, Zd, levels=10, cmap='Blues', alpha=0.55)
ax.contour(Xd2, Yd2, Zd, levels=10, colors=[TS_GRAY], alpha=0.4, linewidths=0.7)
ax.quiver(Xq2, Yq2, S1/norm_s, S2/norm_s,
          color=TS_AMBER, alpha=0.8, scale=22, width=0.005)
ax.plot(*mu_bvn, '*', color=TS_RED, markersize=14, zorder=6,
        label=r'Mean $\boldsymbol{\mu}$ (score $=\mathbf{0}$)')
ax.set_xlabel('$x_1$'); ax.set_ylabel('$x_2$')
ax.set_title(r'Score field $\nabla_{\mathbf{x}}\log p(\mathbf{x})$' + '\npoints toward higher density')
ax.legend(fontsize=9)
ax.set_aspect('equal')

# Right: Stein's lemma verification
ax2 = axes[1]
rng = np.random.default_rng(42)
N_vals = [10, 50, 200, 1000, 5000, 20000, 100000]
mean_score_norms = []
for N in N_vals:
    samp = rng.multivariate_normal(mu_bvn, Sigma_bvn, N)
    sc = score_mvn(samp, mu_bvn, Sigma_inv)
    mean_score_norms.append(np.linalg.norm(sc.mean(axis=0)))

ax2.loglog(N_vals, mean_score_norms, 'o-', color=TS_BLUE, lw=2, markersize=6,
           label=r'$\|\mathbb{E}_N[\nabla_x \log p]\|$')
# Reference: 1/√N decay
ref = mean_score_norms[0] * np.sqrt(N_vals[0]) / np.sqrt(np.array(N_vals))
ax2.loglog(N_vals, ref, '--', color=TS_GRAY, lw=1.5, label=r'$O(N^{-1/2})$ reference')
ax2.set_xlabel('Sample size $N$')
ax2.set_ylabel(r'$\|\hat{\mathbb{E}}[\nabla_x \log p]\|$')
ax2.set_title("Stein's lemma: empirical mean score $\\to \\mathbf{0}$")
ax2.legend(fontsize=9)
ax2.text(0.05, 0.10,
         r'$\mathbb{E}_p[\nabla_x \log p(\mathbf{X})] = \mathbf{0}$  (exact)',
         transform=ax2.transAxes, fontsize=10, color=TS_GREEN,
         bbox=dict(facecolor=TS_LIGHT, alpha=0.8))

plt.suptitle('Score vector field and Stein\'s lemma', fontsize=11, y=1.01)
plt.tight_layout()
plt.show()

print('Score at the mean (should be 0):', score_mvn(mu_bvn, mu_bvn, Sigma_inv))
print(f'Empirical mean score (N=100000): {score_mvn(rng.multivariate_normal(mu_bvn, Sigma_bvn, 100000), mu_bvn, Sigma_inv).mean(axis=0)}')
```

**What do you see?**

- **Left**: The score field points toward the mean (red star) from every direction — arrows in the low-density tails point strongly inward, arrows near the mean are near zero. At the mean itself, the score is exactly $\boldsymbol{0}$: $\nabla_{\boldsymbol{x}}\log p(\boldsymbol{\mu}) = -\boldsymbol{\Sigma}^{-1}(\boldsymbol{\mu} - \boldsymbol{\mu}) = \boldsymbol{0}$. The score field has zero divergence everywhere — at the mean there are no sources or sinks — and its average over the density is exactly $\boldsymbol{0}$.
- **Right**: The empirical mean $\hat{\mathbb{E}}_N[\nabla_{\boldsymbol{x}}\log p]$ decays toward zero at the $O(N^{-1/2})$ rate of the central limit theorem. Stein's lemma says the true expectation is $\boldsymbol{0}$; the empirical average converges to it.

## Summary

| Concept | Statement |
|---|---|
| Vector field | $\boldsymbol{F} : \mathbb{R}^n \to \mathbb{R}^n$ — vector at every point |
| Del operator | $\nabla = (\partial_x, \partial_y, \partial_z)$ — operator, not a vector |
| Divergence | $\nabla \cdot \boldsymbol{F} = \partial_x P + \partial_y Q + \partial_z R$ — scalar; measures net outward flow |
| Curl (in $\mathbb{R}^3$) | $\nabla \times \boldsymbol{F} = (\partial_y R - \partial_z Q,\; \partial_z P - \partial_x R,\; \partial_x Q - \partial_y P)$ — measures rotation |
| Laplacian | $\nabla^2 f = \partial^2_x f + \partial^2_y f + \partial^2_z f$ — average curvature |
| Harmonic | $\nabla^2 f = 0$ everywhere |
| Key identities | $\nabla \times (\nabla f) = \boldsymbol{0}$; $\;\nabla \cdot (\nabla \times \boldsymbol{F}) = 0$ |
| Score field | $\boldsymbol{s}(\boldsymbol{x}) = \nabla_{\boldsymbol{x}}\log p(\boldsymbol{x})$ — vector field on data space |
| Stein's lemma | $\mathbb{E}_p[\nabla_{\boldsymbol{x}}\log p(\boldsymbol{X})] = \boldsymbol{0}$ |
| MVN score | $\nabla_{\boldsymbol{x}}\log p(\boldsymbol{x}) = -\boldsymbol{\Sigma}^{-1}(\boldsymbol{x} - \boldsymbol{\mu})$ |
