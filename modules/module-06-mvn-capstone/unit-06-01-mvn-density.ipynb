## 1. The Density — Definition

A random vector $\boldsymbol{X} \in \mathbb{R}^n$ follows a **multivariate normal distribution** with mean vector $\boldsymbol{\mu} \in \mathbb{R}^n$ and covariance matrix $\boldsymbol{\Sigma} \in \mathbb{R}^{n \times n}$, written $\boldsymbol{X} \sim \mathcal{N}(\boldsymbol{\mu}, \boldsymbol{\Sigma})$, if its probability density function is:

$$
\boxed{f(\boldsymbol{x}) = \frac{1}{(2\pi)^{n/2}|\boldsymbol{\Sigma}|^{1/2}}\exp\!\left(-\frac{1}{2}(\boldsymbol{x} - \boldsymbol{\mu})^{\top}\boldsymbol{\Sigma}^{-1}(\boldsymbol{x} - \boldsymbol{\mu})\right)}
$$

**Reading every factor:**

| Factor | Origin | Role |
|---|---|---|
| $(2\pi)^{n/2}$ | Module 05 Unit 02 | Gaussian normalization — one $\sqrt{2\pi}$ per dimension |
| $|\boldsymbol{\Sigma}|^{1/2}$ | Jacobian of the whitening transform | Accounts for the volume scaling by $\boldsymbol{\Sigma}$ |
| $(\boldsymbol{x} - \boldsymbol{\mu})^{\top}\boldsymbol{\Sigma}^{-1}(\boldsymbol{x} - \boldsymbol{\mu})$ | Module 00, 04 | Squared Mahalanobis distance — dimensionless measure of how far $\boldsymbol{x}$ is from $\boldsymbol{\mu}$ |
| $-\frac{1}{2}$ | Single-variable analogy | Matches $e^{-x^2/(2\sigma^2)}$ when $n=1$, $\Sigma = \sigma^2$ |

**Required conditions on $\boldsymbol{\Sigma}$:** The covariance matrix must be **symmetric positive definite** (SPD):
- **Symmetric**: $\boldsymbol{\Sigma} = \boldsymbol{\Sigma}^{\top}$ — because $\text{Cov}(X_i, X_j) = \text{Cov}(X_j, X_i)$
- **Positive definite**: $\boldsymbol{v}^{\top}\boldsymbol{\Sigma}\boldsymbol{v} > 0$ for all $\boldsymbol{v} \neq \boldsymbol{0}$ — ensures $\boldsymbol{\Sigma}^{-1}$ exists, $|\boldsymbol{\Sigma}| > 0$, and the density is integrable
- **Why PD is natural**: $\boldsymbol{\Sigma} = \mathbb{E}[(\boldsymbol{X}-\boldsymbol{\mu})(\boldsymbol{X}-\boldsymbol{\mu})^{\top}]$ is always PSD by construction, and PD when no component is a deterministic linear combination of others

## 2. Why the Normalization Constant is $(2\pi)^{n/2}|\boldsymbol{\Sigma}|^{1/2}$

We verify that $f$ integrates to 1 by tracing through the whitening transformation.

**Step 1. Whitening.** Since $\boldsymbol{\Sigma}$ is SPD, it has a Cholesky decomposition $\boldsymbol{\Sigma} = \boldsymbol{L}\boldsymbol{L}^{\top}$. Define the **whitening transform**:

$$
\boldsymbol{z} = \boldsymbol{L}^{-1}(\boldsymbol{x} - \boldsymbol{\mu})
$$

**Step 2. The Jacobian.** The transformation is $\boldsymbol{x} = \boldsymbol{L}\boldsymbol{z} + \boldsymbol{\mu}$, so:

$$
\frac{\partial \boldsymbol{x}}{\partial \boldsymbol{z}} = \boldsymbol{L} \qquad \Rightarrow \qquad |\det(\boldsymbol{J})| = |\det(\boldsymbol{L})| = |\boldsymbol{\Sigma}|^{1/2}
$$

because $|\det(\boldsymbol{L})|^2 = |\det(\boldsymbol{L}\boldsymbol{L}^{\top})| = |\boldsymbol{\Sigma}|$.

**Step 3. The exponent in $\boldsymbol{z}$.** With $\boldsymbol{z} = \boldsymbol{L}^{-1}(\boldsymbol{x}-\boldsymbol{\mu})$:

$$
(\boldsymbol{x}-\boldsymbol{\mu})^{\top}\boldsymbol{\Sigma}^{-1}(\boldsymbol{x}-\boldsymbol{\mu}) = \boldsymbol{z}^{\top}(\boldsymbol{L}^{\top})^{-1}(\boldsymbol{L}\boldsymbol{L}^{\top})^{-1}\boldsymbol{L}^{-1}\boldsymbol{L}\boldsymbol{z} = \boldsymbol{z}^{\top}\boldsymbol{z} = \|\boldsymbol{z}\|^2
$$

**Step 4. The integral.** By the change-of-variables formula (Module 05 Unit 02):

$$
\int_{\mathbb{R}^n} f(\boldsymbol{x})\,d\boldsymbol{x}
= \frac{|\boldsymbol{\Sigma}|^{1/2}}{(2\pi)^{n/2}|\boldsymbol{\Sigma}|^{1/2}} \int_{\mathbb{R}^n} e^{-\|\boldsymbol{z}\|^2/2}\,d\boldsymbol{z}
= \frac{1}{(2\pi)^{n/2}} \cdot (2\pi)^{n/2} = 1
$$

where the last step uses $\int_{\mathbb{R}^n} e^{-\|\boldsymbol{z}\|^2/2}\,d\boldsymbol{z} = (2\pi)^{n/2}$ from Module 05 Unit 03.

**Summary of the normalization:** the $|\boldsymbol{\Sigma}|^{1/2}$ in the denominator exactly cancels the Jacobian $|\det(\boldsymbol{L})| = |\boldsymbol{\Sigma}|^{1/2}$ introduced by the whitening transform.

## 3. The Mahalanobis Distance

The quantity in the exponent has a name and a geometric interpretation:

$$
d_{\mathcal{M}}(\boldsymbol{x}, \boldsymbol{\mu}) \coloneqq \sqrt{(\boldsymbol{x} - \boldsymbol{\mu})^{\top}\boldsymbol{\Sigma}^{-1}(\boldsymbol{x} - \boldsymbol{\mu})}
$$

This is the **Mahalanobis distance** from $\boldsymbol{x}$ to $\boldsymbol{\mu}$. It generalises Euclidean distance by accounting for the scale and correlation structure of $\boldsymbol{\Sigma}$.

**Properties:**
- When $\boldsymbol{\Sigma} = \boldsymbol{I}$: $d_{\mathcal{M}} = \|\boldsymbol{x} - \boldsymbol{\mu}\|_2$ — reduces to Euclidean distance
- When $\boldsymbol{\Sigma} = \text{diag}(\sigma_1^2, \ldots, \sigma_n^2)$: $d_{\mathcal{M}}^2 = \sum_i (x_i - \mu_i)^2/\sigma_i^2$ — normalized squared deviations
- **Scale invariance**: $d_{\mathcal{M}}$ is invariant under affine transformations that preserve $\boldsymbol{\Sigma}$ — it measures distance in units of standard deviations, not raw coordinates

**Level sets.** The level sets of the MVN density are the sets where $d_{\mathcal{M}} = c$ for constant $c$:

$$
\{\boldsymbol{x} : (\boldsymbol{x} - \boldsymbol{\mu})^{\top}\boldsymbol{\Sigma}^{-1}(\boldsymbol{x} - \boldsymbol{\mu}) = c^2\}
$$

These are **ellipsoids** centered at $\boldsymbol{\mu}$. The shape and orientation of the ellipsoids are determined by the eigendecomposition of $\boldsymbol{\Sigma}$.

---

## 4. Eigendecomposition of $\boldsymbol{\Sigma}$ and the Shape of the Density

Since $\boldsymbol{\Sigma}$ is symmetric positive definite, it has an eigendecomposition:

$$
\boldsymbol{\Sigma} = \boldsymbol{Q}\boldsymbol{\Lambda}\boldsymbol{Q}^{\top}
$$

where $\boldsymbol{Q} = [\boldsymbol{q}_1 \mid \cdots \mid \boldsymbol{q}_n]$ is orthonormal (eigenvectors) and $\boldsymbol{\Lambda} = \text{diag}(\lambda_1, \ldots, \lambda_n)$ with all $\lambda_i > 0$ (eigenvalues).

Then $\boldsymbol{\Sigma}^{-1} = \boldsymbol{Q}\boldsymbol{\Lambda}^{-1}\boldsymbol{Q}^{\top}$, and the Mahalanobis distance becomes:

$$
d_{\mathcal{M}}^2 = (\boldsymbol{x}-\boldsymbol{\mu})^{\top}\boldsymbol{Q}\boldsymbol{\Lambda}^{-1}\boldsymbol{Q}^{\top}(\boldsymbol{x}-\boldsymbol{\mu}) = \sum_{i=1}^n \frac{[(\boldsymbol{x}-\boldsymbol{\mu}) \cdot \boldsymbol{q}_i]^2}{\lambda_i}
$$

Setting $d_{\mathcal{M}}^2 = c^2$ gives the equation of an ellipsoid in the **eigenbasis** of $\boldsymbol{\Sigma}$:

$$
\sum_{i=1}^n \frac{u_i^2}{\lambda_i} = c^2 \qquad \text{where } u_i = (\boldsymbol{x}-\boldsymbol{\mu}) \cdot \boldsymbol{q}_i
$$

**Reading the geometry:**

| Feature | Determined by |
|---|---|
| Axes of the ellipsoid | Eigenvectors $\boldsymbol{q}_i$ of $\boldsymbol{\Sigma}$ — the principal directions |
| Semi-axis lengths | $c\sqrt{\lambda_i}$ — larger eigenvalue → longer axis → more spread in that direction |
| Orientation | $\boldsymbol{Q}$ — the ellipsoid is rotated by the eigenvector matrix |
| Volume scaling | $|\boldsymbol{\Sigma}|^{1/2} = \prod_i \lambda_i^{1/2}$ — product of semi-axis lengths |

**Diagonal $\boldsymbol{\Sigma}$ (no correlation):** Ellipsoid axes align with the coordinate axes. Semi-axes are $c\sigma_i = c\sqrt{\Sigma_{ii}}$.

**Identity $\boldsymbol{\Sigma} = \boldsymbol{I}$ (spherical):** All eigenvalues equal 1. Level sets are spheres. The density is rotationally symmetric.

### 4.1 Reduction to the Bivariate Normal

For $n=2$ with $\boldsymbol{\mu} = (\mu_X, \mu_Y)^{\top}$ and:

$$
\boldsymbol{\Sigma} = \begin{pmatrix} \sigma_X^2 & \rho\sigma_X\sigma_Y \\ \rho\sigma_X\sigma_Y & \sigma_Y^2 \end{pmatrix}
$$

we compute $|\boldsymbol{\Sigma}| = \sigma_X^2\sigma_Y^2 - \rho^2\sigma_X^2\sigma_Y^2 = \sigma_X^2\sigma_Y^2(1-\rho^2)$, so $|\boldsymbol{\Sigma}|^{1/2} = \sigma_X\sigma_Y\sqrt{1-\rho^2}$.

The inverse:

$$
\boldsymbol{\Sigma}^{-1} = \frac{1}{\sigma_X^2\sigma_Y^2(1-\rho^2)}\begin{pmatrix} \sigma_Y^2 & -\rho\sigma_X\sigma_Y \\ -\rho\sigma_X\sigma_Y & \sigma_X^2 \end{pmatrix}
$$

The quadratic form $(\boldsymbol{x}-\boldsymbol{\mu})^{\top}\boldsymbol{\Sigma}^{-1}(\boldsymbol{x}-\boldsymbol{\mu})$ expands to:

$$
\frac{1}{1-\rho^2}\left[\frac{(x-\mu_X)^2}{\sigma_X^2} - \frac{2\rho(x-\mu_X)(y-\mu_Y)}{\sigma_X\sigma_Y} + \frac{(y-\mu_Y)^2}{\sigma_Y^2}\right]
$$

Substituting into the general MVN formula:

$$
f(x,y) = \frac{1}{(2\pi)^{2/2}\,\sigma_X\sigma_Y\sqrt{1-\rho^2}}\exp\!\left(-\frac{1}{2(1-\rho^2)}\left[\cdots\right]\right)
= \frac{1}{2\pi\sigma_X\sigma_Y\sqrt{1-\rho^2}}\exp\!\left(-\frac{1}{2(1-\rho^2)}\left[\cdots\right]\right)
$$

This is exactly the bivariate normal density from Module 05 Unit 03. ✓

## 5. Statistical Bridge — Interpreting $\boldsymbol{\Sigma}$ in a Fitted Model `[PD]`

When we fit a Gaussian model to data, the estimated $\hat{\boldsymbol{\Sigma}}$ encodes the correlation structure of the variables. Reading it requires the eigendecomposition perspective from Section 4.

**The precision matrix** $\boldsymbol{\Lambda} = \boldsymbol{\Sigma}^{-1}$ (also called the **concentration matrix** or **information matrix**) appears directly in the density exponent. Its off-diagonal entry $\Lambda_{ij}$ encodes the **partial correlation** between $X_i$ and $X_j$ conditional on all other variables — not the marginal correlation in $\Sigma_{ij}$. In Gaussian graphical models (used in network inference and neuroscience), $\Lambda_{ij} = 0$ if and only if $X_i$ and $X_j$ are conditionally independent given all others.

**Principal directions of variation** are the eigenvectors of $\hat{\boldsymbol{\Sigma}}$. PCA (principal component analysis) is exactly the eigendecomposition of the sample covariance matrix — the first principal component is $\boldsymbol{q}_1$, the direction of maximum variance $\lambda_1$. This is why PCA is a natural consequence of the MVN structure rather than a separate algorithm.

### Section 1 — Symbolic Verification: Bivariate Reduction

Verify algebraically using SymPy that the general MVN formula with $n=2$ reduces to the bivariate normal density from Module 05 Unit 03.

```python
# ============================================================
# Section 1 — Symbolic MVN → bivariate normal reduction
# ============================================================
sx, sy, rho, x, y = sp.symbols('sigma_x sigma_y rho x y', positive=True)
mu_x, mu_y = sp.symbols('mu_x mu_y', real=True)

# Covariance matrix for n=2
Sigma = Matrix([[sx**2, rho*sx*sy],
                [rho*sx*sy, sy**2]])

# Compute |Σ| and Σ⁻¹
det_Sigma = Sigma.det()
Sigma_inv = Sigma.inv()

print('|Σ| =', sp.factor(det_Sigma))
print('|Σ|^(1/2) =', sp.factor(sp.sqrt(det_Sigma)))

# Quadratic form (x - μ)ᵀ Σ⁻¹ (x - μ)
delta = Matrix([x - mu_x, y - mu_y])
Q_form = (delta.T * Sigma_inv * delta)[0, 0]
Q_simplified = sp.simplify(Q_form)
print('\nQuadratic form (x-μ)ᵀ Σ⁻¹ (x-μ):')
sp.pprint(Q_simplified)

# Full density
norm_const = (2*pi)**(sp.Rational(2,2)) * sp.sqrt(det_Sigma)
f_mvn_n2 = sp.exp(-Q_simplified / 2) / norm_const

# Known bivariate form
Q_bvn = (((x-mu_x)/sx)**2 - 2*rho*(x-mu_x)*(y-mu_y)/(sx*sy) + ((y-mu_y)/sy)**2) / (1-rho**2)
f_bvn = sp.exp(-Q_bvn/2) / (2*pi*sx*sy*sp.sqrt(1-rho**2))

diff = sp.simplify(f_mvn_n2 - f_bvn)
print('\nMVN (n=2) - bivariate normal formula = 0?', diff == 0)
if diff != 0:
    print('  Simplified difference:', sp.trigsimp(diff))
```

**What do you see?** SymPy confirms that $|\boldsymbol{\Sigma}|^{1/2} = \sigma_X\sigma_Y\sqrt{1-\rho^2}$ and that the general MVN formula with $n=2$ reduces exactly to the bivariate normal density from Module 05 Unit 03. The algebraic path matches Section 4.1 exactly.

### Section 2 — Eigendecomposition and the Geometry of Level Sets

Visualize how different $\boldsymbol{\Sigma}$ matrices produce differently shaped and oriented density level sets. The eigenvectors give the axis directions; the eigenvalues give the axis lengths.

```python
# ============================================================
# Section 2 — Eigendecomposition and density level sets
# ============================================================
def mvn_density(X, Y, mu, Sigma):
    """Evaluate 2D MVN density on a meshgrid."""
    n = 2
    Sigma_inv = np.linalg.inv(Sigma)
    det_Sigma = np.linalg.det(Sigma)
    norm = (2*np.pi)**n * det_Sigma
    dx = np.stack([X - mu[0], Y - mu[1]], axis=-1)   # (..., 2)
    # Quadratic form at each point
    Q = np.einsum('...i,ij,...j->...', dx, Sigma_inv, dx)
    return np.exp(-Q/2) / np.sqrt(norm)

def plot_eigenvectors(ax, Sigma, mu, scale=1.5, colors=(TS_BLUE, TS_AMBER)):
    """Draw eigenvector arrows scaled by sqrt(eigenvalue)."""
    eigvals, eigvecs = np.linalg.eigh(Sigma)
    for i, (lam, q, col) in enumerate(zip(eigvals, eigvecs.T, colors)):
        v = scale * np.sqrt(lam) * q
        ax.annotate('', xy=mu + v, xytext=mu,
                    arrowprops=dict(arrowstyle='->', color=col, lw=2.5))
        ax.annotate('', xy=mu - v, xytext=mu,
                    arrowprops=dict(arrowstyle='->', color=col, lw=2.5, alpha=0.4))
        ax.text(*(mu + v*1.1), fr'$\sqrt{{\lambda_{i+1}}}\mathbf{{q}}_{i+1}$',
                color=col, fontsize=9, fontweight='bold')

configs = [
    (np.array([[1.0, 0.0],[0.0, 1.0]]),   'Spherical: $\\Sigma = I$'),
    (np.array([[3.0, 0.0],[0.0, 0.5]]),   'Axis-aligned: $\\Sigma = \\mathrm{diag}(3, 0.5)$'),
    (np.array([[2.0, 1.5],[1.5, 2.0]]),   'Correlated: $\\Sigma_{12}=1.5$'),
    (np.array([[1.0,-0.8],[-0.8, 1.0]]),  'Negatively correlated: $\\Sigma_{12}=-0.8$'),
]

mu0 = np.array([0.0, 0.0])
xg = np.linspace(-4, 4, 200)
Xg, Yg = np.meshgrid(xg, xg)

fig, axes = plt.subplots(2, 2, figsize=(13, 11))

for ax, (Sigma_cfg, title) in zip(axes.flat, configs):
    Fg = mvn_density(Xg, Yg, mu0, Sigma_cfg)
    cf = ax.contourf(Xg, Yg, Fg, levels=10, cmap='Blues', alpha=0.65)
    ax.contour(Xg, Yg, Fg, levels=10, colors=[TS_GRAY], alpha=0.4, linewidths=0.7)

    plot_eigenvectors(ax, Sigma_cfg, mu0)

    eigvals, _ = np.linalg.eigh(Sigma_cfg)
    ax.set_title(f'{title}\n'
                 fr'$\lambda_1={eigvals[0]:.2f}$, $\lambda_2={eigvals[1]:.2f}$, '
                 fr'$|\Sigma|^{{1/2}}={np.sqrt(np.linalg.det(Sigma_cfg)):.3f}$',
                 fontsize=10)
    ax.set_aspect('equal')
    ax.set_xlabel('$x_1$'); ax.set_ylabel('$x_2$')
    ax.set_xlim(-4, 4); ax.set_ylim(-4, 4)

plt.suptitle('MVN level sets — eigenvectors give axis directions, '
             'eigenvalues give axis lengths',
             fontsize=12, y=1.01)
plt.tight_layout()
plt.show()
```

**What do you see?**

- **Spherical**: circular level sets. All eigenvalues equal, so all directions have equal variance. The eigenvector arrows are all the same length.
- **Axis-aligned**: ellipses aligned with the coordinate axes. The longer axis corresponds to $\lambda_1 = 3$ (more variance in $x_1$). No off-diagonal terms → eigenvectors are the coordinate directions.
- **Correlated**: ellipses tilted at ~45°. Positive correlation rotates the principal axes toward the diagonal. The eigenvectors point in the direction of maximum and minimum variance.
- **Negatively correlated**: tilted in the opposite direction. Larger $x_1$ is associated with smaller $x_2$.

In all cases, $|\boldsymbol{\Sigma}|^{1/2}$ — printed in each title — is the product $\sqrt{\lambda_1\lambda_2}$, which equals the area of the ellipse divided by $\pi c^2$. Larger $|\boldsymbol{\Sigma}|$ means a more spread-out (lower-peaked) density.

### Section 3 — Mahalanobis Distance and Anomaly Detection

Compute the Mahalanobis distance for points in a 2D dataset and show how it differs from Euclidean distance when the data has correlation structure. Points with large Mahalanobis distance are anomalies under the fitted Gaussian model.

```python
# ============================================================
# Section 3 — Mahalanobis distance and anomaly detection
# ============================================================
rng = np.random.default_rng(42)
n_samp = 300

# Generate correlated 2D data
mu_true   = np.array([1.0, 2.0])
Sigma_true = np.array([[2.5, 1.8], [1.8, 2.0]])
data = rng.multivariate_normal(mu_true, Sigma_true, n_samp)

# Add a handful of true anomalies
anomalies = np.array([[5.0, 5.0], [-2.0, 5.0], [4.0, -1.0],
                       [0.0, 6.0], [-3.0, 0.5]])
all_pts = np.vstack([data, anomalies])

# Fit mean and covariance
mu_hat    = data.mean(axis=0)
Sigma_hat = np.cov(data.T)
Sigma_inv = np.linalg.inv(Sigma_hat)

def mahalanobis(pts, mu, Sigma_inv):
    delta = pts - mu
    return np.sqrt(np.einsum('ni,ij,nj->n', delta, Sigma_inv, delta))

def euclidean(pts, mu):
    return np.linalg.norm(pts - mu, axis=1)

d_mah = mahalanobis(all_pts, mu_hat, Sigma_inv)
d_euc = euclidean(all_pts, mu_hat)

# Chi-squared threshold: 95% of MVN mass within d_M² < chi2(0.95, df=2)
threshold = np.sqrt(stats.chi2.ppf(0.95, df=2))
is_anomaly = d_mah > threshold

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Left: scatter with Mahalanobis contours
ax = axes[0]
xg = np.linspace(-6, 8, 200)
Xg, Yg = np.meshgrid(xg, xg)
Fg = mvn_density(Xg, Yg, mu_hat, Sigma_hat)
ax.contour(Xg, Yg, Fg, levels=8, colors=[TS_GRAY], alpha=0.5, linewidths=0.8)

# Normal points
ax.scatter(all_pts[~is_anomaly, 0], all_pts[~is_anomaly, 1],
           c=d_mah[~is_anomaly], cmap='Blues', s=20, alpha=0.6,
           vmin=0, vmax=threshold*1.5)
# Anomalies
ax.scatter(all_pts[is_anomaly, 0], all_pts[is_anomaly, 1],
           color=TS_RED, s=100, marker='X', zorder=6,
           label=f'Anomaly ($d_M > {threshold:.2f}$)')
ax.plot(*mu_hat, '*', color=TS_AMBER, markersize=14, zorder=7, label=r'$\hat{\mu}$')
ax.set_xlabel('$x_1$'); ax.set_ylabel('$x_2$')
ax.set_title('Mahalanobis anomaly detection\n(colored by $d_M$, red X = anomaly)')
ax.legend(fontsize=9)

# Right: Mahalanobis vs Euclidean distance for anomalies
ax2 = axes[1]
ax2.scatter(d_euc[:n_samp], d_mah[:n_samp], color=TS_BLUE,
            alpha=0.3, s=15, label='Normal points')
ax2.scatter(d_euc[n_samp:], d_mah[n_samp:], color=TS_RED,
            s=120, marker='X', zorder=6, label='True anomalies')
ax2.axhline(threshold, color=TS_RED, lw=1.5, ls='--',
            label=f'$d_M$ threshold = {threshold:.2f} (95%)')
ax2.set_xlabel('Euclidean distance from $\\hat{\\mu}$')
ax2.set_ylabel('Mahalanobis distance $d_M$')
ax2.set_title('Mahalanobis vs Euclidean:\nnot always the same ordering')
ax2.legend(fontsize=9)

plt.suptitle('Mahalanobis distance — accounts for covariance structure',
             fontsize=11, y=1.01)
plt.tight_layout()
plt.show()

print(f'Fitted μ̂  = {np.round(mu_hat, 3)}')
print(f'Fitted Σ̂  =\n{np.round(Sigma_hat, 3)}')
print(f'\nAnomaly threshold (χ²₂, 95%): d_M > {threshold:.3f}')
print(f'Anomalies detected: {is_anomaly.sum()} out of {len(all_pts)} points')
```

**What do you see?**

- **Left**: The density level curves (ellipses) are aligned with the correlation structure of the data. Points colored darker are further from $\hat{\boldsymbol{\mu}}$ in Mahalanobis distance. The red X anomalies lie outside the 95% mass ellipsoid.
- **Right**: Euclidean distance and Mahalanobis distance do not always agree. A point can be Euclidean-close to $\hat{\boldsymbol{\mu}}$ but Mahalanobis-far (if it lies off the principal axes), and vice versa. Mahalanobis distance is the correct notion of distance for a Gaussian model.
- **The chi-squared threshold**: Under $\boldsymbol{X} \sim \mathcal{N}(\boldsymbol{\mu}, \boldsymbol{\Sigma})$, the squared Mahalanobis distance $d_{\mathcal{M}}^2$ follows a $\chi^2_n$ distribution. The 95th percentile of $\chi^2_2$ gives the radius of the ellipsoid containing 95% of the probability mass — this is how the anomaly threshold is set.

**Connection to AI security**: Mahalanobis-distance anomaly detection under a fitted Gaussian is a standard baseline for detecting out-of-distribution inputs, distributional shift, and adversarial examples in production ML systems.
