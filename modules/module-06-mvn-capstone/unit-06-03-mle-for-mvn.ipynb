## 1. The MVN Log-Likelihood

### 1.1 Setup

Observe $n$ iid samples $\boldsymbol{x}_1, \ldots, \boldsymbol{x}_n \overset{\text{iid}}{\sim} \mathcal{N}(\boldsymbol{\mu}, \boldsymbol{\Sigma})$. The log-likelihood is:

$$
\ell(\boldsymbol{\mu}, \boldsymbol{\Sigma}) = \sum_{i=1}^n \log f(\boldsymbol{x}_i) = -\frac{n}{2}\log|\boldsymbol{\Sigma}| - \frac{n \cdot p}{2}\log(2\pi) - \frac{1}{2}\sum_{i=1}^n (\boldsymbol{x}_i - \boldsymbol{\mu})^{\top}\boldsymbol{\Sigma}^{-1}(\boldsymbol{x}_i - \boldsymbol{\mu})
$$

where $p$ is the dimension (using $p$ to avoid confusion with the sample size $n$).

### 1.2 Trace Trick

The quadratic form sum can be rewritten using the **trace trick**: for a scalar $a = \boldsymbol{v}^{\top}\boldsymbol{A}\boldsymbol{v}$, we have $a = \text{tr}(\boldsymbol{v}^{\top}\boldsymbol{A}\boldsymbol{v}) = \text{tr}(\boldsymbol{A}\boldsymbol{v}\boldsymbol{v}^{\top})$. Applying this:

$$
\sum_{i=1}^n (\boldsymbol{x}_i - \boldsymbol{\mu})^{\top}\boldsymbol{\Sigma}^{-1}(\boldsymbol{x}_i - \boldsymbol{\mu}) = \text{tr}\!\left(\boldsymbol{\Sigma}^{-1}\sum_{i=1}^n (\boldsymbol{x}_i - \boldsymbol{\mu})(\boldsymbol{x}_i - \boldsymbol{\mu})^{\top}\right)
$$

Define the **scatter matrix** $\boldsymbol{S}(\boldsymbol{\mu}) = \sum_{i=1}^n (\boldsymbol{x}_i - \boldsymbol{\mu})(\boldsymbol{x}_i - \boldsymbol{\mu})^{\top}$. The log-likelihood becomes:

$$
\ell(\boldsymbol{\mu}, \boldsymbol{\Sigma}) = -\frac{np}{2}\log(2\pi) - \frac{n}{2}\log|\boldsymbol{\Sigma}| - \frac{1}{2}\text{tr}\!\left(\boldsymbol{\Sigma}^{-1}\boldsymbol{S}(\boldsymbol{\mu})\right)
$$

This form separates the dependence on $\boldsymbol{\mu}$ (through $\boldsymbol{S}$) from the dependence on $\boldsymbol{\Sigma}$.

---

## 2. MLE for $\boldsymbol{\mu}$

### 2.1 Gradient with Respect to $\boldsymbol{\mu}$

Only the scatter matrix term depends on $\boldsymbol{\mu}$. Using the gradient rule $\nabla_{\boldsymbol{\mu}}\,(\boldsymbol{a} - \boldsymbol{\mu})^{\top}\boldsymbol{A}(\boldsymbol{a} - \boldsymbol{\mu}) = -2\boldsymbol{A}(\boldsymbol{a} - \boldsymbol{\mu})$ (from Module 03 Unit 01):

$$
\nabla_{\boldsymbol{\mu}}\,\ell = -\frac{1}{2}\sum_{i=1}^n \nabla_{\boldsymbol{\mu}}\,(\boldsymbol{x}_i - \boldsymbol{\mu})^{\top}\boldsymbol{\Sigma}^{-1}(\boldsymbol{x}_i - \boldsymbol{\mu}) = \boldsymbol{\Sigma}^{-1}\sum_{i=1}^n (\boldsymbol{x}_i - \boldsymbol{\mu})
$$

### 2.2 Setting to Zero

$$
\boldsymbol{\Sigma}^{-1}\sum_{i=1}^n (\boldsymbol{x}_i - \hat{\boldsymbol{\mu}}) = \boldsymbol{0}
$$

Since $\boldsymbol{\Sigma}^{-1}$ is invertible, this requires $\sum_{i=1}^n (\boldsymbol{x}_i - \hat{\boldsymbol{\mu}}) = \boldsymbol{0}$, giving:

$$
\boxed{\hat{\boldsymbol{\mu}}_{\text{MLE}} = \frac{1}{n}\sum_{i=1}^n \boldsymbol{x}_i = \bar{\boldsymbol{x}}}
$$

The MLE of the mean is the sample mean — the direct multivariate generalization of $\hat{\mu} = \bar{x}$ from Module 03 Unit 01.

---

## 3. MLE for $\boldsymbol{\Sigma}$

Substituting $\hat{\boldsymbol{\mu}} = \bar{\boldsymbol{x}}$, the profile log-likelihood in $\boldsymbol{\Sigma}$ is:

$$
\ell(\boldsymbol{\Sigma}) = \text{const} - \frac{n}{2}\log|\boldsymbol{\Sigma}| - \frac{1}{2}\text{tr}\!\left(\boldsymbol{\Sigma}^{-1}\hat{\boldsymbol{S}}\right)
$$

where $\hat{\boldsymbol{S}} = \sum_{i=1}^n (\boldsymbol{x}_i - \bar{\boldsymbol{x}})(\boldsymbol{x}_i - \bar{\boldsymbol{x}})^{\top}$ is the scatter matrix at the MLE mean.

### 3.1 Matrix Derivative Rules

Two results from matrix calculus (derivations available in the exercise PDF):

$$
\frac{\partial}{\partial \boldsymbol{\Sigma}}\log|\boldsymbol{\Sigma}| = \boldsymbol{\Sigma}^{-\top} = \boldsymbol{\Sigma}^{-1} \quad (\text{since } \boldsymbol{\Sigma} \text{ is symmetric})
$$

$$
\frac{\partial}{\partial \boldsymbol{\Sigma}}\text{tr}\!\left(\boldsymbol{\Sigma}^{-1}\hat{\boldsymbol{S}}\right) = -(\boldsymbol{\Sigma}^{-1}\hat{\boldsymbol{S}}\boldsymbol{\Sigma}^{-1})^{\top} = -\boldsymbol{\Sigma}^{-1}\hat{\boldsymbol{S}}\boldsymbol{\Sigma}^{-1}
$$

### 3.2 Setting the Derivative to Zero

$$
\frac{\partial \ell}{\partial \boldsymbol{\Sigma}} = -\frac{n}{2}\boldsymbol{\Sigma}^{-1} + \frac{1}{2}\boldsymbol{\Sigma}^{-1}\hat{\boldsymbol{S}}\boldsymbol{\Sigma}^{-1} = \boldsymbol{0}
$$

Multiply both sides on the left and right by $\boldsymbol{\Sigma}$:

$$
-n\boldsymbol{I} + \hat{\boldsymbol{S}}\boldsymbol{\Sigma}^{-1} = \boldsymbol{0} \qquad \Longrightarrow \qquad \hat{\boldsymbol{S}}\boldsymbol{\Sigma}^{-1} = n\boldsymbol{I}
$$

$$
\boldsymbol{\Sigma} = \frac{1}{n}\hat{\boldsymbol{S}} = \frac{1}{n}\sum_{i=1}^n (\boldsymbol{x}_i - \bar{\boldsymbol{x}})(\boldsymbol{x}_i - \bar{\boldsymbol{x}})^{\top}
$$

$$
\boxed{\hat{\boldsymbol{\Sigma}}_{\text{MLE}} = \frac{1}{n}\sum_{i=1}^n (\boldsymbol{x}_i - \bar{\boldsymbol{x}})(\boldsymbol{x}_i - \bar{\boldsymbol{x}})^{\top}}
$$

The MLE of the covariance is the **biased sample covariance matrix** (divides by $n$, not $n-1$). The unbiased estimate divides by $n-1$ — this correction is the **Bessel correction**, and it accounts for the fact that we estimate $\boldsymbol{\mu}$ from the data before computing the scatter.

---

## 4. Fisher Information and the Precision Matrix

### 4.1 Fisher Information for $\boldsymbol{\mu}$

The Fisher information matrix for $\boldsymbol{\mu}$ (with $\boldsymbol{\Sigma}$ known) is the negative expected Hessian of the log-likelihood:

$$
\mathcal{I}(\boldsymbol{\mu}) = -\mathbb{E}\!\left[\nabla_{\boldsymbol{\mu}}^2\,\ell\right] = n\boldsymbol{\Sigma}^{-1}
$$

The Fisher information for the mean is the precision matrix $\boldsymbol{\Sigma}^{-1}$, scaled by $n$. More data means more information — the precision accumulates linearly with sample size.

### 4.2 Cramér-Rao Bound

The Cramér-Rao lower bound states that for any unbiased estimator $\hat{\boldsymbol{\mu}}$ of $\boldsymbol{\mu}$:

$$
\text{Cov}(\hat{\boldsymbol{\mu}}) \geq \mathcal{I}(\boldsymbol{\mu})^{-1} = \frac{1}{n}\boldsymbol{\Sigma}
$$

The MLE $\hat{\boldsymbol{\mu}} = \bar{\boldsymbol{x}}$ achieves this bound — it is the **minimum variance unbiased estimator** of $\boldsymbol{\mu}$. This is the formal sense in which the sample mean is the best possible estimator.

### 4.3 The Full Circle

The Fisher information $\mathcal{I}(\boldsymbol{\mu}) = n\boldsymbol{\Sigma}^{-1}$ connects every major object in this course:

- $\boldsymbol{\Sigma}^{-1}$ appears in the MVN density exponent (Unit 01)
- $\boldsymbol{\Sigma}^{-1}$ is the Schur complement's building block (Unit 02)
- $\mathcal{I}(\boldsymbol{\mu}) = -\boldsymbol{H}\ell$ is the negative Hessian (Module 04 Unit 01)
- The score function $\nabla_{\boldsymbol{\mu}}\ell = \boldsymbol{\Sigma}^{-1}\sum_i(\boldsymbol{x}_i - \boldsymbol{\mu})$ is zero at the MLE (Module 03 Unit 01)
- The Bayesian posterior precision $\boldsymbol{S}_n^{-1} = \boldsymbol{S}_0^{-1} + n\boldsymbol{\Sigma}^{-1}$ accumulates Fisher information (Unit 02)

The precision matrix $\boldsymbol{\Sigma}^{-1}$ is the thread that runs through all of them.

### Section 1 — MLE Derivation: Numerical Gradient Check

Verify the MLE formulas by numerically maximizing the log-likelihood and confirming the optimum matches $\bar{\boldsymbol{x}}$ and $\hat{\boldsymbol{\Sigma}}$.

```python
# ============================================================
# Section 1 — MLE formulas vs numerical optimization
# ============================================================
from scipy.optimize import minimize

rng = np.random.default_rng(42)
p = 3;  n_obs = 80

mu_true = np.array([1.0, -0.5, 2.0])
# Random SPD covariance
A = rng.standard_normal((p, p))
Sigma_true = A @ A.T / p + 0.5 * np.eye(p)

data = rng.multivariate_normal(mu_true, Sigma_true, n_obs)

# --- Analytical MLE ---
mu_mle    = data.mean(axis=0)
Sigma_mle = np.cov(data.T, bias=True)   # bias=True → divide by n (MLE)
Sigma_unbiased = np.cov(data.T)          # default → divide by n-1

# --- Log-likelihood function ---
def mvn_loglik(data, mu, Sigma):
    """Sum of log-densities for n iid MVN observations."""
    rv = stats.multivariate_normal(mean=mu, cov=Sigma, allow_singular=False)
    return rv.logpdf(data).sum()

# --- Verify gradient is zero at MLE ---
eps = 1e-5
grad_mu_at_mle = np.linalg.solve(Sigma_mle, data.sum(axis=0) - n_obs * mu_mle)

print('=== Verification that gradient = 0 at MLE ===')
print(f'‖∇_μ ℓ‖ at μ̂_MLE  = {np.linalg.norm(grad_mu_at_mle):.2e}  (should be ≈ 0)')

# Confirm MLE is a maximum by checking nearby points have lower ℓ
ll_at_mle   = mvn_loglik(data, mu_mle, Sigma_mle)
ll_perturbed = mvn_loglik(data, mu_mle + 0.1, Sigma_mle)
print(f'\nℓ at μ̂_MLE       = {ll_at_mle:.4f}')
print(f'ℓ at μ̂_MLE + 0.1 = {ll_perturbed:.4f}  (should be < above)')
print(f'MLE is a maximum: {ll_at_mle > ll_perturbed}')

print(f'\n=== Estimates ===')
print(f'True μ         = {np.round(mu_true, 4)}')
print(f'MLE  μ̂         = {np.round(mu_mle, 4)}')
print(f'\nMLE Σ̂ (bias=n):\n{np.round(Sigma_mle, 4)}')
print(f'\nUnbiased Σ̂ (÷n-1):\n{np.round(Sigma_unbiased, 4)}')
print(f'\nTrue Σ:\n{np.round(Sigma_true, 4)}')
print(f'\nBessel correction: MLE vs unbiased differ by factor {n_obs/(n_obs-1):.4f}')
```

**What do you see?** The gradient $\nabla_{\boldsymbol{\mu}}\,\ell$ evaluated at $\hat{\boldsymbol{\mu}} = \bar{\boldsymbol{x}}$ is numerically zero (machine precision). Perturbing the mean away from the MLE reduces the log-likelihood — confirming the MLE is a maximum. The biased MLE $\hat{\boldsymbol{\Sigma}}$ and unbiased estimate differ by the Bessel correction factor $n/(n-1)$, which is negligible for large $n$.

### Section 2 — Log-Likelihood Surface and the MLE as a Maximum

Visualize the MVN log-likelihood as a function of $\boldsymbol{\mu}$ (with $\boldsymbol{\Sigma}$ fixed) and as a function of one covariance entry, confirming the MLE is the global maximum.

```python
# ============================================================
# Section 2 — Log-likelihood surfaces
# ============================================================
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# --- Left: ℓ as a function of (μ1, μ2) with μ3, Σ fixed ---
ax = axes[0]
mu1_g = np.linspace(mu_mle[0]-1.5, mu_mle[0]+1.5, 100)
mu2_g = np.linspace(mu_mle[1]-1.5, mu_mle[1]+1.5, 100)
M1, M2 = np.meshgrid(mu1_g, mu2_g)
LL_mu = np.zeros_like(M1)
for i in range(M1.shape[0]):
    for j in range(M1.shape[1]):
        mu_ij = np.array([M1[i,j], M2[i,j], mu_mle[2]])
        LL_mu[i,j] = mvn_loglik(data, mu_ij, Sigma_mle)

cf = ax.contourf(M1, M2, LL_mu, levels=20, cmap='Blues', alpha=0.70)
ax.contour(M1, M2, LL_mu, levels=20, colors=[TS_GRAY], alpha=0.4, linewidths=0.7)
plt.colorbar(cf, ax=ax, label=r'$\ell(\mu_1, \mu_2)$')
ax.plot(*mu_mle[:2], '*', color=TS_AMBER, markersize=14, zorder=6,
        label=fr'MLE $\hat{{\mu}} = ({mu_mle[0]:.2f}, {mu_mle[1]:.2f})$')
ax.set_xlabel(r'$\mu_1$'); ax.set_ylabel(r'$\mu_2$')
ax.set_title(r'$\ell(\mu_1, \mu_2)$ with $\mu_3, \Sigma$ fixed')
ax.legend(fontsize=9)

# --- Right: ℓ as a function of one variance entry Σ11 ---
ax2 = axes[1]
sigma11_grid = np.linspace(Sigma_mle[0,0]*0.3, Sigma_mle[0,0]*2.5, 200)
ll_sigma11 = []
for s11 in sigma11_grid:
    S_perturb = Sigma_mle.copy()
    S_perturb[0,0] = s11
    # Ensure PD
    eigvals = np.linalg.eigvalsh(S_perturb)
    if eigvals.min() > 1e-8:
        ll_sigma11.append(mvn_loglik(data, mu_mle, S_perturb))
    else:
        ll_sigma11.append(np.nan)

ax2.plot(sigma11_grid, ll_sigma11, color=TS_BLUE, lw=2.5)
ax2.axvline(Sigma_mle[0,0], color=TS_AMBER, lw=2, ls='--',
            label=fr'MLE $\hat{{\Sigma}}_{{11}} = {Sigma_mle[0,0]:.3f}$')
ax2.set_xlabel(r'$\Sigma_{11}$ (variance of $X_1$)')
ax2.set_ylabel(r'$\ell$')
ax2.set_title(r'$\ell$ as a function of $\Sigma_{11}$ (other entries fixed at MLE)')
ax2.legend(fontsize=9)

plt.suptitle('MVN log-likelihood — concave in $\\mu$, concave in $\\Sigma$',
             fontsize=11, y=1.01)
plt.tight_layout()
plt.show()
```

**What do you see?**

- **Left**: The log-likelihood in $(\mu_1, \mu_2)$ is a concave surface with a unique maximum at the MLE (amber star). The level curves are ellipses — their shape is determined by $\boldsymbol{\Sigma}^{-1}$ (the Hessian of the log-likelihood in $\boldsymbol{\mu}$ is $-n\boldsymbol{\Sigma}^{-1}$). This is the matrix version of the scalar plot from Module 03 Unit 01.
- **Right**: The log-likelihood as a function of $\Sigma_{11}$ is also concave with a unique maximum at the MLE. The asymmetry — it falls steeply for small $\Sigma_{11}$ and gently for large — reflects the $-\frac{n}{2}\log|\boldsymbol{\Sigma}|$ term: too-small variance is heavily penalized.

### Section 3 — Fitting an MVN to Real Data: Iris Dataset

Apply the MLE formulas to the Iris dataset, fit a class-conditional MVN to each species, visualize the fitted densities, and use Mahalanobis distance for classification.

```python
# ============================================================
# Section 3 — MVN fitted to Iris: MLE and classification
# ============================================================
iris = load_iris()
X_iris = iris.data[:, :2]   # use first two features: sepal length, sepal width
y_iris = iris.target
class_names = iris.target_names
colors_cls = [TS_BLUE, TS_AMBER, TS_GREEN]

# --- MLE per class ---
models = {}
for c in range(3):
    Xc = X_iris[y_iris == c]
    mu_c     = Xc.mean(axis=0)
    Sigma_c  = np.cov(Xc.T, bias=True)   # MLE: divide by n
    n_c      = len(Xc)
    models[c] = {'mu': mu_c, 'Sigma': Sigma_c, 'n': n_c,
                 'rv': stats.multivariate_normal(mu_c, Sigma_c)}
    print(f'{class_names[c]:12s}  n={n_c}')
    print(f'  μ̂ = {np.round(mu_c, 3)}')
    print(f'  Σ̂ =\n{np.round(Sigma_c, 4)}')
    eigs = np.linalg.eigvalsh(Sigma_c)
    print(f'  eigenvalues = {np.round(eigs, 4)}  (both > 0: {np.all(eigs>0)})')

# --- Classification by max log-likelihood (LDA-style) ---
log_probs = np.column_stack([models[c]['rv'].logpdf(X_iris) for c in range(3)])
y_pred = log_probs.argmax(axis=1)
accuracy = (y_pred == y_iris).mean()
print(f'\nMVN classification accuracy: {accuracy:.3f}')

# --- Visualization ---
xg = np.linspace(3.8, 8.5, 200)
yg = np.linspace(1.5, 4.8, 200)
Xg, Yg = np.meshgrid(xg, yg)
pts = np.stack([Xg.ravel(), Yg.ravel()], axis=1)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Left: scatter + fitted contours
ax = axes[0]
for c in range(3):
    Xc = X_iris[y_iris == c]
    ax.scatter(Xc[:,0], Xc[:,1], color=colors_cls[c],
               alpha=0.6, s=35, label=class_names[c])
    Z = models[c]['rv'].pdf(pts).reshape(200,200)
    ax.contour(Xg, Yg, Z, levels=5, colors=[colors_cls[c]], alpha=0.7, linewidths=1.5)
    mu_c = models[c]['mu']
    ax.plot(*mu_c, '*', color=colors_cls[c], markersize=14, zorder=6)

ax.set_xlabel('Sepal length (cm)')
ax.set_ylabel('Sepal width (cm)')
ax.set_title('Fitted MVN per class — MLE estimates')
ax.legend(fontsize=9)

# Right: Mahalanobis distance to each class
ax2 = axes[1]
d_mah_all = np.column_stack([
    np.sqrt(np.einsum('ni,ij,nj->n',
                      X_iris - models[c]['mu'],
                      np.linalg.inv(models[c]['Sigma']),
                      X_iris - models[c]['mu']))
    for c in range(3)
])

for c in range(3):
    Xc_idx = y_iris == c
    ax2.scatter(np.where(Xc_idx)[0], d_mah_all[Xc_idx, c],
                color=colors_cls[c], s=25, alpha=0.7, label=f'{class_names[c]} → own class')

ax2.axhline(np.sqrt(stats.chi2.ppf(0.95, df=2)), color=TS_RED,
            ls='--', lw=1.5, label='95% threshold ($\\chi^2_2$)')
ax2.set_xlabel('Sample index')
ax2.set_ylabel('Mahalanobis distance $d_M$')
ax2.set_title('Mahalanobis distance to own class')
ax2.legend(fontsize=8)

plt.suptitle(f'Iris: MVN MLE + Mahalanobis classification (accuracy={accuracy:.1%})',
             fontsize=11, y=1.01)
plt.tight_layout()
plt.show()
```

**What do you see?**

- **Left**: Three fitted MVN densities — one per species. The elliptical contours capture the correlation structure within each class. Setosa (blue) is well-separated; Versicolor and Virginica overlap, which is why the accuracy is not perfect.
- **Right**: Mahalanobis distance from each observation to its own class center. Most points fall below the 95% threshold — they are typical members of their class. Outliers above the threshold are atypical observations (genuine boundary cases or measurement noise).
- The classifier assigns each observation to the class with the highest log-likelihood — equivalent to the class with smallest Mahalanobis distance, adjusted for each class's covariance volume $|\boldsymbol{\Sigma}_c|^{1/2}$.

**This is quadratic discriminant analysis (QDA)** — a direct consequence of fitting a separate MVN per class and classifying by maximum likelihood. When all class covariances are forced equal, it reduces to linear discriminant analysis (LDA) and the decision boundaries become hyperplanes.
