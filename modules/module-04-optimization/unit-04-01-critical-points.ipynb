## Why This Unit

Module 03 gave you the gradient and showed how it drives optimization via gradient descent. But gradient descent only tells you how to move — it does not tell you what you have found when you stop. When $\nabla f(\boldsymbol{x}^*) = \boldsymbol{0}$, have you found a minimum, a maximum, or a saddle point?

The answer requires second-order information — how the gradient is changing, not just where it is zero. The **Hessian matrix** captures this, and the **second-derivative test** classifies what you have found.

This unit builds the Hessian and the classification machinery. Unit 02 extends this to convexity — the global property that guarantees a unique minimum exists — and to Newton's method, which uses the Hessian to accelerate optimization.

## 1. Critical Points

**Definition.** A point $\boldsymbol{x}^* \in \mathbb{R}^n$ is a **critical point** (or **stationary point**) of $f : \mathbb{R}^n \to \mathbb{R}$ if:

$$
\nabla f(\boldsymbol{x}^*) = \boldsymbol{0}
$$

That is, all partial derivatives vanish simultaneously at $\boldsymbol{x}^*$.

**Why necessary.** At a local minimum or maximum in the interior of the domain, $f$ cannot be increasing or decreasing in any direction. Since the directional derivative $D_{\boldsymbol{u}}f(\boldsymbol{x}^*) = \nabla f(\boldsymbol{x}^*) \cdot \boldsymbol{u}$ must be zero for every unit vector $\boldsymbol{u}$, the only possibility is $\nabla f(\boldsymbol{x}^*) = \boldsymbol{0}$.

> **Critical $\neq$ extremum.** The condition $\nabla f = \boldsymbol{0}$ is necessary but not sufficient for a minimum or maximum. A critical point can also be a **saddle point** — a point where the gradient is zero but $f$ has neither a local minimum nor a local maximum.

### 1.1 Types of Critical Points

| Type | Description | Single-variable analogy |
|---|---|---|
| **Local minimum** | $f(\boldsymbol{x}) \geq f(\boldsymbol{x}^*)$ for $\boldsymbol{x}$ near $\boldsymbol{x}^*$ | $f''(x^*) > 0$ |
| **Local maximum** | $f(\boldsymbol{x}) \leq f(\boldsymbol{x}^*)$ for $\boldsymbol{x}$ near $\boldsymbol{x}^*$ | $f''(x^*) < 0$ |
| **Saddle point** | Neither; $f$ increases in some directions and decreases in others | $f''(x^*) = 0$ (inconclusive) |
| **Global minimum** | $f(\boldsymbol{x}) \geq f(\boldsymbol{x}^*)$ for all $\boldsymbol{x}$ | — |

**Worked Example 1.1.** Find the critical points of $f(x, y) = x^3 - 3x + y^2 - 2y$.

$$
\frac{\partial f}{\partial x} = 3x^2 - 3 = 0 \implies x = \pm 1
$$
$$
\frac{\partial f}{\partial y} = 2y - 2 = 0 \implies y = 1
$$

Two critical points: $(1, 1)$ and $(-1, 1)$. Classification requires the second-derivative test.

## 2. The Hessian Matrix

The Hessian is the matrix of all second-order partial derivatives. It is the multivariable analogue of $f''(x)$.

**Definition.** Let $f : \mathbb{R}^n \to \mathbb{R}$ be twice differentiable. The **Hessian matrix** of $f$ at $\boldsymbol{x}$ is:

$$
\boldsymbol{H}f(\boldsymbol{x}) \coloneqq \nabla^2 f(\boldsymbol{x}) \coloneqq \begin{pmatrix} \dfrac{\partial^2 f}{\partial x_1^2} & \dfrac{\partial^2 f}{\partial x_1 \partial x_2} & \cdots & \dfrac{\partial^2 f}{\partial x_1 \partial x_n} \\[10pt] \dfrac{\partial^2 f}{\partial x_2 \partial x_1} & \dfrac{\partial^2 f}{\partial x_2^2} & \cdots & \dfrac{\partial^2 f}{\partial x_2 \partial x_n} \\[6pt] \vdots & \vdots & \ddots & \vdots \\[4pt] \dfrac{\partial^2 f}{\partial x_n \partial x_1} & \dfrac{\partial^2 f}{\partial x_n \partial x_2} & \cdots & \dfrac{\partial^2 f}{\partial x_n^2} \end{pmatrix} \in \mathbb{R}^{n \times n}
$$

Entry $(i, j)$ is $\frac{\partial^2 f}{\partial x_i \partial x_j}$ — differentiate first with respect to $x_j$, then with respect to $x_i$.

**Symmetry.** Under mild smoothness conditions (Schwarz's theorem), mixed partial derivatives commute: $\frac{\partial^2 f}{\partial x_i \partial x_j} = \frac{\partial^2 f}{\partial x_j \partial x_i}$. Therefore the Hessian is always **symmetric**: $\boldsymbol{H}f = (\boldsymbol{H}f)^{\top}$. This symmetry is essential — it means the Hessian has real eigenvalues and an orthonormal eigenbasis.

**The Hessian as the Jacobian of the gradient.** Notice that $\boldsymbol{H}f = \boldsymbol{J}_{\nabla f}$ — the Hessian is exactly the Jacobian of the gradient map $\nabla f : \mathbb{R}^n \to \mathbb{R}^n$. Every tool from Module 03 Unit 03 applies here.

**Worked Example 2.1.** Compute the Hessian of $f(x, y) = x^3 - 3x + y^2 - 2y$ from Example 1.1.

First-order partials: $f_x = 3x^2 - 3$, $f_y = 2y - 2$.

Second-order partials:
$$
\frac{\partial^2 f}{\partial x^2} = 6x \qquad \frac{\partial^2 f}{\partial y^2} = 2 \qquad \frac{\partial^2 f}{\partial x \partial y} = 0
$$

$$
\boldsymbol{H}f(x, y) = \begin{pmatrix} 6x & 0 \\ 0 & 2 \end{pmatrix}
$$

At the two critical points:
$$
\boldsymbol{H}f(1, 1) = \begin{pmatrix} 6 & 0 \\ 0 & 2 \end{pmatrix} \qquad \boldsymbol{H}f(-1, 1) = \begin{pmatrix} -6 & 0 \\ 0 & 2 \end{pmatrix}
$$

**Worked Example 2.2.** Compute the Hessian of $f(x, y) = e^x \sin y$.

$$
f_x = e^x \sin y \qquad f_y = e^x \cos y
$$
$$
f_{xx} = e^x \sin y \qquad f_{yy} = -e^x \sin y \qquad f_{xy} = e^x \cos y
$$

$$
\boldsymbol{H}f(x, y) = e^x\begin{pmatrix} \sin y & \cos y \\ \cos y & -\sin y \end{pmatrix}
$$

Note that $f_{xy} = f_{yx} = e^x \cos y$ — Schwarz's theorem confirmed.

## 3. The Second-Derivative Test in Two Variables

For $f : \mathbb{R}^2 \to \mathbb{R}$ with a critical point at $(x^*, y^*)$, the second-derivative test uses two scalar quantities derived from the Hessian:

$$
\boldsymbol{H}f(x^*, y^*) = \begin{pmatrix} f_{xx} & f_{xy} \\ f_{xy} & f_{yy} \end{pmatrix}
$$

Define the **discriminant**:

$$
D \coloneqq \det(\boldsymbol{H}f) = f_{xx}\,f_{yy} - f_{xy}^2
$$

**Second-Derivative Test:**

| Condition | Classification |
|---|---|
| $D > 0$ and $f_{xx} > 0$ | **Local minimum** |
| $D > 0$ and $f_{xx} < 0$ | **Local maximum** |
| $D < 0$ | **Saddle point** |
| $D = 0$ | **Inconclusive** — higher-order analysis needed |

**Why the determinant?** The discriminant $D = f_{xx}f_{yy} - f_{xy}^2$ is the product of the eigenvalues of the Hessian (since $\det(\boldsymbol{H}) = \lambda_1 \lambda_2$). A positive determinant means both eigenvalues have the same sign (both positive → minimum, both negative → maximum). A negative determinant means one eigenvalue is positive and one negative — a saddle point.

**Worked Example 3.1.** Classify the critical points of $f(x, y) = x^3 - 3x + y^2 - 2y$ from Example 1.1.

**At $(1, 1)$:** $\boldsymbol{H}f = \begin{pmatrix} 6 & 0 \\ 0 & 2 \end{pmatrix}$.

$D = (6)(2) - 0 = 12 > 0$ and $f_{xx} = 6 > 0$ $\implies$ **local minimum**.

**At $(-1, 1)$:** $\boldsymbol{H}f = \begin{pmatrix} -6 & 0 \\ 0 & 2 \end{pmatrix}$.

$D = (-6)(2) - 0 = -12 < 0$ $\implies$ **saddle point**.

**Worked Example 3.2.** Find and classify critical points of $f(x, y) = x^2 + xy + y^2 - 3x$.

$$
f_x = 2x + y - 3 = 0 \qquad f_y = x + 2y = 0
$$

From $f_y$: $x = -2y$. Substituting into $f_x$: $2(-2y) + y - 3 = 0 \implies -3y = 3 \implies y = -1$, $x = 2$.

Critical point: $(2, -1)$.

$$
\boldsymbol{H}f = \begin{pmatrix} 2 & 1 \\ 1 & 2 \end{pmatrix} \qquad D = 4 - 1 = 3 > 0, \quad f_{xx} = 2 > 0 \implies \text{local minimum.}
$$

### 3.1 Extension to $n$ Variables

For $f : \mathbb{R}^n \to \mathbb{R}$, the second-derivative test generalises using eigenvalues of the $n \times n$ Hessian:

- All eigenvalues of $\boldsymbol{H}f(\boldsymbol{x}^*)$ positive $\implies$ **local minimum** (positive definite Hessian)
- All eigenvalues negative $\implies$ **local maximum** (negative definite)
- Mixed signs $\implies$ **saddle point** (indefinite)
- Any eigenvalue zero $\implies$ **inconclusive** (positive/negative semi-definite)

We develop positive definiteness fully in Unit 02.

## 4. Second-Order Taylor Expansion

The Hessian arises naturally from the Taylor expansion of $f$ near a critical point. For $f : \mathbb{R}^n \to \mathbb{R}$ twice differentiable at $\boldsymbol{x}^*$:

$$
f(\boldsymbol{x}^* + \boldsymbol{h}) = f(\boldsymbol{x}^*) + \underbrace{\nabla f(\boldsymbol{x}^*)^{\top} \boldsymbol{h}}_{= 0 \text{ at critical point}} + \frac{1}{2}\boldsymbol{h}^{\top}\boldsymbol{H}f(\boldsymbol{x}^*)\boldsymbol{h} + O(\|\boldsymbol{h}\|^3)
$$

At a critical point, the linear term vanishes and the behavior is determined by the **quadratic form** $\boldsymbol{h}^{\top}\boldsymbol{H}f\,\boldsymbol{h}$:

$$
f(\boldsymbol{x}^* + \boldsymbol{h}) - f(\boldsymbol{x}^*) \approx \frac{1}{2}\boldsymbol{h}^{\top}\boldsymbol{H}f(\boldsymbol{x}^*)\boldsymbol{h}
$$

- If $\boldsymbol{h}^{\top}\boldsymbol{H}f\,\boldsymbol{h} > 0$ for all $\boldsymbol{h} \neq \boldsymbol{0}$: $f$ increases in every direction from $\boldsymbol{x}^*$ → **local minimum**.
- If $\boldsymbol{h}^{\top}\boldsymbol{H}f\,\boldsymbol{h} < 0$ for all $\boldsymbol{h} \neq \boldsymbol{0}$: $f$ decreases in every direction → **local maximum**.
- If the sign depends on $\boldsymbol{h}$: $f$ increases in some directions, decreases in others → **saddle point**.

This is the formal connection between the second-derivative test and the geometry of $f$: the test is asking whether the quadratic form $\boldsymbol{h}^{\top}\boldsymbol{H}f\,\boldsymbol{h}$ is always positive, always negative, or sign-indefinite.

## 5. Statistical Bridge — Hessian of the Gaussian Log-Likelihood `[MLE]`

In Module 03 Unit 01 we found the MLE for the Gaussian model by setting $\nabla_{\boldsymbol{\theta}}\,\ell = \boldsymbol{0}$ and solving:

$$
\hat{\mu}_{\text{MLE}} = \bar{x} \qquad \hat{\sigma}^2_{\text{MLE}} = \frac{1}{n}\sum_{i=1}^n (x_i - \bar{x})^2
$$

We stated that this was a maximum — but never proved it. The Hessian closes that gap.

### 5.1 The Log-Likelihood

For $x_1, \ldots, x_n \overset{\text{iid}}{\sim} \mathcal{N}(\mu, \sigma^2)$, writing $v = \sigma^2 > 0$:

$$
\ell(\mu, v) = -\frac{n}{2}\log(2\pi v) - \frac{1}{2v}\sum_{i=1}^n (x_i - \mu)^2
$$

### 5.2 Second-Order Partials

We already have the first-order partials from Module 03. The second-order partials are:

$$
\frac{\partial^2 \ell}{\partial \mu^2} = -\frac{n}{v}
$$

$$
\frac{\partial^2 \ell}{\partial v^2} = \frac{n}{2v^2} - \frac{1}{v^3}\sum_{i=1}^n (x_i - \mu)^2
$$

$$
\frac{\partial^2 \ell}{\partial \mu \partial v} = -\frac{1}{v^2}\sum_{i=1}^n (x_i - \mu)
$$

### 5.3 Evaluating at the MLE

Substitute $\hat{\mu} = \bar{x}$ and $\hat{v} = \hat{\sigma}^2 = \frac{1}{n}\sum(x_i - \bar{x})^2$.

**Off-diagonal:** $\sum_i (x_i - \bar{x}) = 0$ by definition of $\bar{x}$, so $\frac{\partial^2 \ell}{\partial \mu \partial v}\big|_{\text{MLE}} = 0$.

**Diagonal entries:**

$$
\frac{\partial^2 \ell}{\partial \mu^2}\bigg|_{\text{MLE}} = -\frac{n}{\hat{\sigma}^2} < 0
$$

$$
\frac{\partial^2 \ell}{\partial v^2}\bigg|_{\text{MLE}} = \frac{n}{2\hat{\sigma}^4} - \frac{n\hat{\sigma}^2}{\hat{\sigma}^6} = \frac{n}{2\hat{\sigma}^4} - \frac{n}{\hat{\sigma}^4} = -\frac{n}{2\hat{\sigma}^4} < 0
$$

### 5.4 The Hessian at the MLE

$$
\boldsymbol{H}\ell\big|_{\text{MLE}} = \begin{pmatrix} -\dfrac{n}{\hat{\sigma}^2} & 0 \\[8pt] 0 & -\dfrac{n}{2\hat{\sigma}^4} \end{pmatrix}
$$

Both diagonal entries are negative, so:

$$
D = \left(-\frac{n}{\hat{\sigma}^2}\right)\left(-\frac{n}{2\hat{\sigma}^4}\right) - 0 = \frac{n^2}{2\hat{\sigma}^6} > 0
$$

and $\ell_{\mu\mu} < 0$. The second-derivative test confirms: $\hat{\boldsymbol{\theta}}_{\text{MLE}} = (\bar{x}, \hat{\sigma}^2)^{\top}$ is a **local maximum** of the log-likelihood. Since $\ell$ is concave (as the Python section confirms), it is in fact the **global maximum**.

> **Fisher information.** The negative Hessian of the log-likelihood, evaluated at the MLE, is the **observed Fisher information matrix**: $\mathcal{I}(\hat{\boldsymbol{\theta}}) = -\boldsymbol{H}\ell|_{\text{MLE}}$. Its inverse gives the asymptotic covariance of $\hat{\boldsymbol{\theta}}_{\text{MLE}}$ — a measure of estimation uncertainty. For the Gaussian above: $\text{Var}(\hat{\mu}) \approx \hat{\sigma}^2 / n$ and $\text{Var}(\hat{\sigma}^2) \approx 2\hat{\sigma}^4 / n$.

### Section 1 — Symbolic Critical Points and Hessian Verification

Use SymPy to find critical points, compute Hessians, and apply the second-derivative test for the worked examples.

```python
# ============================================================
# Section 1 — Symbolic critical points and Hessian
# ============================================================
x, y = symbols('x y', real=True)

# --- Example 1.1 / 3.1: f = x³ - 3x + y² - 2y ---
f1 = x**3 - 3*x + y**2 - 2*y
grad_f1 = [diff(f1, x), diff(f1, y)]
crit_pts = sp.solve(grad_f1, [x, y])

print('f(x,y) = x³ − 3x + y² − 2y')
print('Critical points:', crit_pts)

H1 = hessian(f1, [x, y])
print('\nHessian:')
sp.pprint(H1)

for pt in crit_pts:
    H_at = H1.subs([(x, pt[0]), (y, pt[1])])
    D = H_at.det()
    fxx = H_at[0, 0]
    if D > 0 and fxx > 0:
        ctype = 'LOCAL MINIMUM'
    elif D > 0 and fxx < 0:
        ctype = 'LOCAL MAXIMUM'
    elif D < 0:
        ctype = 'SADDLE POINT'
    else:
        ctype = 'INCONCLUSIVE'
    print(f'  At {pt}: D={D}, f_xx={fxx}  →  {ctype}')

# --- Example 3.2: f = x² + xy + y² - 3x ---
print('\n' + '─'*50)
f2 = x**2 + x*y + y**2 - 3*x
grad_f2 = [diff(f2, x), diff(f2, y)]
crit_pts2 = sp.solve(grad_f2, [x, y])
H2 = hessian(f2, [x, y])
D2 = H2.det()

print('f(x,y) = x² + xy + y² − 3x')
print('Critical points:', crit_pts2)
print('Hessian (constant):')
sp.pprint(H2)
print(f'det(H) = {D2}, H[0,0] = {H2[0,0]}  →  LOCAL MINIMUM' if D2 > 0 and H2[0,0] > 0 else 'Other')
```

**What do you see?** SymPy finds the two critical points of $f_1$ and correctly classifies them: $(1,1)$ as a local minimum and $(-1,1)$ as a saddle point — matching the hand calculation. For $f_2$ the constant Hessian confirms a unique critical point that is a local minimum.

### Section 2 — Visualizing Critical Points and Saddle Points

Plot the surface and contours of $f_1$, marking the local minimum and saddle point. The saddle point is easy to recognize in 3D — it looks like the center of a saddle or a mountain pass.

```python
# ============================================================
# Section 2 — Surface and contour plot with critical points
# ============================================================
xg = np.linspace(-2.5, 2.5, 250)
yg = np.linspace(-0.5, 2.5, 250)
X_, Y_ = np.meshgrid(xg, yg)
Z_ = X_**3 - 3*X_ + Y_**2 - 2*Y_

fig = plt.figure(figsize=(14, 5))

# --- Left: 3D surface ---
ax1 = fig.add_subplot(121, projection='3d')
surf = ax1.plot_surface(X_, Y_, Z_, cmap='Blues', alpha=0.75,
                         linewidth=0, antialiased=True)
# Mark critical points
for px, py, label, col in [(1, 1, 'min', TS_GREEN), (-1, 1, 'saddle', TS_RED)]:
    pz = px**3 - 3*px + py**2 - 2*py
    ax1.scatter([px], [py], [pz], s=80, color=col, zorder=10)
    ax1.text(px, py, pz + 0.4, label, color=col, fontsize=10, fontweight='bold')

ax1.set_xlabel('x'); ax1.set_ylabel('y'); ax1.set_zlabel('f(x,y)')
ax1.set_title(r'$f(x,y) = x^3 - 3x + y^2 - 2y$')
ax1.view_init(elev=28, azim=-55)

# --- Right: contour plot ---
ax2 = fig.add_subplot(122)
levels = np.linspace(Z_.min(), Z_.max(), 28)
cf = ax2.contourf(X_, Y_, Z_, levels=levels, cmap='Blues', alpha=0.6)
cs = ax2.contour(X_, Y_, Z_, levels=levels, colors=[TS_GRAY],
                  alpha=0.4, linewidths=0.7)
plt.colorbar(cf, ax=ax2)

# Gradient field (sub-sampled)
xq = np.linspace(-2.2, 2.2, 14)
yq = np.linspace(-0.2, 2.2, 10)
Xq, Yq = np.meshgrid(xq, yq)
Gx = 3*Xq**2 - 3
Gy = 2*Yq - 2
nrm = np.sqrt(Gx**2 + Gy**2) + 1e-8
ax2.quiver(Xq, Yq, Gx/nrm, Gy/nrm,
           color=TS_BLUE, alpha=0.35, scale=22, width=0.003)

# Mark critical points
ax2.plot(1, 1, 'o', color=TS_GREEN, markersize=11, zorder=6,
         label='Local min $(1,1)$')
ax2.plot(-1, 1, 's', color=TS_RED, markersize=11, zorder=6,
         label='Saddle $(-1,1)$')
ax2.legend(fontsize=9)
ax2.set_xlabel('x'); ax2.set_ylabel('y')
ax2.set_title('Contours + gradient field')
ax2.set_aspect('equal')

plt.tight_layout()
plt.show()
```

**What do you see?**

- **3D surface**: The local minimum at $(1,1)$ is a clear bowl. The saddle point at $(-1,1)$ is a pass — a descent from one direction and an ascent from another. This is the defining geometric signature of a saddle.
- **Contour plot**: At the local minimum, gradient arrows converge inward (flowing into the minimum). At the saddle, arrows flow in along one axis and out along the other — the gradient is zero at the point itself, but the surrounding flow pattern reveals the saddle structure.

**Key observation**: a gradient descent algorithm started anywhere to the right of $x \approx 0$ will descend to the local minimum at $(1, 1)$. Started to the left, it descends away from the saddle. Saddle points are problematic for optimization — gradient descent can slow dramatically near them because $\|\nabla f\|$ is small, even though the point is not a minimum.

### Section 3 — Statistical Bridge: Log-Likelihood Surface and Hessian at the MLE

Visualize the Gaussian log-likelihood as a surface in $(\mu, \sigma^2)$, confirm it is concave (every cross-section curves downward), locate the MLE, and verify the Hessian at that point is negative definite.

```python
# ============================================================
# Section 3 — Gaussian log-likelihood: surface, MLE, Hessian
# ============================================================
rng = np.random.default_rng(42)
n_obs = 30
true_mu, true_sigma = 2.5, 1.2
data = rng.normal(true_mu, true_sigma, size=n_obs)
x_bar    = data.mean()
s2_hat   = data.var()       # MLE for σ²

def log_lik(mu, v, x):
    """Gaussian log-likelihood in (mu, v=sigma^2). v must be > 0."""
    n = len(x)
    return (-n/2)*np.log(2*np.pi*v) - np.sum((x - mu)**2)/(2*v)

# Grid over (mu, sigma^2)
mu_g  = np.linspace(x_bar - 1.5, x_bar + 1.5, 200)
v_g   = np.linspace(s2_hat * 0.3, s2_hat * 2.5, 200)
MU, V = np.meshgrid(mu_g, v_g)
LL    = np.vectorize(lambda m, v: log_lik(m, v, data))(MU, V)

fig = plt.figure(figsize=(14, 5))

# Left: 3D surface
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(MU, V, LL, cmap='Blues', alpha=0.75,
                 linewidth=0, antialiased=True)
ll_mle = log_lik(x_bar, s2_hat, data)
ax1.scatter([x_bar], [s2_hat], [ll_mle],
            s=80, color=TS_AMBER, zorder=10)
ax1.text(x_bar, s2_hat, ll_mle + 1.0, 'MLE', color=TS_AMBER,
         fontsize=10, fontweight='bold')
ax1.set_xlabel(r'$\mu$'); ax1.set_ylabel(r'$\sigma^2$')
ax1.set_zlabel(r'$\ell(\mu, \sigma^2)$')
ax1.set_title('Gaussian log-likelihood surface')
ax1.view_init(elev=28, azim=-60)

# Right: contour + gradient flow
ax2 = fig.add_subplot(122)
cf2 = ax2.contourf(MU, V, LL, levels=25, cmap='Blues', alpha=0.65)
ax2.contour(MU, V, LL, levels=25, colors=[TS_GRAY], alpha=0.4, linewidths=0.7)
plt.colorbar(cf2, ax=ax2, label=r'$\ell(\mu, \sigma^2)$')
ax2.plot(x_bar, s2_hat, '*', color=TS_AMBER, markersize=14, zorder=6,
         label=fr'MLE: $\bar{{x}}={x_bar:.3f}$, $\hat{{\sigma}}^2={s2_hat:.3f}$')
ax2.axvline(true_mu, color=TS_GREEN, lw=1.0, ls=':',
            label=fr'True $\mu={true_mu}$')
ax2.axhline(true_sigma**2, color=TS_GREEN, lw=1.0, ls=':')
ax2.set_xlabel(r'$\mu$'); ax2.set_ylabel(r'$\sigma^2$')
ax2.set_title('Log-likelihood contours — MLE at the peak')
ax2.legend(fontsize=9)

plt.tight_layout()
plt.show()

# --- Numerical Hessian at the MLE ---
eps = 1e-5
def ll_scalar(params):
    return log_lik(params[0], params[1], data)

theta_mle = np.array([x_bar, s2_hat])
H_num = np.zeros((2, 2))
for i in range(2):
    for j in range(2):
        ei = np.zeros(2); ei[i] = eps
        ej = np.zeros(2); ej[j] = eps
        H_num[i,j] = (ll_scalar(theta_mle + ei + ej)
                     - ll_scalar(theta_mle + ei - ej)
                     - ll_scalar(theta_mle - ei + ej)
                     + ll_scalar(theta_mle - ei - ej)) / (4 * eps**2)

eigenvalues = np.linalg.eigvalsh(H_num)
print('Numerical Hessian at MLE:')
print(np.round(H_num, 4))
print(f'\nEigenvalues: {np.round(eigenvalues, 4)}')
print(f'Both negative: {np.all(eigenvalues < 0)}  →  MLE is a MAXIMUM ✓')
print(f'\nAnalytical Hessian diagonal:')
print(f'  ∂²ℓ/∂μ²    = {-n_obs/s2_hat:.4f}')
print(f'  ∂²ℓ/∂(σ²)² = {-n_obs/(2*s2_hat**2):.4f}')
print(f'\nObserved Fisher information (= −H):')
print(np.round(-H_num, 4))
print(f'Asymptotic Var(μ̂) ≈ σ²/n = {s2_hat/n_obs:.4f}')
print(f'Asymptotic Var(σ̂²) ≈ 2σ⁴/n = {2*s2_hat**2/n_obs:.4f}')
```

**What do you see?**

- The log-likelihood surface is a concave cap — it rises to a unique peak at the MLE and falls away in all directions. This is the visual signature of a negative definite Hessian.
- The contour plot shows elliptical level curves centered near the MLE. The ellipse orientation reflects the off-diagonal structure of the Hessian — here nearly axis-aligned because the mixed partial vanishes at $\hat{\mu} = \bar{x}$.
- The numerical Hessian confirms both eigenvalues are negative, and the diagonal entries match the analytical formulas from Section 5.3.
- The observed Fisher information (negative Hessian) diagonal gives the asymptotic variances: $\hat{\sigma}^2/n$ for $\hat{\mu}$ and $2\hat{\sigma}^4/n$ for $\hat{\sigma}^2$.

## Summary

| Concept | Statement |
|---|---|
| Critical point | $\nabla f(\boldsymbol{x}^*) = \boldsymbol{0}$ — necessary for interior extremum |
| Hessian definition | $\boldsymbol{H}f \in \mathbb{R}^{n \times n}$, entry $(i,j) = \partial^2 f / \partial x_i \partial x_j$ |
| Hessian symmetry | $\boldsymbol{H}f = (\boldsymbol{H}f)^{\top}$ (Schwarz's theorem) |
| Hessian as Jacobian | $\boldsymbol{H}f = \boldsymbol{J}_{\nabla f}$ |
| Second-order Taylor | $f(\boldsymbol{x}^* + \boldsymbol{h}) \approx f(\boldsymbol{x}^*) + \frac{1}{2}\boldsymbol{h}^{\top}\boldsymbol{H}f\,\boldsymbol{h}$ at a critical point |
| $2 \times 2$ test: min | $D > 0$, $f_{xx} > 0$ |
| $2 \times 2$ test: max | $D > 0$, $f_{xx} < 0$ |
| $2 \times 2$ test: saddle | $D < 0$ |
| General test | Sign of all eigenvalues of $\boldsymbol{H}f(\boldsymbol{x}^*)$ |
| MLE Hessian | Negative definite at $\hat{\boldsymbol{\theta}}_{\text{MLE}}$ confirms global maximum |
| Fisher information | $\mathcal{I}(\hat{\boldsymbol{\theta}}) = -\boldsymbol{H}\ell|_{\text{MLE}}$; inverse gives asymptotic covariance |
