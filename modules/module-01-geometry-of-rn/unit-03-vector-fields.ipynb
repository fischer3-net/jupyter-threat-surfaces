## 1. Vector Fields

### 1.1 Definition

In Unit 01-01 we defined a **scalar field** as a function $f: \mathbb{R}^n \to \mathbb{R}$ — one that assigns a single number to each point in space. A **vector field** assigns a *vector* to each point instead.

**Definition.** A **vector field** on $\mathbb{R}^n$ is a function
$$
\boldsymbol{F}: D \subseteq \mathbb{R}^n \to \mathbb{R}^n
$$
that assigns to each point $\boldsymbol{x} \in D$ a vector $\boldsymbol{F}(\boldsymbol{x}) \in \mathbb{R}^n$.

For $n = 2$ we write $\boldsymbol{F}(x,y) = (P(x,y),\, Q(x,y))^{\top}$, where $P$ and $Q$ are scalar fields giving the $x$- and $y$-components of the output vector at each point. For $n = 3$, $\boldsymbol{F}(x,y,z) = (P, Q, R)^{\top}$.

We visualise a vector field by drawing an arrow at each point $(x, y)$, pointing in the direction $\boldsymbol{F}(x,y)$ with length proportional to $\|\boldsymbol{F}(x,y)\|$. The result is a **flow diagram** — the arrows show how something would move if it followed the field.

### 1.2 Three Canonical Vector Fields

**Constant field.** $\boldsymbol{F}(x,y) = (1, 0)^{\top}$ — all arrows point to the right with equal length. This is the simplest possible field: direction and magnitude are the same everywhere. It models uniform flow (e.g., a steady wind).

**Radial (source) field.** $\boldsymbol{F}(x,y) = (x, y)^{\top}$ — at each point, the arrow points directly away from the origin with magnitude $\|(x,y)\| = \sqrt{x^2 + y^2}$. This field points outward in all directions from the origin, which acts as a *source*. Near the origin the arrows are short (the field is weak); far from the origin they are long (the field is strong).

**Circulation field.** $\boldsymbol{F}(x,y) = (-y, x)^{\top}$ — at each point, the arrow is perpendicular to the position vector $(x, y)^{\top}$ and rotated 90° counterclockwise. All arrows have magnitude $\sqrt{x^2 + y^2}$. The field produces counterclockwise rotation around the origin.

### 1.3 The Gradient as a Vector Field

The most important vector field in this course is the one built from a scalar field $f$ by the operation of taking the gradient. We write it $\nabla f$ (read "del $f$" or "grad $f$"). Its formal definition — as a vector of partial derivatives — arrives in Module 03. For now, the geometric definition:

> At each point $\boldsymbol{x}_0$, the vector $\nabla f(\boldsymbol{x}_0)$ points in the direction in which $f$ increases most rapidly, and its magnitude equals the rate of that increase.

This makes $\nabla f: \mathbb{R}^n \to \mathbb{R}^n$ a vector field — it takes a point in the domain of $f$ and returns a vector in $\mathbb{R}^n$.

Gradient fields have a special structure that most vector fields lack: they are *conservative* (also called *irrotational*). In a conservative field, the work done in moving from point $A$ to point $B$ depends only on $A$ and $B$, not on the path taken. This property has deep implications in both physics and optimization.

## 2. The Gradient: Geometric Picture

### 2.1 Perpendicularity to Level Curves

The central geometric fact about the gradient:

> **Theorem (informal).** At any point $\boldsymbol{x}_0$ where $\nabla f(\boldsymbol{x}_0) \neq \boldsymbol{0}$, the gradient vector $\nabla f(\boldsymbol{x}_0)$ is perpendicular to the level curve (or level set) of $f$ passing through $\boldsymbol{x}_0$.

We will prove this rigorously in Module 03. For now, here is the intuition:

Moving *along* a level curve keeps $f$ constant — the rate of change of $f$ in the tangent direction is zero. The direction in which $f$ increases most rapidly must therefore be perpendicular to the tangent of the level curve. Since the tangent direction and the direction of steepest ascent are orthogonal, and since a perpendicular to the tangent is the normal to the curve, the gradient points along the normal — i.e., *perpendicular to the level curve*.

### 2.2 The Gradient Points Uphill

"Perpendicular to the level curve" does not uniquely specify a direction — there are two perpendicular directions at each point (inward and outward). The gradient convention resolves this: $\nabla f$ points in the direction of *increasing* $f$, not decreasing $f$.

Concretely:
- $\nabla f(\boldsymbol{x}_0)$ points from the level curve $\{f = c\}$ toward the level curve $\{f = c + \epsilon\}$ for small $\epsilon > 0$.
- The magnitude $\|\nabla f(\boldsymbol{x}_0)\|$ equals the rate of increase in that direction — large where the level curves are close together (steep surface), small where they are far apart (gentle surface).
- $-\nabla f(\boldsymbol{x}_0)$ points in the direction of steepest *descent* — toward lower level curves.

**Worked example — paraboloid.** For $f(x,y) = x^2 + y^2$, the level curves are concentric circles. The direction perpendicular to a circle at any point is the radial direction — pointing directly away from the centre. And indeed, the gradient of $f$ is $(2x, 2y)^{\top}$, which points radially outward. At $(1, 1)$, the gradient is $(2, 2)^{\top}$ — outward and uphill. At the origin, the gradient is $(0, 0)^{\top}$ — the minimum, where there is no preferred direction.

**Worked example — Gaussian bell.** For $f(x,y) = e^{-(x^2+y^2)/2}$, the level curves are again concentric circles, but now the maximum is at the origin and $f$ decreases outward. The gradient points *inward* toward the peak — radially toward the origin. At $(1, 0)$, the gradient points in the $(-1, 0)$ direction, toward the centre.

### 2.3 Gradient Descent: Moving Opposite the Gradient

Most optimization in machine learning involves *minimising* a loss function. Since $\nabla f$ points toward increasing $f$, moving in the direction $-\nabla f$ decreases $f$ as rapidly as possible. This is the update rule for **gradient descent**:

$$
\boldsymbol{\theta}_{t+1} = \boldsymbol{\theta}_t - \eta\,\nabla_{\boldsymbol{\theta}}\mathcal{L}(\boldsymbol{\theta}_t),
$$

where $\eta > 0$ is the **learning rate** (step size) and $\mathcal{L}$ is the loss.

Reading this geometrically: start at $\boldsymbol{\theta}_t$ on the loss surface. The gradient $\nabla \mathcal{L}(\boldsymbol{\theta}_t)$ is the vector perpendicular to the level curve through $\boldsymbol{\theta}_t$, pointing uphill. Take a step of size $\eta$ in the opposite direction — downhill. Repeat.

The learning rate $\eta$ controls how far we step between iterations. If $\eta$ is too large, we overshoot the minimum and may diverge. If $\eta$ is too small, convergence is slow. The curvature of the loss surface — quantified by the **Hessian** in Module 04 — determines the optimal step size.

### 2.4 Where the Gradient Is Zero

At a point where $\nabla f(\boldsymbol{x}_0) = \boldsymbol{0}$, there is no preferred direction of ascent — the surface is locally flat in all directions. These are **critical points**: candidates for local maxima, local minima, or saddle points. The classification of critical points is the subject of Module 04.

Gradient descent halts (in the limit) at a critical point. For a convex loss function, the unique critical point is the global minimum. For non-convex losses — like those of neural networks — gradient descent may halt at a local minimum or a saddle point.

## 3. Reading a Vector Field Plot

A **quiver plot** (named after the container for arrows) draws a vector field by placing one arrow per grid point. The arrow at $(x_0, y_0)$ has direction $\boldsymbol{F}(x_0, y_0)$ and length proportional to $\|\boldsymbol{F}(x_0, y_0)\|$. In practice, arrows are often normalised to a common length for clarity, with colour or a separate scalar plot encoding magnitude.

When reading a quiver plot:

| Feature | What it signals |
|---|---|
| Arrows converging to a point | *Sink* — the field is pulling toward that point; local minimum of the potential |
| Arrows diverging from a point | *Source* — the field is pushing away; local maximum of the potential |
| Arrows circling a point | *Vortex* — the field is rotating; NOT a gradient field (gradient fields have no rotation) |
| Arrows all parallel | Uniform field — constant gradient; the potential is a linear function |
| Long arrows in one region, short in another | Steep vs. gentle regions of the underlying scalar field |
| Arrows perpendicular to a visible contour | Gradient field — the arrow is the gradient of the scalar field whose level curves are shown |

The last row is the key check: if you overlay a gradient field on the contour map of its scalar field, every arrow should be exactly perpendicular to the nearest contour. This is a built-in consistency check for any gradient computation.

## 4. Statistical Bridge — The Score Function as a Vector Field [MLE]

### 4.1 The Score Function

The **score function** is the gradient of the log-likelihood with respect to the parameters:

$$
s(\boldsymbol{\theta}) = \nabla_{\boldsymbol{\theta}}\,\ell(\boldsymbol{\theta}).
$$

For a model with parameter vector $\boldsymbol{\theta} \in \mathbb{R}^p$, the log-likelihood $\ell(\boldsymbol{\theta})$ is a scalar field on $\mathbb{R}^p$ and the score $s(\boldsymbol{\theta})$ is a vector field on $\mathbb{R}^p$ — it assigns a $p$-dimensional vector to each parameter value.

The score encodes: *if I am currently at parameter value $\boldsymbol{\theta}$, in which direction should I move to increase the log-likelihood most rapidly?* The MLE is the point where the score is zero — the level curves of $\ell$ are flat, no direction is preferred, and we are at the peak.

### 4.2 Gaussian Score in Two Parameters

For the Gaussian log-likelihood $\ell(\mu, \sigma^2)$ from Unit 01-01, the score has two components:

$$
s_\mu(\mu, \sigma^2) = \frac{\partial \ell}{\partial \mu} = \frac{1}{\sigma^2}\sum_{i=1}^{n}(x_i - \mu) = \frac{n(\bar{x} - \mu)}{\sigma^2}
$$

$$
s_{\sigma^2}(\mu, \sigma^2) = \frac{\partial \ell}{\partial \sigma^2} = -\frac{n}{2\sigma^2} + \frac{1}{2(\sigma^2)^2}\sum_{i=1}^{n}(x_i - \mu)^2
$$

The score vector $s(\mu, \sigma^2) = (s_\mu,\, s_{\sigma^2})^{\top}$ is a vector field on the $(\mu, \sigma^2)$ parameter plane. Several observations:

- $s_\mu > 0$ when $\mu < \bar{x}$ (the log-likelihood increases as $\mu$ moves toward $\bar{x}$); $s_\mu < 0$ when $\mu > \bar{x}$.
- $s_{\sigma^2} > 0$ when $\sigma^2 < \hat{\sigma}^2_{\text{MLE}}$; $s_{\sigma^2} < 0$ when $\sigma^2 > \hat{\sigma}^2_{\text{MLE}}$.
- At the MLE $(\bar{x}, \hat{\sigma}^2)$, both components are zero — the score vector vanishes. This is the defining property of the MLE.
- The score vector at any point is perpendicular to the likelihood contour through that point — exactly as the gradient of any scalar field is perpendicular to its level curves.

### 4.3 The Fisher Information as Variance of the Score

The **Fisher information matrix** is defined as the covariance of the score:

$$
\boldsymbol{I}(\boldsymbol{\theta}) = \mathbb{E}\left[s(\boldsymbol{\theta})\,s(\boldsymbol{\theta})^{\top}\right] = -\mathbb{E}\left[\nabla^2_{\boldsymbol{\theta}}\,\ell(\boldsymbol{\theta})\right],
$$

where $\nabla^2$ denotes the Hessian matrix of second derivatives (Module 04). Large Fisher information means the score changes rapidly near the MLE — the likelihood surface is sharply curved — and the MLE is estimated precisely. Small Fisher information means a flat surface and an imprecise estimate. The Fisher information matrix determines the shape and size of the confidence ellipses from Unit 01-02.

> **Forward link.** In Module 02 we will compute partial derivatives and make the formulas above precise. In Module 03 we will define the gradient formally and prove the perpendicularity property. In Module 04 the Hessian of the log-likelihood connects the curvature of the score field to the precision of the MLE.

## Section 1 — The Three Canonical Vector Fields

We plot the constant, radial, and circulation fields from Section 1.2. Each plot shows normalised arrows (constant length) so the *direction* is clear, with colour encoding the magnitude $\|\boldsymbol{F}\|$ at each point.

```python
# --- Grid for quiver plots ---
lim   = 2.2
steps = 16
x_q   = np.linspace(-lim, lim, steps)
y_q   = np.linspace(-lim, lim, steps)
X_q, Y_q = np.meshgrid(x_q, y_q)

# ---- Three vector fields ----
# 1. Constant: F = (1, 0)
U_const = np.ones_like(X_q)
V_const = np.zeros_like(Y_q)

# 2. Radial: F = (x, y)
U_radial = X_q.copy()
V_radial = Y_q.copy()

# 3. Circulation: F = (-y, x)
U_circ = -Y_q.copy()
V_circ =  X_q.copy()

# Magnitude for colouring
M_const  = np.ones_like(X_q)
M_radial = np.sqrt(U_radial**2 + V_radial**2)
M_circ   = np.sqrt(U_circ**2  + V_circ**2)

fields = [
    (U_const,  V_const,  M_const,  r'Constant field  $\boldsymbol{F}=(1,0)^\top$',
     'Uniform magnitude — same everywhere'),
    (U_radial, V_radial, M_radial, r'Radial (source) field  $\boldsymbol{F}=(x,y)^\top$',
     'Magnitude grows with distance from origin'),
    (U_circ,   V_circ,   M_circ,   r'Circulation field  $\boldsymbol{F}=(-y,x)^\top$',
     'Counterclockwise rotation — NOT a gradient field'),
]

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for ax, (U, V, M, title, note) in zip(axes, fields):
    # Normalise arrow lengths for direction clarity; colour encodes magnitude
    norm = np.sqrt(U**2 + V**2)
    norm[norm == 0] = 1  # avoid division by zero
    q = ax.quiver(X_q, Y_q, U / norm, V / norm, M,
                  cmap='Blues', scale=18, width=0.004,
                  clim=(M.min(), M.max()))
    plt.colorbar(q, ax=ax, label=r'$\|\boldsymbol{F}\|$')
    ax.set_title(title, fontweight='bold', color=TS_GRAY)
    ax.set_xlabel('$x$'); ax.set_ylabel('$y$')
    ax.set_aspect('equal')
    ax.axhline(0, color=TS_GRAY, lw=0.5, ls='--')
    ax.axvline(0, color=TS_GRAY, lw=0.5, ls='--')
    ax.text(0.02, 0.02, note, transform=ax.transAxes,
            fontsize=8, color=TS_BLUE, style='italic')

plt.suptitle('Module 01, Unit 03 — Three Canonical Vector Fields',
             fontsize=14, fontweight='bold', color=TS_GRAY)
plt.tight_layout()
plt.savefig('module-01-unit-03-canonical-fields.png', dpi=150, bbox_inches='tight')
plt.show()
print('Figure saved.')
```

### What Do You See?

1. **Constant field**: All arrows point right with the same length. This field is the gradient of $f(x,y) = x$ — a tilted plane. What are its level curves, and does the arrow direction match your expectation?

2. **Radial field**: The arrows point away from the origin in all directions. The colour darkens with distance — magnitude grows. This is the gradient of $f(x,y) = \frac{1}{2}(x^2 + y^2)$ — the paraboloid, rescaled. Does the arrow direction match the gradient-perpendicular-to-level-curves rule for the paraboloid?

3. **Circulation field**: The arrows form a counterclockwise whirlpool. There is no scalar field whose gradient produces this pattern — this is a fundamentally non-gradient (non-conservative) field. Notice that the arrows are *tangent* to circles centred at the origin, not perpendicular to them. Gradient fields always produce arrows perpendicular to the level curves; circulation fields produce arrows parallel to them. This distinction — gradient vs. non-gradient — becomes precise when we study curl in Module 07.

## Section 2 — Gradient Fields Overlaid on Level Curves

We now overlay the gradient vector field of a scalar field on top of its contour map. This is the core visual verification: every gradient arrow should be **perpendicular to the nearest contour line**.

We use numerical gradients (via NumPy's `np.gradient`) since we have not yet formally defined partial derivatives. The result is exactly what Module 03 will produce analytically.

```python
# --- Dense grid for contours ---
lim_c  = 2.5
n_fine = 300
x_c    = np.linspace(-lim_c, lim_c, n_fine)
y_c    = np.linspace(-lim_c, lim_c, n_fine)
X_c, Y_c = np.meshgrid(x_c, y_c)

# --- Coarse grid for gradient arrows ---
n_coarse = 18
x_g  = np.linspace(-lim_c * 0.9, lim_c * 0.9, n_coarse)
y_g  = np.linspace(-lim_c * 0.9, lim_c * 0.9, n_coarse)
X_g, Y_g = np.meshgrid(x_g, y_g)

# ---- Function 1: Paraboloid  f = x^2 + y^2 ----
Z_bowl = X_c**2 + Y_c**2
# Analytical gradient: (2x, 2y)
Ug_bowl = 2 * X_g
Vg_bowl = 2 * Y_g

# ---- Function 2: Gaussian bell  f = exp(-(x^2+y^2)/2) ----
Z_bell = np.exp(-(X_c**2 + Y_c**2) / 2)
# Analytical gradient: -x*exp(-(x^2+y^2)/2), -y*exp(-(x^2+y^2)/2)
bell_g   = np.exp(-(X_g**2 + Y_g**2) / 2)
Ug_bell  = -X_g * bell_g
Vg_bell  = -Y_g * bell_g

fig, axes = plt.subplots(1, 2, figsize=(15, 6))

for ax, (Z, Ug, Vg, title, levels, note, marker_pt) in zip(axes, [
    (Z_bowl, Ug_bowl, Vg_bowl,
     r'Paraboloid  $f = x^2 + y^2$',
     [0.5, 1, 2, 3, 4, 5],
     'Gradient (blue arrows) points outward — away from minimum',
     (0, 0)),
    (Z_bell, Ug_bell, Vg_bell,
     r'Gaussian bell  $f = e^{-(x^2+y^2)/2}$',
     [0.15, 0.30, 0.50, 0.70, 0.85, 0.95],
     'Gradient (blue arrows) points inward — toward maximum',
     (0, 0)),
]):
    # Contour map
    cf = ax.contourf(X_c, Y_c, Z, levels=20, cmap='Greys', alpha=0.35)
    cs = ax.contour(X_c, Y_c, Z, levels=levels, colors=[TS_GRAY], linewidths=1.4)
    ax.clabel(cs, fontsize=8, colors=TS_GRAY)

    # Gradient arrows — normalised direction, colour = magnitude
    mag = np.sqrt(Ug**2 + Vg**2)
    mag[mag == 0] = 1
    ax.quiver(X_g, Y_g, Ug / mag, Vg / mag, mag,
              cmap='Blues', scale=22, width=0.004,
              clim=(0, mag.max()), zorder=3)

    # Mark the critical point
    ax.scatter([marker_pt[0]], [marker_pt[1]], color=TS_AMBER,
               s=100, zorder=5)

    ax.set_title(title, fontweight='bold', color=TS_GRAY)
    ax.set_xlabel('$x$'); ax.set_ylabel('$y$')
    ax.set_aspect('equal')
    ax.text(0.02, 0.02, note, transform=ax.transAxes,
            fontsize=9, color=TS_BLUE, style='italic')

plt.suptitle('Module 01, Unit 03 — Gradient Field Overlaid on Level Curves\n'
             'Every arrow is perpendicular to the nearest contour',
             fontsize=13, fontweight='bold', color=TS_GRAY)
plt.tight_layout()
plt.savefig('module-01-unit-03-gradient-overlay.png', dpi=150, bbox_inches='tight')
plt.show()
print('Figure saved.')
```

### What Do You See?

1. **Perpendicularity check**: Pick any contour line on either panel and trace it with your eye. Now look at the gradient arrows nearby. Are they pointing along the contour, or across it? They should be crossing the contour at a right angle. This is the geometric definition of the gradient made visible.

2. **Paraboloid**: The arrows point outward (away from the origin) and grow longer with distance. The orange dot marks the minimum, where the gradient is zero — no arrow is drawn there. A gradient descent update at any point would move *opposite* these arrows — inward toward the minimum.

3. **Gaussian bell**: The arrows point inward (toward the peak at the origin) and are longer where the surface is steep (midrange radius) than where it is flat (near the peak or far from it). A gradient ascent step — following the arrows — leads to the maximum. A gradient *descent* step on the *negative* Gaussian would lead away from the minimum at the origin.

4. **Comparison**: The paraboloid arrows grow without bound as you move away from the origin; the Gaussian bell arrows shrink toward zero both near the peak and far from it (the surface is flat in both regions). This has direct implications for optimisation: gradient descent on a quadratic (paraboloid) has predictable step behaviour, while gradient descent in the tails of a Gaussian-shaped surface has very small steps — the gradient is nearly zero far from the peak.

## Section 3 — Statistical Bridge: The Score Function as a Vector Field [MLE]

We now visualise the Gaussian score function
$$
s(\mu, \sigma^2) = \left(\frac{n(\bar{x} - \mu)}{\sigma^2},\;\; -\frac{n}{2\sigma^2} + \frac{\sum(x_i - \mu)^2}{2(\sigma^2)^2}\right)^{\!\top}
$$
as a vector field over the $(\mu, \sigma^2)$ parameter plane, overlaid on the log-likelihood contours from Unit 01-02. Every score arrow should point perpendicular to the nearest likelihood contour — and they should all converge toward the MLE.

```python
# --- Reproduce data from Units 01-01 and 01-02 ---
rng = np.random.default_rng(seed=42)
n_data  = 30
mu_true = 2.0
s2_true = 1.0
data    = rng.normal(loc=mu_true, scale=np.sqrt(s2_true), size=n_data)
x_bar   = data.mean()
s2_hat  = data.var()
ss      = np.sum((data - x_bar)**2)  # sum of squared deviations from MLE mu

# --- Grid for log-likelihood contours (dense) ---
mu_lo, mu_hi = x_bar - 1.1, x_bar + 1.1
s2_lo, s2_hi = max(0.1, s2_hat - 0.9), s2_hat + 1.3

mu_fine = np.linspace(mu_lo, mu_hi, 350)
s2_fine = np.linspace(s2_lo, s2_hi, 350)
MU_f, S2_f = np.meshgrid(mu_fine, s2_fine)

resid_sq = np.sum((data[:, None, None] - MU_f[None, :, :])**2, axis=0)
LOGLIK   = -n_data / 2 * np.log(2 * np.pi * S2_f) - resid_sq / (2 * S2_f)
ell_max  = LOGLIK.max()

# --- Grid for score vectors (coarse) ---
n_arr  = 16
mu_arr = np.linspace(mu_lo + 0.05, mu_hi - 0.05, n_arr)
s2_arr = np.linspace(s2_lo + 0.05, s2_hi - 0.05, n_arr)
MU_a, S2_a = np.meshgrid(mu_arr, s2_arr)

# --- Score components (exact partial derivatives of ell) ---
# s_mu    = n * (x_bar - mu) / sigma^2
# s_sigma2 = -n/(2*sigma^2) + sum((x_i-mu)^2) / (2*sigma^4)
# For the coarse grid, compute sum((x_i - mu)^2) via broadcasting
ss_arr   = np.sum((data[:, None, None] - MU_a[None, :, :])**2, axis=0)
U_score  = n_data * (x_bar - MU_a) / S2_a
V_score  = -n_data / (2 * S2_a) + ss_arr / (2 * S2_a**2)

mag_score = np.sqrt(U_score**2 + V_score**2)
mag_score[mag_score == 0] = 1

print(f'MLE: mu_hat = {x_bar:.4f},  sigma^2_hat = {s2_hat:.4f}')
print(f'Score at MLE: s_mu = {n_data * (x_bar - x_bar) / s2_hat:.4f},  '
      f's_s2 = {-n_data/(2*s2_hat) + ss/(2*s2_hat**2):.6f}  (both should be ~0)')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# ---- Left: score field overlaid on log-likelihood contours ----
ax = axes[0]

# Log-likelihood contour map
drop_range = 20
cf = ax.contourf(MU_f, S2_f, LOGLIK,
                 levels=np.linspace(ell_max - drop_range, ell_max, 25),
                 cmap='Greens', alpha=0.6)
plt.colorbar(cf, ax=ax, label='$\\ell(\\mu, \\sigma^2)$')
cs = ax.contour(MU_f, S2_f, LOGLIK,
                levels=np.linspace(ell_max - drop_range, ell_max, 10),
                colors=[TS_GRAY], linewidths=0.8, alpha=0.7)

# Score vector field — normalised direction, colour = magnitude
ax.quiver(MU_a, S2_a,
          U_score / mag_score, V_score / mag_score,
          mag_score, cmap='Blues',
          scale=22, width=0.004, zorder=4)

# MLE and true parameters
ax.scatter([x_bar],  [s2_hat],  color=TS_AMBER, s=120, marker='*', zorder=6,
           label=f'MLE $\\hat{{\\mu}}={x_bar:.2f},\\,\\hat{{\\sigma}}^2={s2_hat:.2f}$')
ax.scatter([mu_true],[s2_true], color=TS_BLUE,  s=70,  zorder=6,
           label=f'True $(\\mu^*={mu_true},\\,\\sigma^{{2*}}={s2_true})$')

ax.set_xlabel('$\\mu$'); ax.set_ylabel('$\\sigma^2$')
ax.set_title('Score field (blue arrows) on log-likelihood contours\n'
             'Arrows perpendicular to contours, converging to MLE',
             fontweight='bold', color=TS_GRAY)
ax.legend(fontsize=9)

# ---- Right: score components separately ----
ax2 = axes[1]

# Plot s_mu as a function of mu (with sigma^2 fixed at MLE)
mu_line  = np.linspace(mu_lo, mu_hi, 300)
s_mu_line = n_data * (x_bar - mu_line) / s2_hat

# Plot s_sigma2 as a function of sigma^2 (with mu fixed at MLE)
s2_line    = np.linspace(s2_lo, s2_hi, 300)
s_s2_line  = -n_data / (2 * s2_line) + ss / (2 * s2_line**2)

ax2_twin = ax2.twinx()

ax2.plot(mu_line, s_mu_line, color=TS_BLUE, lw=2,
         label=r'$s_\mu(\mu)$  (left axis, $\sigma^2$ fixed at MLE)')
ax2.axhline(0, color=TS_BLUE, lw=0.8, ls='--', alpha=0.5)
ax2.axvline(x_bar, color=TS_AMBER, lw=1.5, ls=':', label=f'$\\bar{{x}} = {x_bar:.2f}$')
ax2.set_xlabel('$\\mu$ (left)  /  $\\sigma^2$ (right axis for $s_{{\\sigma^2}}$)')
ax2.set_ylabel(r'Score component $s_\mu$', color=TS_BLUE)
ax2.tick_params(axis='y', labelcolor=TS_BLUE)

ax2_twin.plot(s2_line, s_s2_line, color=TS_GREEN, lw=2,
              label=r'$s_{\sigma^2}(\sigma^2)$  (right axis, $\mu$ fixed at MLE)')
ax2_twin.axhline(0, color=TS_GREEN, lw=0.8, ls='--', alpha=0.5)
ax2_twin.axvline(s2_hat, color=TS_AMBER, lw=1.5, ls=':')
ax2_twin.set_ylabel(r'Score component $s_{\sigma^2}$', color=TS_GREEN)
ax2_twin.tick_params(axis='y', labelcolor=TS_GREEN)

# Combined legend
lines1, labels1 = ax2.get_legend_handles_labels()
lines2, labels2 = ax2_twin.get_legend_handles_labels()
ax2.legend(lines1 + lines2, labels1 + labels2, fontsize=8, loc='upper right')

ax2.set_title('Score components cross zero at the MLE\n'
              r'$s_\mu = 0$ at $\mu = \bar{x}$;  $s_{\sigma^2} = 0$ at $\sigma^2 = \hat{\sigma}^2$',
              fontweight='bold', color=TS_GRAY)

plt.suptitle('Module 01, Unit 03 — Gaussian Score Function as a Vector Field [MLE]',
             fontsize=13, fontweight='bold', color=TS_GRAY)
plt.tight_layout()
plt.savefig('module-01-unit-03-score-field.png', dpi=150, bbox_inches='tight')
plt.show()
print('Figure saved.')
```

### What Do You See?

1. **Left panel — perpendicularity**: Trace a likelihood contour (green curve) with your eye and compare the score arrows (blue) nearby. They should be crossing the contour at right angles, pointing inward toward the MLE star. This is the same perpendicularity you verified for the paraboloid and Gaussian bell in Section 2, now on the parameter space of a statistical model.

2. **Left panel — convergence**: The score arrows from all parts of the parameter space converge toward the MLE (orange star). Running gradient ascent on the log-likelihood — following the score — from any starting point would trace a path to the MLE, crossing level curves all the way.

3. **Right panel — zero-crossing**: $s_\mu$ crosses zero exactly at $\mu = \bar{x}$ (the MLE for $\mu$, marked by the dotted vertical line). $s_{\sigma^2}$ crosses zero at $\sigma^2 = \hat{\sigma}^2$. This is the defining property of the MLE: the score function is zero at the maximum. For $n = 30$ observations, is the zero of $s_\mu$ close to the true $\mu^* = 2.0$? What would happen as $n$ increases?

## Summary

| Concept | Definition / Key result |
|---|---|
| Vector field | $\boldsymbol{F}: \mathbb{R}^n \to \mathbb{R}^n$ — assigns a vector to every point in space |
| Gradient field | The vector field $\nabla f$ built from a scalar field $f$ — a conservative field |
| Gradient direction | Perpendicular to the level curve/set of $f$, pointing toward increasing $f$ |
| Gradient magnitude | $\|\nabla f\|$ = rate of steepest ascent; large where level curves are close |
| Critical point | $\nabla f = \boldsymbol{0}$ — no preferred direction; candidate for extremum or saddle |
| Gradient descent | $\boldsymbol{\theta}_{t+1} = \boldsymbol{\theta}_t - \eta\,\nabla\mathcal{L}(\boldsymbol{\theta}_t)$ — moves opposite the gradient |
| Score function | $s(\boldsymbol{\theta}) = \nabla_{\boldsymbol{\theta}}\,\ell(\boldsymbol{\theta})$ — the gradient of the log-likelihood; a vector field on parameter space |
| MLE condition | $s(\hat{\boldsymbol{\theta}}_{\text{MLE}}) = \boldsymbol{0}$ — score is zero at the maximum |
| Fisher information | $\boldsymbol{I}(\boldsymbol{\theta}) = \mathbb{E}[s\,s^{\top}]$ — measures how rapidly the score changes; encodes precision of the MLE |
| Non-gradient field | Circulation field — arrows tangent to level curves, no scalar potential; identified by non-zero curl (Module 07) |
