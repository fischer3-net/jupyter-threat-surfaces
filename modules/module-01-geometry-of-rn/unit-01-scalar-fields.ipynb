## 1. Functions of Multiple Variables

### 1.1 The Basic Definition

In single-variable calculus, a function $f: \mathbb{R} \to \mathbb{R}$ takes one real number as input and returns one real number as output. Multivariable calculus generalises the *input* side: we allow the domain to be $\mathbb{R}^n$ for any $n \geq 2$.

**Definition.** A **function of $n$ variables** is a rule $f: D \subseteq \mathbb{R}^n \to \mathbb{R}$ that assigns to each point $\boldsymbol{x} = (x_1, x_2, \ldots, x_n)^{\top}$ in its domain $D$ a unique real number $f(\boldsymbol{x}) \in \mathbb{R}$.

For $n = 2$ we typically write $f(x, y)$; for $n = 3$, $f(x, y, z)$. For general $n$ we write $f(\boldsymbol{x})$ or $f(x_1, \ldots, x_n)$.

The **domain** $D$ is the set of inputs for which $f$ is defined. Unless otherwise stated, we take the *natural domain* — the largest subset of $\mathbb{R}^n$ on which the expression makes sense.

The **range** (or image) is $\{\, f(\boldsymbol{x}) \mid \boldsymbol{x} \in D \,\} \subseteq \mathbb{R}$.

**Example 1.** Let $f(x, y) = x^2 + y^2$.
- Domain: all of $\mathbb{R}^2$ (the expression is defined everywhere).
- Range: $[0, \infty)$ (the sum of squares is always non-negative, and every non-negative value is achieved).
- Evaluation: $f(1, 2) = 1 + 4 = 5$; $\;f(-3, 0) = 9$.

**Example 2.** Let $f(x, y) = \log(x^2 + y^2)$.

Here $\log$ denotes the natural logarithm (the convention throughout this course). The expression requires $x^2 + y^2 > 0$, so the natural domain is $\mathbb{R}^2 \setminus \{(0,0)\}$ — the whole plane minus the origin.

**Example 3.** Let $f(x, y) = \dfrac{1}{\sqrt{1 - x^2 - y^2}}$.

The expression requires $1 - x^2 - y^2 > 0$, i.e. $x^2 + y^2 < 1$. The natural domain is the open unit disk in $\mathbb{R}^2$.

### 1.2 Higher-Dimensional Inputs

For $n \geq 3$ there is no difficulty in the definition — only in visualisation. We can no longer draw the graph (which would live in $\mathbb{R}^{n+1}$), but we can still:
- Evaluate $f$ at any point.
- Ask where $f$ is large, where it is small, and how it changes as we move in a given direction.

All the machinery we build in Modules 01–04 applies in full generality to $f: \mathbb{R}^n \to \mathbb{R}$. The $n = 2$ case is where we develop intuition. Higher $n$ is where the statistics lives.

> **Example (machine learning).** The training loss of a neural network is a function $\mathcal{L}: \mathbb{R}^p \to \mathbb{R}$, where $p$ is the number of parameters — often millions. It is a scalar field on a very high-dimensional space. Every gradient descent step is a move on that surface, guided by the gradient of $\mathcal{L}$.

## 2. The Landscape Metaphor

For $f: \mathbb{R}^2 \to \mathbb{R}$, the graph of $f$ is the set of all triples

$$
\{\,(x,\, y,\, f(x,y)) \mid (x,y) \in D\,\} \subseteq \mathbb{R}^3.
$$

This is a **surface** in three-dimensional space. Think of it as terrain: the $xy$-plane is the ground, and $f(x,y)$ gives the elevation at the location $(x,y)$. Peaks are local maxima, valleys are local minima, and saddle points are places where the terrain rises in one direction and falls in another.

This metaphor is not decorative — it is how working practitioners think about optimisation surfaces. When we say a loss function has a "flat region", a "sharp minimum", or a "ravine", we are using landscape language precisely.

### 2.1 Reading a Surface

Given a surface $z = f(x,y)$, learn to ask:

| Feature | What to look for |
|---|---|
| **Local maximum** | A peak — $f$ decreases in every direction from this point |
| **Local minimum** | A valley — $f$ increases in every direction from this point |
| **Saddle point** | Neither — $f$ increases in some directions, decreases in others |
| **Ridge** | An elongated region of high values — the surface is curved in one direction, nearly flat in another |
| **Flat region** | $f$ is nearly constant — the surface is approximately horizontal |
| **Steep region** | $f$ changes rapidly — the surface is nearly vertical |

Identifying these features by eye is the first skill. In Modules 02–04 we will identify them analytically.

### 2.2 Three Canonical Surfaces

These three examples recur constantly throughout the course. Commit them to memory.

**Paraboloid** (bowl):
$$
f(x,y) = x^2 + y^2
$$
Has a unique global minimum at the origin. Convex. The loss surface of unregularised least squares regression is exactly a paraboloid in coefficient space.

**Saddle**:
$$
f(x,y) = x^2 - y^2
$$
The origin is a critical point that is neither a maximum nor a minimum: $f$ increases along the $x$-axis and decreases along the $y$-axis. This is the simplest saddle. In high-dimensional optimisation, saddle points are far more common than local minima.

**Gaussian bell**:
$$
f(x,y) = e^{-(x^2 + y^2)/2}
$$
Has a unique global maximum at the origin; $f \to 0$ as $\|(x,y)\| \to \infty$. This is the radially symmetric version of the normal density. Its level curves (introduced in Unit 01-02) are concentric circles.

## 3. Scalar Fields

A function $f: \mathbb{R}^n \to \mathbb{R}$ is called a **scalar field** when we want to emphasise that $f$ assigns a *scalar value* to each *point in space*, rather than merely thinking of it as an abstract function.

The word "field" carries the physical intuition: imagine the domain as a region of space, and at every location in that region, there is a number — temperature, air pressure, electrostatic potential, log-likelihood. The field is the totality of all those numbers at all those locations.

**Examples of scalar fields in physics and engineering:**
- Temperature field $T(x, y, z)$: assigns the temperature at each point in a room.
- Pressure field $p(x, y, z)$: assigns air pressure at each point in the atmosphere.
- Gravitational potential $\Phi(x, y, z)$: assigns potential energy per unit mass at each point.

**Examples of scalar fields in statistics and machine learning:**
- Loss surface $\mathcal{L}(\boldsymbol{\theta})$: assigns training loss to each parameter vector.
- Log-posterior $\log p(\boldsymbol{\theta} \mid \boldsymbol{x})$: assigns a log-probability to each parameter value.
- Probability density $p(x_1, x_2)$: assigns a density value to each point in data space.

### 3.1 Scalar Fields vs. Vector Fields

Not all fields assign scalars. A **vector field** $\boldsymbol{F}: \mathbb{R}^n \to \mathbb{R}^n$ assigns a *vector* — rather than a number — to each point in space. Wind velocity is a vector field: at each point in the atmosphere, the wind has both a speed (magnitude) and a direction.

The distinction:

| | Input | Output | Example |
|---|---|---|---|
| **Scalar field** | point in $\mathbb{R}^n$ | real number | temperature $T(x,y,z)$ |
| **Vector field** | point in $\mathbb{R}^n$ | vector in $\mathbb{R}^n$ | wind velocity $\boldsymbol{v}(x,y,z)$ |

This distinction matters because the **gradient** of a scalar field is a vector field. When we write $\nabla f$ in Module 03, we are building a vector field from a scalar field: at each point $\boldsymbol{x}$, the gradient $\nabla f(\boldsymbol{x})$ is a vector that points in the direction of steepest ascent of $f$.

Vector fields receive their own treatment in Unit 01-03. This unit focuses on scalar fields.

## 4. Statistical Bridge — Probability Densities as Scalar Fields [PD]

A probability density function (PDF) is a scalar field. This connection is worth spelling out carefully because it makes the geometry of distributions immediately accessible.

### 4.1 The Univariate Gaussian as a Scalar Field on $\mathbb{R}$

The Gaussian PDF with mean $\mu$ and variance $\sigma^2$ is

$$
p(x) = \frac{1}{\sqrt{2\pi\sigma^2}}\,\exp\!\left(-\frac{(x - \mu)^2}{2\sigma^2}\right).
$$

This is a function $p: \mathbb{R} \to \mathbb{R}_{\geq 0}$. It assigns a non-negative number to each point $x \in \mathbb{R}$. The landscape is the familiar bell curve.

### 4.2 The Bivariate Gaussian as a Scalar Field on $\mathbb{R}^2$

The bivariate Gaussian with mean vector $\boldsymbol{\mu} = (\mu_1, \mu_2)^{\top}$ and covariance matrix $\boldsymbol{\Sigma}$ is

$$
p(x_1, x_2) = \frac{1}{2\pi\,|\boldsymbol{\Sigma}|^{1/2}}\,\exp\!\left(-\frac{1}{2}(\boldsymbol{x} - \boldsymbol{\mu})^{\top}\boldsymbol{\Sigma}^{-1}(\boldsymbol{x} - \boldsymbol{\mu})\right).
$$

This is a function $p: \mathbb{R}^2 \to \mathbb{R}_{\geq 0}$ — a scalar field on the plane. Its graph is a bell surface; its level curves (sets of constant density) are ellipses. We will return to this in Unit 01-02.

### 4.3 The PDF as a Scalar Field Over *Parameter* Space

There is a second, equally important way to think about the Gaussian as a scalar field. Fix a data set $\{x_1, \ldots, x_n\}$ and let the *parameters* $\mu$ and $\sigma^2$ vary. The **likelihood function** is

$$
L(\mu, \sigma^2) = \prod_{i=1}^{n} \frac{1}{\sqrt{2\pi\sigma^2}}\,\exp\!\left(-\frac{(x_i - \mu)^2}{2\sigma^2}\right),
$$

and its logarithm — the **log-likelihood** — is

$$
\ell(\mu, \sigma^2) = -\frac{n}{2}\log(2\pi\sigma^2) - \frac{1}{2\sigma^2}\sum_{i=1}^{n}(x_i - \mu)^2.
$$

This is a scalar field on *parameter space* $\mathbb{R} \times \mathbb{R}_{>0}$ (where the two axes are $\mu$ and $\sigma^2$). It assigns a log-likelihood value to each pair of parameter values $(\mu, \sigma^2)$.

In Module 00, Unit 03 we found the MLE for $\mu$ by treating $\ell$ as a function of one variable (holding $\sigma^2$ fixed). In Module 02 we will differentiate $\ell$ with respect to *both* parameters simultaneously — the full multivariable treatment. The surface we plot in the Python lab below is exactly that function.

> **Key insight.** Maximum likelihood estimation is an optimisation problem on a scalar field: find the point $(\mu, \sigma^2)$ at which $\ell(\mu, \sigma^2)$ is maximised. The geometry of that surface — whether it has a single sharp peak or a broad ridge — determines how well the optimisation problem is conditioned.

## Section 1 — The Three Canonical Surfaces

We plot the paraboloid $f(x,y) = x^2 + y^2$, the saddle $f(x,y) = x^2 - y^2$, and the Gaussian bell $f(x,y) = e^{-(x^2+y^2)/2}$ side by side.

Each plot shows both the 3D surface and a 2D contour map beneath it. Read both views together: the contour lines are the "shadow" of the surface projected onto the floor of the plot — they show regions of equal height.

```python
# --- Grid ---
lim = 2.5
x_vals = np.linspace(-lim, lim, 200)
y_vals = np.linspace(-lim, lim, 200)
X, Y = np.meshgrid(x_vals, y_vals)

# --- Three canonical scalar fields ---
Z_bowl   = X**2 + Y**2
Z_saddle = X**2 - Y**2
Z_bell   = np.exp(-(X**2 + Y**2) / 2)

surfaces = [
    (Z_bowl,   r'Paraboloid  $f = x^2 + y^2$',  'Blues',   'minimum at origin'),
    (Z_saddle, r'Saddle  $f = x^2 - y^2$',       'RdBu',    'saddle at origin'),
    (Z_bell,   r'Gaussian bell  $f = e^{-(x^2+y^2)/2}$', 'Greens', 'maximum at origin'),
]

fig = plt.figure(figsize=(18, 6))

for col, (Z, title, cmap, note) in enumerate(surfaces):
    ax = fig.add_subplot(1, 3, col + 1, projection='3d')
    ax.plot_surface(X, Y, Z, cmap=cmap, alpha=0.85, linewidth=0, antialiased=True)
    ax.contour(X, Y, Z, zdir='z', offset=Z.min() - 0.1 * (Z.max() - Z.min()),
               levels=10, cmap=cmap, alpha=0.6)
    ax.set_title(title, fontweight='bold', color=TS_GRAY, pad=10)
    ax.set_xlabel('$x$', labelpad=4)
    ax.set_ylabel('$y$', labelpad=4)
    ax.set_zlabel('$f(x,y)$', labelpad=4)
    ax.tick_params(labelsize=8)
    ax.text2D(0.5, -0.04, note, transform=ax.transAxes,
              ha='center', fontsize=9, color=TS_BLUE, style='italic')

plt.suptitle('Module 01, Unit 01 — Three Canonical Scalar Fields',
             fontsize=14, fontweight='bold', color=TS_GRAY, y=1.01)
plt.tight_layout()
plt.savefig('module-01-unit-01-canonical-surfaces.png', dpi=150, bbox_inches='tight')
plt.show()
print('Figure saved.')
```

### What Do You See?

1. **Paraboloid**: The contour lines projected onto the floor are concentric circles. What does the spacing between contour lines tell you about the steepness of the surface? Are the circles equally spaced in radius, or do they get closer together as you move outward?

2. **Saddle**: The 3D shape rises as you move along the $x$-axis and falls as you move along the $y$-axis. What do the contour lines look like? Can you identify the saddle point at the origin just from the contour pattern?

3. **Gaussian bell**: Compare the contour spacing here to the paraboloid. Near the peak the surface is flat (contours widely spaced); near the flanks it falls rapidly (contours bunched together). How does this compare to the paraboloid's behaviour?

## Section 2 — Exploring Scalar Fields

We now look at two less symmetric scalar fields to practice feature identification: a function with multiple critical points, and a function with a ridge structure.

For each we plot both the 3D surface (left panel) and a flat contour map (right panel). The contour map is often easier to read for identifying topology — we will rely on it heavily in Unit 01-02.

```python
# --- Function 1: a surface with multiple critical points ---
# f(x,y) = sin(x) * cos(y) on [-pi, pi]^2
# This has a checkerboard of maxima and minima — a good pattern-reading exercise.

x_sc = np.linspace(-np.pi, np.pi, 300)
y_sc = np.linspace(-np.pi, np.pi, 300)
X_sc, Y_sc = np.meshgrid(x_sc, y_sc)
Z_sc = np.sin(X_sc) * np.cos(Y_sc)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Left: 3D surface
ax3d = fig.add_subplot(1, 2, 1, projection='3d')
ax3d.plot_surface(X_sc, Y_sc, Z_sc, cmap='coolwarm', alpha=0.9,
                  linewidth=0, antialiased=True)
ax3d.set_title(r'$f(x,y) = \sin(x)\cos(y)$ — surface', fontweight='bold', color=TS_GRAY)
ax3d.set_xlabel('$x$')
ax3d.set_ylabel('$y$')
ax3d.set_zlabel('$f$')
ax3d.tick_params(labelsize=8)

# Right: contour map
ax2d = axes[1]
cf = ax2d.contourf(X_sc, Y_sc, Z_sc, levels=20, cmap='coolwarm')
ax2d.contour(X_sc, Y_sc, Z_sc, levels=20, colors='k', linewidths=0.4, alpha=0.5)
plt.colorbar(cf, ax=ax2d, label='$f(x,y)$')
ax2d.set_title(r'$f(x,y) = \sin(x)\cos(y)$ — contour map', fontweight='bold', color=TS_GRAY)
ax2d.set_xlabel('$x$')
ax2d.set_ylabel('$y$')
ax2d.set_aspect('equal')

# Remove the empty second subplot from axes array
fig.delaxes(axes[0])

plt.suptitle('Module 01, Unit 01 — Multiple Critical Points',
             fontsize=13, fontweight='bold', color=TS_GRAY)
plt.tight_layout()
plt.savefig('module-01-unit-01-sin-cos.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# --- Function 2: a ridge surface ---
# f(x,y) = exp(-y^2) — independent of x, so it forms an infinite ridge along the x-axis.
# This mimics the shape of a likelihood surface that is poorly identified in one direction.

x_r = np.linspace(-3, 3, 300)
y_r = np.linspace(-3, 3, 300)
X_r, Y_r = np.meshgrid(x_r, y_r)
Z_r = np.exp(-Y_r**2)  # constant in x — a ridge

fig2 = plt.figure(figsize=(14, 5))

# Left: 3D
ax3d2 = fig2.add_subplot(1, 2, 1, projection='3d')
ax3d2.plot_surface(X_r, Y_r, Z_r, cmap='Blues', alpha=0.9,
                   linewidth=0, antialiased=True)
ax3d2.set_title(r'$f(x,y) = e^{-y^2}$ — ridge surface', fontweight='bold', color=TS_GRAY)
ax3d2.set_xlabel('$x$')
ax3d2.set_ylabel('$y$')
ax3d2.set_zlabel('$f$')
ax3d2.tick_params(labelsize=8)

# Right: contour
ax2d2 = fig2.add_subplot(1, 2, 2)
cf2 = ax2d2.contourf(X_r, Y_r, Z_r, levels=15, cmap='Blues')
ax2d2.contour(X_r, Y_r, Z_r, levels=15, colors='k', linewidths=0.4, alpha=0.5)
plt.colorbar(cf2, ax=ax2d2, label='$f(x,y)$')
ax2d2.set_title(r'$f(x,y) = e^{-y^2}$ — contour map (parallel lines = ridge)',
                fontweight='bold', color=TS_GRAY)
ax2d2.set_xlabel('$x$')
ax2d2.set_ylabel('$y$')
ax2d2.set_aspect('equal')

plt.suptitle('Module 01, Unit 01 — Ridge Structure',
             fontsize=13, fontweight='bold', color=TS_GRAY)
plt.tight_layout()
plt.savefig('module-01-unit-01-ridge.png', dpi=150, bbox_inches='tight')
plt.show()

print()
print('Ridge observation: the contour lines are straight and parallel.')
print('This means f is constant in the x-direction — moving along x changes nothing.')
print('An optimiser on this surface has a whole line of equally good solutions.')
```

### What Do You See?

1. **$\sin(x)\cos(y)$**: How many maxima can you count on the contour map? How many minima? The function has the same value of $+1$ at all maxima and $-1$ at all minima — can you verify this from the colour scale?

2. **Ridge surface**: The contour lines are parallel and horizontal (constant in $x$). What does this mean for an optimiser trying to maximise $f$? If you told the optimiser "move to a higher value", and it moved along the ridge (changed $x$ without changing $y$), would $f$ increase, decrease, or stay the same?

   This ridge geometry is directly relevant to poorly identified statistical models, where moving along certain directions in parameter space does not change the likelihood — a condition called *non-identifiability*.

## Section 3 — Statistical Bridge: The Log-Likelihood as a Scalar Field [PD]

We now visualise the Gaussian log-likelihood as a scalar field over the two-dimensional parameter space $(\mu, \sigma^2)$.

**Setup.** Generate $n = 30$ observations from a Gaussian with true parameters $\mu^* = 2.0$, $\sigma^{2*} = 1.0$. Then evaluate

$$
\ell(\mu, \sigma^2) = -\frac{n}{2}\log(2\pi\sigma^2) - \frac{1}{2\sigma^2}\sum_{i=1}^{n}(x_i - \mu)^2
$$

over a grid of $(\mu, \sigma^2)$ values and plot the resulting surface. The maximum of this surface is the MLE.

```python
# --- Generate data ---
rng = np.random.default_rng(seed=42)
n       = 30
mu_true = 2.0
s2_true = 1.0   # sigma^2
data    = rng.normal(loc=mu_true, scale=np.sqrt(s2_true), size=n)

x_bar = data.mean()
s2_hat = data.var()   # MLE: (1/n) * sum((x_i - x_bar)^2)

print(f'Sample size n          = {n}')
print(f'True mu                = {mu_true:.4f}')
print(f'True sigma^2           = {s2_true:.4f}')
print(f'MLE mu_hat  = x_bar    = {x_bar:.4f}')
print(f'MLE sigma^2_hat        = {s2_hat:.4f}')
```

```python
# --- Evaluate log-likelihood on a grid ---
mu_vals = np.linspace(mu_true - 2.0, mu_true + 2.0, 300)
s2_vals = np.linspace(0.2, 4.0, 300)           # sigma^2 must be > 0
MU, S2  = np.meshgrid(mu_vals, s2_vals)

# Vectorised log-likelihood
# ell(mu, s2) = -n/2 * log(2*pi*s2) - (1/(2*s2)) * sum((x_i - mu)^2)
# Broadcasting: data shape (n,) against MU/S2 shape (300, 300)
residuals_sq = np.sum((data[:, None, None] - MU[None, :, :])**2, axis=0)  # shape (300,300)
LOGLIK = -n / 2 * np.log(2 * np.pi * S2) - residuals_sq / (2 * S2)

# Locate the numerical maximum
max_idx    = np.unravel_index(np.argmax(LOGLIK), LOGLIK.shape)
mu_peak    = MU[max_idx]
s2_peak    = S2[max_idx]
print(f'Numerical MLE from grid: mu = {mu_peak:.4f},  sigma^2 = {s2_peak:.4f}')
print(f'Analytical MLE:          mu = {x_bar:.4f},  sigma^2 = {s2_hat:.4f}')
print('(Small discrepancy is grid resolution — they converge as the grid is refined.)')
```

```python
# --- Plot: surface + contour map side by side ---
fig = plt.figure(figsize=(16, 6))

# ---- Left: 3D surface ----
ax3d = fig.add_subplot(1, 2, 1, projection='3d')
surf = ax3d.plot_surface(MU, S2, LOGLIK, cmap='Greens', alpha=0.88,
                          linewidth=0, antialiased=True)
ax3d.scatter([mu_peak], [s2_peak], [LOGLIK[max_idx]],
             color=TS_AMBER, s=80, zorder=5, label='Numerical MLE')
ax3d.set_xlabel('$\\mu$', labelpad=6)
ax3d.set_ylabel('$\\sigma^2$', labelpad=6)
ax3d.set_zlabel('$\\ell(\\mu, \\sigma^2)$', labelpad=6)
ax3d.set_title('Log-likelihood surface', fontweight='bold', color=TS_GRAY)
ax3d.tick_params(labelsize=8)
ax3d.legend(fontsize=9)

# ---- Right: contour map ----
ax2d = fig.add_subplot(1, 2, 2)
levels = np.linspace(LOGLIK.max() - 60, LOGLIK.max(), 25)
cf = ax2d.contourf(MU, S2, LOGLIK, levels=levels, cmap='Greens')
ax2d.contour(MU, S2, LOGLIK, levels=levels, colors='k', linewidths=0.4, alpha=0.4)
plt.colorbar(cf, ax=ax2d, label='$\\ell(\\mu, \\sigma^2)$')

# Mark true parameters and MLE
ax2d.scatter([mu_true], [s2_true], color=TS_BLUE, s=80, zorder=5,
             label=f'True $(\\mu^*={mu_true},\\, \\sigma^{{2*}}={s2_true})$')
ax2d.scatter([x_bar], [s2_hat], color=TS_AMBER, s=80, marker='*', zorder=5,
             label=f'MLE $(\\hat{{\\mu}}={x_bar:.2f},\\, \\hat{{\\sigma}}^2={s2_hat:.2f})$')
ax2d.axvline(x_bar, color=TS_AMBER, lw=1, ls='--', alpha=0.6)
ax2d.axhline(s2_hat, color=TS_AMBER, lw=1, ls='--', alpha=0.6)

ax2d.set_xlabel('$\\mu$')
ax2d.set_ylabel('$\\sigma^2$')
ax2d.set_title('Log-likelihood contour map — parameter space',
               fontweight='bold', color=TS_GRAY)
ax2d.legend(fontsize=9, loc='upper right')

plt.suptitle('Module 01, Unit 01 — Gaussian Log-Likelihood as a Scalar Field [PD]',
             fontsize=13, fontweight='bold', color=TS_GRAY)
plt.tight_layout()
plt.savefig('module-01-unit-01-loglik-surface.png', dpi=150, bbox_inches='tight')
plt.show()
print('Figure saved.')
```

## Summary

| Concept | Key result |
|---|---|
| Function of $n$ variables | $f: D \subseteq \mathbb{R}^n \to \mathbb{R}$; inputs are points, output is a scalar |
| Natural domain | Largest subset of $\mathbb{R}^n$ on which $f$ is defined |
| Scalar field | A function $f: \mathbb{R}^n \to \mathbb{R}$ viewed as assigning a value to every point in space |
| Vector field | A function $\boldsymbol{F}: \mathbb{R}^n \to \mathbb{R}^n$ assigning a vector to every point — distinct from a scalar field |
| Paraboloid | $f = x^2 + y^2$ — bowl, unique global minimum at origin, convex |
| Saddle | $f = x^2 - y^2$ — critical point at origin that is neither max nor min |
| Gaussian bell | $f = e^{-(x^2+y^2)/2}$ — unique global maximum at origin, decays to zero |
| Ridge | Constant in one direction; contours are parallel lines; gradient is zero in that direction |
| Log-likelihood as scalar field | $\ell(\mu, \sigma^2)$ is a scalar field on parameter space; MLE is its argmax |
