## 1. Limits, Continuity, and Differentiability

These three properties form a hierarchy. Differentiability implies continuity; continuity implies the limit exists at that point — but neither implication runs the other way.

### 1.1 Limits

**Definition.** We say $\lim_{x \to a} f(x) = L$ if for every $\varepsilon > 0$ there exists $\delta > 0$ such that

$$
0 < |x - a| < \delta \implies |f(x) - L| < \varepsilon.
$$

The formal definition is rarely used in computation, but its logic is worth absorbing once: the limit is about what $f$ *approaches*, not what it *equals* at $a$. The value $f(a)$ need not exist for $\lim_{x \to a} f(x)$ to be defined.

For the functions we encounter in this course — polynomials, exponentials, logarithms, and compositions of these — limits at interior points of the domain can almost always be evaluated by direct substitution.

> **Notation check.** Throughout this course, $\log$ denotes the natural logarithm (base $e$), consistent with the convention in statistics and machine learning. When we write $\log p(x \mid \theta)$, we mean $\ln p(x \mid \theta)$. We never use $\log_{10}$.

### 1.2 Continuity

**Definition.** A function $f$ is **continuous** at $a$ if:
1. $f(a)$ is defined,
2. $\lim_{x \to a} f(x)$ exists, and
3. $\lim_{x \to a} f(x) = f(a)$.

Informally: no jumps, no holes, no vertical asymptotes at $a$. All three conditions collapse into one when $f$ is, say, a polynomial — which is why polynomials are continuous everywhere.

For our purposes, continuity matters most because the optimization theorems we use (the Extreme Value Theorem, Fermat's Theorem) require it. A log-likelihood that is discontinuous in $\theta$ would be pathological — and fortunately, for the distributions we study, it never is.

### 1.3 Differentiability

**Definition.** The **derivative** of $f$ at $a$ is

$$
f'(a) = \lim_{h \to 0} \frac{f(a + h) - f(a)}{h},
$$

provided the limit exists. If it does, $f$ is **differentiable** at $a$.

The derivative is simultaneously:
- The **slope of the tangent line** to the graph of $f$ at $(a, f(a))$,
- The **instantaneous rate of change** of $f$ at $a$, and
- The **best linear approximation** to $f$ near $a$: $f(a + h) \approx f(a) + f'(a)\,h$ for small $h$.

This third interpretation — derivative as best linear approximation — is the one that generalizes cleanest to multiple variables. In Module 01 we will replace $f'(a)$ with a gradient vector, and the same approximation idea carries over.

**Differentiability implies continuity.** The converse is false. The canonical counterexample is $f(x) = |x|$, which is continuous at $0$ but not differentiable there — the left- and right-hand limits of the difference quotient disagree.

## 2. Differentiation Rules

We assume fluency with the standard rules. This section records them in the notation used throughout the course and highlights the ones that appear most often in the statistical material.

### 2.1 Basic Rules

Let $f$ and $g$ be differentiable functions of $x$, and let $c$ be a constant.

| Rule | Statement |
|---|---|
| Constant | $\frac{d}{dx}[c] = 0$ |
| Power | $\frac{d}{dx}[x^n] = n\,x^{n-1}$ |
| Constant multiple | $\frac{d}{dx}[c\,f(x)] = c\,f'(x)$ |
| Sum | $\frac{d}{dx}[f(x) + g(x)] = f'(x) + g'(x)$ |
| Exponential | $\frac{d}{dx}[e^x] = e^x$ |
| Natural log | $\frac{d}{dx}[\log x] = \dfrac{1}{x}$ |

### 2.2 Product Rule

$$
\frac{d}{dx}[f(x)\,g(x)] = f'(x)\,g(x) + f(x)\,g'(x)
$$

**Example.** Let $h(x) = x^2 e^x$. Then

$$
h'(x) = 2x\,e^x + x^2\,e^x = e^x(x^2 + 2x).
$$

### 2.3 Chain Rule

The chain rule is the single most important differentiation rule in this course. It appears every time we differentiate a composition, and the multivariable version (Module 02) is a direct generalization.

**Chain Rule.** If $h(x) = f(g(x))$, then

$$
h'(x) = f'(g(x)) \cdot g'(x).
$$

Read this as: *derivative of the outside, evaluated at the inside, times the derivative of the inside.*

**Example 1.** Let $h(x) = e^{-x^2/2}$. Take $f(u) = e^u$ and $g(x) = -x^2/2$. Then $f'(u) = e^u$ and $g'(x) = -x$, so

$$
h'(x) = e^{-x^2/2} \cdot (-x) = -x\,e^{-x^2/2}.
$$

This is the derivative of the Gaussian kernel — one of the most frequently differentiated expressions in the course.

**Example 2.** Let $h(\theta) = \log(1 + e^\theta)$ (the softplus function). Take $f(u) = \log u$ and $g(\theta) = 1 + e^\theta$. Then

$$
h'(\theta) = \frac{1}{1 + e^\theta} \cdot e^\theta = \frac{e^\theta}{1 + e^\theta} = \sigma(\theta),
$$

where $\sigma(\theta) = 1/(1 + e^{-\theta})$ is the **logistic sigmoid** function. This chain rule calculation appears verbatim in the GLM modules.

> ⚠ **Watch out.** The most common chain rule error is forgetting $g'(x)$ — writing $h'(x) = f'(g(x))$ and stopping. Always ask: *did I multiply by the derivative of the inside?*

If chain rule applications feel uncertain, stop here and visit:
→ [Single-Variable Chain Rule refresher](../../refreshers/single-variable-chain-rule/)

### 2.4 Quotient Rule

$$
\frac{d}{dx}\left[\frac{f(x)}{g(x)}\right] = \frac{f'(x)\,g(x) - f(x)\,g'(x)}{[g(x)]^2}
$$

In practice, quotient rule situations in this course are usually handled by rewriting the quotient as a product (e.g., $f(x) \cdot [g(x)]^{-1}$) and applying the product and chain rules. Either approach is correct; choose the one you find less error-prone.

## 3. Integration

We use integration in two distinct ways in this course: to compute normalizing constants for probability distributions (Module 05), and occasionally to verify a result by reversing a differentiation. The level of integration fluency required is therefore moderate — you need to recognize and evaluate standard antiderivatives, apply integration by substitution, and state the Fundamental Theorem.

### 3.1 The Fundamental Theorem of Calculus

**FTC, Part 1.** If $F'(x) = f(x)$, then

$$
\int_a^b f(x)\,dx = F(b) - F(a).
$$

**FTC, Part 2.** The function $G(x) = \int_a^x f(t)\,dt$ is differentiable and $G'(x) = f(x)$.

Part 2 is the one that often surprises students: differentiation and integration are inverse operations. This is what makes antiderivatives — also called indefinite integrals — the correct tool for evaluating definite integrals.

### 3.2 Standard Antiderivatives

The table below records the antiderivatives used most often in this course. Each is a direct consequence of reversing the corresponding differentiation rule.

| Function $f(x)$ | Antiderivative $F(x)$ |
|---|---|
| $x^n$ ($ n \neq -1$) | $\dfrac{x^{n+1}}{n+1}$ |
| $\dfrac{1}{x}$ | $\log|x|$ |
| $e^x$ | $e^x$ |
| $e^{cx}$ | $\dfrac{1}{c}e^{cx}$ |
| $\cos x$ | $\sin x$ |
| $\sin x$ | $-\cos x$ |

All antiderivatives carry an arbitrary constant $+\,C$, which cancels when evaluating a definite integral via $F(b) - F(a)$.

### 3.3 Substitution

Substitution is the integration analogue of the chain rule. If the integrand contains a recognizable inner function $u = g(x)$ and its derivative $g'(x)$ appears as a factor, substitute:

$$
\int f(g(x))\,g'(x)\,dx = \int f(u)\,du.
$$

**Example.** Compute $\displaystyle\int_0^1 x\,e^{-x^2}\,dx$.

Let $u = -x^2$, so $du = -2x\,dx$ and $x\,dx = -\tfrac{1}{2}\,du$. When $x=0$, $u=0$; when $x=1$, $u=-1$. Therefore:

$$
\int_0^1 x\,e^{-x^2}\,dx
= \int_0^{-1} e^u \cdot \left(-\tfrac{1}{2}\right)du
= \frac{1}{2}\int_{-1}^{0} e^u\,du
= \frac{1}{2}\left[e^u\right]_{-1}^{0}
= \frac{1}{2}(1 - e^{-1}).
$$

This integral type — $\int x\,e^{-x^2}\,dx$ — appears in the Gaussian normalization computations of Module 05 and Module 06.

## Statistical Bridge: Log-Likelihood as a Function to Differentiate [MLE]

Everything in Sections 1–3 converges here. Maximum likelihood estimation is, at its core, a single-variable (or later, multivariable) calculus problem: find the parameter value $\theta$ that maximises $\ell(\theta)$.

### Setup

Suppose we observe $n$ independent values $x_1, x_2, \ldots, x_n$, each drawn from a normal distribution $\mathcal{N}(\mu, \sigma^2)$ with **known** $\sigma^2$ and **unknown** $\mu$. The likelihood is the joint density, viewed as a function of $\mu$:

$$
L(\mu) = \prod_{i=1}^n \frac{1}{\sqrt{2\pi\sigma^2}} \exp\!\left(-\frac{(x_i - \mu)^2}{2\sigma^2}\right).
$$

Taking the natural log and using $\log(ab) = \log a + \log b$ and $\log(e^u) = u$:

$$
\ell(\mu)
= \log L(\mu)
= -\frac{n}{2}\log(2\pi\sigma^2) - \frac{1}{2\sigma^2}\sum_{i=1}^n (x_i - \mu)^2.
$$

The first term does not depend on $\mu$, so it vanishes when we differentiate.

### Finding the MLE

Differentiate $\ell(\mu)$ with respect to $\mu$, applying the chain rule to $(x_i - \mu)^2$:

$$
\frac{d\ell}{d\mu}
= -\frac{1}{2\sigma^2} \cdot \sum_{i=1}^n 2(x_i - \mu)(-1)
= \frac{1}{\sigma^2}\sum_{i=1}^n (x_i - \mu).
$$

Set this equal to zero and solve:

$$
\sum_{i=1}^n (x_i - \mu) = 0
\implies \sum_{i=1}^n x_i = n\mu
\implies \hat{\mu}_{\text{MLE}} = \frac{1}{n}\sum_{i=1}^n x_i = \bar{x}.
$$

The MLE for the mean of a Gaussian is the sample mean. This result is important not because it is surprising, but because the **method** — write the log-likelihood, differentiate, set to zero — is the template we will apply to every distribution in the course. The chain rule step ($\frac{d}{d\mu}(x_i - \mu)^2 = -2(x_i - \mu)$) is exactly the single-variable chain rule from Section 2.3.

To confirm this is a maximum (not a minimum), note that $\ell$ is a concave function of $\mu$ — the leading coefficient of the quadratic $\sum(x_i - \mu)^2$ is positive, so $\ell(\mu)$ is a downward-opening parabola in $\mu$. A unique global maximum exists.

> **Forward link.** In Module 02 we will differentiate the log-likelihood with respect to *both* $\mu$ and $\sigma^2$ simultaneously — the multivariable version of this calculation. In Module 03 the gradient of $\ell$ replaces the single derivative. In Module 04 we confirm the maximum using the Hessian.

```python
# --- Define symbolic variables ---
x, theta, mu, sigma2 = sp.symbols('x theta mu sigma2', real=True)

print('Symbolic variables defined:', x, theta, mu, sigma2)
```

```python
# --- Example 1: Chain rule on the Gaussian kernel ---
# h(x) = exp(-x^2 / 2)
# Expected: h'(x) = -x * exp(-x^2 / 2)

h1 = sp.exp(-x**2 / 2)
dh1 = sp.diff(h1, x)

print('h(x) =', h1)
print("h'(x) =", dh1)
print("Simplified h'(x) =", sp.simplify(dh1))

# --- Example 2: Chain rule on softplus ---
# h(theta) = log(1 + exp(theta))
# Expected: h'(theta) = exp(theta) / (1 + exp(theta))  [the logistic sigmoid]

h2 = sp.log(1 + sp.exp(theta))
dh2 = sp.diff(h2, theta)

print('\nh(theta) =', h2)
print("h'(theta) =", dh2)
print("Simplified h'(theta) =", sp.simplify(dh2))
```

```python
# --- Product rule: h(x) = x^2 * exp(x) ---
# Expected: h'(x) = e^x * (x^2 + 2x)

h3 = x**2 * sp.exp(x)
dh3 = sp.diff(h3, x)

print('h(x) =', h3)
print("h'(x) =", dh3)
print("Factored h'(x) =", sp.factor(dh3))

# --- Your turn: differentiate h(x) = log(x^2 + 1) ---
# Write the expected result by hand before running.
h4 = sp.log(x**2 + 1)
dh4 = sp.diff(h4, x)

print('\nh(x) =', h4)
print("h'(x) =", dh4)

# Enter your hand-computed answer here to compare:
my_dh4 = None  # e.g. 2*x / (x**2 + 1)
if my_dh4 is not None:
    print('Match:', sp.simplify(dh4 - my_dh4) == 0)
```

## Section 2 — Tangent Lines and the Derivative as Slope

The derivative $f'(a)$ is the slope of the tangent line to $f$ at $x = a$. The tangent line passes through $(a,\, f(a))$ with slope $f'(a)$, giving the linear approximation:

$$
T_a(x) = f(a) + f'(a)(x - a).
$$

We visualise this for $f(x) = e^{-x^2/2}$ at three different points. This plot also previews what the derivative *function* looks like relative to the original.

```python
# --- Tangent lines to the Gaussian kernel ---

def f(x_val):
    return np.exp(-x_val**2 / 2)

def df(x_val):
    """Derivative: -x * exp(-x^2 / 2)"""
    return -x_val * np.exp(-x_val**2 / 2)

x_vals = np.linspace(-3, 3, 400)
tangent_points = [-1.5, 0.0, 1.0]
colors = [TS_AMBER, TS_RED, TS_GREEN]

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# --- Left panel: f(x) with tangent lines ---
ax = axes[0]
ax.plot(x_vals, f(x_vals), color=TS_BLUE, linewidth=2.5,
        label=r'$f(x) = e^{-x^2/2}$')

for a, col in zip(tangent_points, colors):
    fa  = f(a)
    dfa = df(a)
    # Tangent line: T_a(x) = f(a) + f'(a)(x - a)
    T   = fa + dfa * (x_vals - a)
    # Clip to a visible window so it doesn't dominate
    T_clipped = np.clip(T, -0.5, 1.5)
    ax.plot(x_vals, T_clipped, color=col, linewidth=1.5, linestyle='--', alpha=0.85,
            label=rf'Tangent at $a={a}$, slope $={dfa:.3f}$')
    ax.scatter([a], [fa], color=col, zorder=5, s=60)

ax.set_xlim(-3, 3)
ax.set_ylim(-0.3, 1.35)
ax.set_xlabel('$x$')
ax.set_ylabel('$f(x)$')
ax.set_title(r'$f(x) = e^{-x^2/2}$ with Tangent Lines', fontweight='bold', color=TS_BLUE)
ax.legend(fontsize=9, loc='upper right')
ax.axhline(0, color=TS_GRAY, linewidth=0.8)
ax.axvline(0, color=TS_GRAY, linewidth=0.8)

# --- Right panel: f(x) and f'(x) together ---
ax2 = axes[1]
ax2.plot(x_vals, f(x_vals),  color=TS_BLUE,  linewidth=2.5, label=r"$f(x)$")
ax2.plot(x_vals, df(x_vals), color=TS_AMBER, linewidth=2.5, linestyle='--',
         label=r"$f'(x) = -x\,e^{-x^2/2}$")
ax2.axhline(0, color=TS_GRAY, linewidth=0.8)
ax2.axvline(0, color=TS_GRAY, linewidth=0.8)
# Mark where f is increasing / decreasing
ax2.fill_between(x_vals, 0, df(x_vals),
                 where=df(x_vals) > 0, alpha=0.12, color=TS_BLUE,
                 label=r"$f' > 0$ (f increasing)")
ax2.fill_between(x_vals, 0, df(x_vals),
                 where=df(x_vals) < 0, alpha=0.12, color=TS_RED,
                 label=r"$f' < 0$ (f decreasing)")
ax2.set_xlim(-3, 3)
ax2.set_xlabel('$x$')
ax2.set_title(r"$f(x)$ and $f'(x)$ — Sign Determines Monotonicity",
              fontweight='bold', color=TS_BLUE)
ax2.legend(fontsize=9)

plt.tight_layout()
plt.savefig('m00-u03-tangent-lines.png', dpi=150, bbox_inches='tight')
plt.show()
```

### What Do You See?

1. At $a = 0$, the tangent line is horizontal (slope $= 0$). Why? What does that tell you about $f$ at that point?
2. At $a = -1.5$, the tangent has positive slope; at $a = 1.0$, it has negative slope. How does the right-hand panel confirm this?
3. Where does $f'(x)$ reach its most extreme values? What is happening to $f$ at those $x$-values?

**Analysis.** The left panel shows that each tangent line is a faithful local approximation: near the point of tangency, the dashed line and the curve are nearly indistinguishable. Move far away from $a$ and they diverge — the linear approximation breaks down. The right panel formalises the relationship between $f$ and $f'$: $f$ is increasing exactly where $f' > 0$ (shaded blue) and decreasing where $f' < 0$ (shaded red). The zero of $f'$ at $x = 0$ corresponds to the maximum of $f$. This sign-analysis logic, applied to the log-likelihood in the next section, is how we find MLEs.

## Section 3 — Statistical Bridge: Gaussian Log-Likelihood [MLE]

We now make the MLE calculation from the notes concrete and visual. Given a small dataset, we plot $\ell(\mu)$ as a function of $\mu$ (with $\sigma^2$ fixed), draw the derivative, and confirm that the maximum of $\ell$ occurs exactly at $\bar{x}$.

### Dataset

We use $n = 8$ observations: $\boldsymbol{x} = (2.3,\;3.1,\;2.8,\;3.5,\;2.6,\;3.0,\;2.9,\;3.2)^{\top}$, with known $\sigma^2 = 0.25$.

```python
# --- Data and parameters ---
data    = np.array([2.3, 3.1, 2.8, 3.5, 2.6, 3.0, 2.9, 3.2])
n       = len(data)
sigma2  = 0.25          # Known variance
x_bar   = np.mean(data) # Theoretical MLE

print(f'n        = {n}')
print(f'sigma^2  = {sigma2}')
print(f'x_bar    = {x_bar:.4f}  (this should be the MLE)')

# --- Log-likelihood as a function of mu ---
def log_likelihood(mu_val, x, sigma2):
    """
    ell(mu) = -n/2 * log(2*pi*sigma^2) - 1/(2*sigma^2) * sum((x_i - mu)^2)
    """
    const = -n / 2 * np.log(2 * np.pi * sigma2)
    quad  = -1 / (2 * sigma2) * np.sum((x - mu_val)**2)
    return const + quad

# --- Score (derivative of log-likelihood) as a function of mu ---
def score(mu_val, x, sigma2):
    """
    d/d(mu) ell(mu) = (1/sigma^2) * sum(x_i - mu)
    """
    return (1 / sigma2) * np.sum(x - mu_val)

print('\nLog-likelihood at x_bar:', log_likelihood(x_bar, data, sigma2))
print('Score at x_bar (should be ~0):', score(x_bar, data, sigma2))
```

```python
# --- Plot: log-likelihood and its derivative ---
mu_range = np.linspace(1.5, 4.5, 400)
ell_vals = np.array([log_likelihood(m, data, sigma2) for m in mu_range])
score_vals = np.array([score(m, data, sigma2) for m in mu_range])

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# --- Left: Log-likelihood surface ---
ax = axes[0]
ax.plot(mu_range, ell_vals, color=TS_GREEN, linewidth=2.5,
        label=r'$\ell(\mu)$')
ax.axvline(x_bar, color=TS_AMBER, linewidth=1.8, linestyle='--',
           label=rf'$\hat{{\mu}}_{{\mathrm{{MLE}}}} = \bar{{x}} = {x_bar:.3f}$')
ax.scatter([x_bar], [log_likelihood(x_bar, data, sigma2)],
           color=TS_AMBER, zorder=6, s=80)
# Mark the data points on the x-axis
ax.scatter(data, [log_likelihood(xi, data, sigma2) for xi in data],
           color=TS_BLUE, marker='|', s=200, linewidths=1.5,
           label='Observed data $x_i$', zorder=5)
ax.set_xlabel(r'$\mu$')
ax.set_ylabel(r'$\ell(\mu)$')
ax.set_title('Gaussian Log-Likelihood as a Function of $\\mu$',
             fontweight='bold', color=TS_GREEN)
ax.legend(fontsize=9)

# --- Right: Score (derivative) ---
ax2 = axes[1]
ax2.plot(mu_range, score_vals, color=TS_BLUE, linewidth=2.5,
         label=r"$\ell'(\mu) = \frac{1}{\sigma^2}\sum(x_i - \mu)$")
ax2.axhline(0, color=TS_GRAY, linewidth=1.2)
ax2.axvline(x_bar, color=TS_AMBER, linewidth=1.8, linestyle='--',
            label=rf'$\ell^\prime(\bar{{x}}) = 0$ at $\mu = {x_bar:.3f}$')
ax2.scatter([x_bar], [0], color=TS_AMBER, zorder=6, s=80)
ax2.fill_between(mu_range, 0, score_vals,
                 where=score_vals > 0, alpha=0.13, color=TS_BLUE,
                 label=r"$\ell' > 0$ ($\ell$ increasing)")
ax2.fill_between(mu_range, 0, score_vals,
                 where=score_vals < 0, alpha=0.13, color=TS_RED,
                 label=r"$\ell' < 0$ ($\ell$ decreasing)")
ax2.set_xlabel(r'$\mu$')
ax2.set_ylabel(r"$\ell'(\mu)$  (Score)")
ax2.set_title("Score Function — Zero at the MLE",
              fontweight='bold', color=TS_BLUE)
ax2.legend(fontsize=9)

plt.tight_layout()
plt.savefig('m00-u03-log-likelihood.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# --- Numerical verification ---
from scipy.optimize import minimize_scalar

# Maximise ell <=> minimise -ell
result = minimize_scalar(lambda m: -log_likelihood(m, data, sigma2),
                         bounds=(1.0, 5.0), method='bounded')

print('=== MLE Verification ===')
print(f'Analytical MLE (x_bar):     {x_bar:.6f}')
print(f'Numerical MLE (scipy):       {result.x:.6f}')
print(f'Difference:                  {abs(x_bar - result.x):.2e}')
print(f'Score at analytical MLE:     {score(x_bar, data, sigma2):.6e}  (should be ~0)')
```

## Summary

| Concept | Key result or rule |
|---|---|
| Differentiability | $f'(a) = \lim_{h \to 0} \frac{f(a+h)-f(a)}{h}$; differentiable $\Rightarrow$ continuous |
| Chain rule | $\frac{d}{dx}f(g(x)) = f'(g(x)) \cdot g'(x)$ — outside times inside |
| Gaussian kernel derivative | $\frac{d}{dx}e^{-x^2/2} = -x\,e^{-x^2/2}$ |
| FTC | $\int_a^b f(x)\,dx = F(b) - F(a)$ where $F' = f$ |
| Substitution | $\int f(g(x))\,g'(x)\,dx = \int f(u)\,du$ |
| Log-likelihood (Gaussian) | $\ell(\mu) = -\frac{n}{2}\log(2\pi\sigma^2) - \frac{1}{2\sigma^2}\sum(x_i - \mu)^2$ |
| MLE for $\mu$ | $\hat{\mu}_{\text{MLE}} = \bar{x}$ — found by setting $\frac{d\ell}{d\mu} = 0$ |
| $\log$ convention | Throughout this course, $\log$ means the natural logarithm |
