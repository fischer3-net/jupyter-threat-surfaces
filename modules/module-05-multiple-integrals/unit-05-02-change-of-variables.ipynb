## Why This Unit

Unit 01 showed how to evaluate integrals over regions described in Cartesian coordinates. But many of the most important regions in probability — disks, spheres, ellipsoids — are described much more naturally in polar or spherical coordinates. And the normalizing constant of the Gaussian distribution, $\sqrt{2\pi}$, cannot be derived without switching to polar coordinates.

The change-of-variables formula is also the integration counterpart of the Jacobian from Module 03 Unit 03. There, the Jacobian matrix described how a function stretches and rotates a small neighborhood of the input. Here, $|\det(\boldsymbol{J})|$ quantifies exactly how much area is stretched — and that stretch factor must be included to keep the integral correct after substitution.

---

## 1. The Change-of-Variables Formula

### 1.1 Single-Variable Analogy

Recall from Module 00 Unit 03: for the substitution $x = g(u)$, $dx = g'(u)\,du$:

$$
\int_a^b f(x)\,dx = \int_{g^{-1}(a)}^{g^{-1}(b)} f(g(u))\,|g'(u)|\,du
$$

The factor $|g'(u)|$ corrects for the stretching of the $u$-axis by the substitution. In two dimensions, this stretching becomes area scaling.

### 1.2 The Formula

Let $\boldsymbol{T} : D' \to D$ be an invertible differentiable transformation mapping $(u, v)$-coordinates to $(x, y)$-coordinates:

$$
\boldsymbol{T}(u, v) = (x(u,v),\ y(u,v))
$$

The **Jacobian matrix** of $\boldsymbol{T}$ (from Module 03 Unit 03) is:

$$
\boldsymbol{J}_{\boldsymbol{T}}(u, v) = \begin{pmatrix} \dfrac{\partial x}{\partial u} & \dfrac{\partial x}{\partial v} \\[8pt] \dfrac{\partial y}{\partial u} & \dfrac{\partial y}{\partial v} \end{pmatrix}
$$

**Theorem (Change of Variables).** Under the transformation $\boldsymbol{T}$:

$$
\boxed{\iint_D f(x, y)\,dx\,dy = \iint_{D'} f(x(u,v),\, y(u,v))\,\left|\det(\boldsymbol{J}_{\boldsymbol{T}}(u,v))\right|\,du\,dv}
$$

The factor $|\det(\boldsymbol{J}_{\boldsymbol{T}})|$ — the **absolute Jacobian determinant** — is the local area scaling factor: it tells you how much a small rectangle in $(u,v)$-space is stretched to a parallelogram in $(x,y)$-space by $\boldsymbol{T}$.

### 1.3 Why $|\det(\boldsymbol{J})|$ is the Area Scaling Factor

From Module 03 Unit 03, the Jacobian $\boldsymbol{J}_{\boldsymbol{T}}$ is the best linear approximation to $\boldsymbol{T}$ near a point. A small rectangle with sides $\Delta u$ and $\Delta v$ maps approximately to a parallelogram spanned by:

$$
\boldsymbol{J}_{\boldsymbol{T}} \begin{pmatrix}\Delta u \\ 0\end{pmatrix} \qquad \text{and} \qquad \boldsymbol{J}_{\boldsymbol{T}} \begin{pmatrix}0 \\ \Delta v\end{pmatrix}
$$

The area of a parallelogram spanned by two vectors is $|\det|$ of the matrix formed by those vectors. Therefore the area of the image parallelogram is $|\det(\boldsymbol{J}_{\boldsymbol{T}})| \cdot \Delta u\,\Delta v$. The change-of-variables formula simply corrects for this scaling at every point.

## 2. Polar Coordinates

### 2.1 The Transformation

$$
x = r\cos\theta \qquad y = r\sin\theta \qquad r \geq 0,\ 0 \leq \theta < 2\pi
$$

### 2.2 The Jacobian

$$
\boldsymbol{J}_{\boldsymbol{T}}(r, \theta) = \begin{pmatrix} \dfrac{\partial x}{\partial r} & \dfrac{\partial x}{\partial \theta} \\[8pt] \dfrac{\partial y}{\partial r} & \dfrac{\partial y}{\partial \theta} \end{pmatrix} = \begin{pmatrix} \cos\theta & -r\sin\theta \\ \sin\theta & r\cos\theta \end{pmatrix}
$$

$$
\det(\boldsymbol{J}_{\boldsymbol{T}}) = r\cos^2\theta + r\sin^2\theta = r
$$

Therefore $|\det(\boldsymbol{J}_{\boldsymbol{T}})| = r$. The change-of-variables formula in polar coordinates is:

$$
\iint_D f(x, y)\,dx\,dy = \iint_{D'} f(r\cos\theta,\, r\sin\theta)\cdot r\,dr\,d\theta
$$

The factor of $r$ — the **polar Jacobian** — is easy to forget and almost always costs marks. It has a clear geometric meaning: a small rectangle of area $\Delta r\,\Delta\theta$ in polar space corresponds to a sector-shaped patch of area $r\,\Delta r\,\Delta\theta$ in Cartesian space. At $r = 0$ the patch has zero area; at large $r$ it is large.

**Worked Example 2.1.** Evaluate $\iint_D e^{-(x^2 + y^2)}\,dA$ over the disk $D = \{(x,y) : x^2 + y^2 \leq R^2\}$.

In polar: $x^2 + y^2 = r^2$, so $f(r\cos\theta, r\sin\theta) = e^{-r^2}$. The disk is $0 \leq r \leq R$, $0 \leq \theta \leq 2\pi$.

$$
\int_0^{2\pi}\int_0^R e^{-r^2}\cdot r\,dr\,d\theta
= 2\pi\int_0^R r\,e^{-r^2}\,dr
= 2\pi\left[-\frac{1}{2}e^{-r^2}\right]_0^R
= \pi(1 - e^{-R^2})
$$

As $R \to \infty$: $\iint_{\mathbb{R}^2} e^{-(x^2+y^2)}\,dA = \pi$.

## 3. Statistical Bridge — The Gaussian Normalization Constant `[PD]`

This is one of the most celebrated calculations in probability. The integral $\int_{-\infty}^{\infty} e^{-x^2/2}\,dx$ cannot be evaluated directly — $e^{-x^2/2}$ has no elementary antiderivative. But by converting to two dimensions and applying polar coordinates, the result falls out in a few lines.

### 3.1 The Trick

Let $I = \int_{-\infty}^{\infty} e^{-x^2/2}\,dx$. Then:

$$
I^2 = \left(\int_{-\infty}^{\infty} e^{-x^2/2}\,dx\right)\left(\int_{-\infty}^{\infty} e^{-y^2/2}\,dy\right)
$$

Since the two integrals are over independent variables, they combine into a double integral:

$$
I^2 = \iint_{\mathbb{R}^2} e^{-(x^2 + y^2)/2}\,dx\,dy
$$

### 3.2 Converting to Polar

With $x^2 + y^2 = r^2$ and Jacobian factor $r$:

$$
I^2 = \int_0^{2\pi}\int_0^{\infty} e^{-r^2/2}\cdot r\,dr\,d\theta
= 2\pi\int_0^{\infty} r\,e^{-r^2/2}\,dr
$$

### 3.3 Evaluating the Radial Integral

Substitute $u = r^2/2$, $du = r\,dr$:

$$
\int_0^{\infty} r\,e^{-r^2/2}\,dr = \int_0^{\infty} e^{-u}\,du = 1
$$

Therefore:

$$
I^2 = 2\pi \cdot 1 = 2\pi \qquad \Longrightarrow \qquad \boxed{\int_{-\infty}^{\infty} e^{-x^2/2}\,dx = \sqrt{2\pi}}
$$

### 3.4 Why This Matters

The standard normal density $\phi(x) = \frac{1}{\sqrt{2\pi}}e^{-x^2/2}$ integrates to 1 precisely because $\int e^{-x^2/2}\,dx = \sqrt{2\pi}$. Every probability calculation involving the normal distribution — $z$-scores, confidence intervals, hypothesis tests — implicitly relies on this result. Without the polar substitution, it cannot be derived.

More generally, for $\mathcal{N}(\mu, \sigma^2)$: substitute $x = \mu + \sigma t$ to get $\int_{-\infty}^{\infty} e^{-(x-\mu)^2/(2\sigma^2)}\,dx = \sigma\sqrt{2\pi}$. The denominator $\sqrt{2\pi\sigma^2}$ in the normal density is exactly this constant.

## 4. Spherical Coordinates and Triple Integrals

For integrals over three-dimensional regions, spherical coordinates are the natural generalization of polar.

### 4.1 The Transformation

$$
x = \rho\sin\phi\cos\theta \qquad y = \rho\sin\phi\sin\theta \qquad z = \rho\cos\phi
$$

where $\rho \geq 0$ (radial distance), $0 \leq \phi \leq \pi$ (polar angle from $z$-axis), $0 \leq \theta < 2\pi$ (azimuthal angle).

### 4.2 The Jacobian

$$
|\det(\boldsymbol{J}_{\boldsymbol{T}})| = \rho^2 \sin\phi
$$

The change of variables formula:

$$
\iiint_D f(x,y,z)\,dV = \iiint_{D'} f(\rho\sin\phi\cos\theta,\, \rho\sin\phi\sin\theta,\, \rho\cos\phi)\cdot\rho^2\sin\phi\,d\rho\,d\phi\,d\theta
$$

**Worked Example 4.1.** Volume of the unit ball $\rho \leq 1$:

$$
\int_0^{2\pi}\int_0^{\pi}\int_0^1 \rho^2\sin\phi\,d\rho\,d\phi\,d\theta
= \int_0^{2\pi}d\theta \cdot \int_0^{\pi}\sin\phi\,d\phi \cdot \int_0^1 \rho^2\,d\rho
= 2\pi \cdot 2 \cdot \frac{1}{3} = \frac{4\pi}{3}
$$

**Worked Example 4.2.** Compute $\iiint_{\mathbb{R}^3} e^{-(x^2+y^2+z^2)/2}\,dV$.

In spherical: $x^2+y^2+z^2 = \rho^2$:

$$
\int_0^{2\pi}\int_0^{\pi}\int_0^{\infty} e^{-\rho^2/2}\,\rho^2\sin\phi\,d\rho\,d\phi\,d\theta
= 4\pi \int_0^{\infty} \rho^2 e^{-\rho^2/2}\,d\rho
= 4\pi \cdot \frac{\sqrt{2\pi}}{4} \cdot 2 = (2\pi)^{3/2}
$$

This is the normalizing constant of the 3D standard normal density $\mathcal{N}(\boldsymbol{0}, \boldsymbol{I}_3)$.

### Section 1 — Symbolic Jacobian Determinants and Change of Variables

Verify the polar Jacobian $r$ and the spherical Jacobian $\rho^2\sin\phi$ symbolically, then confirm the worked examples.

```python
# ============================================================
# Section 1 — Symbolic Jacobians and change-of-variables verification
# ============================================================
r, theta, rho, phi = symbols('r theta rho phi', positive=True)

# --- Polar Jacobian ---
x_pol = r * cos(theta)
y_pol = r * sin(theta)
J_pol = Matrix([[x_pol.diff(r), x_pol.diff(theta)],
                [y_pol.diff(r), y_pol.diff(theta)]])
det_pol = simplify(J_pol.det())
print('Polar Jacobian matrix:')
sp.pprint(J_pol)
print('det(J_polar) =', det_pol, '  →  |det| = r ✓')

# --- Spherical Jacobian ---
x_sph = rho * sin(phi) * cos(theta)
y_sph = rho * sin(phi) * sin(theta)
z_sph = rho * cos(phi)
J_sph = Matrix([[x_sph.diff(rho), x_sph.diff(phi), x_sph.diff(theta)],
                [y_sph.diff(rho), y_sph.diff(phi), y_sph.diff(theta)],
                [z_sph.diff(rho), z_sph.diff(phi), z_sph.diff(theta)]])
det_sph = simplify(J_sph.det())
print('\nSpherical Jacobian det:', det_sph)
print('Simplified:', simplify(det_sph - rho**2 * sin(phi)) == 0, ' → det = ρ²sin(φ) ✓')

# --- Example 2.1: ∫∫ e^(−r²) dA over disk radius R ---
R = symbols('R', positive=True)
integral_disk = integrate(integrate(exp(-r**2) * r, (r, 0, R)), (theta, 0, 2*pi))
print('\nExample 2.1  ∫∫ e^(−r²) over disk radius R:')
print(' ', simplify(integral_disk))
print('  R→∞ limit:', sp.limit(integral_disk, R, oo))

# --- Gaussian normalization: I² = 2π ---
I_sq = integrate(integrate(exp(-r**2/2) * r, (r, 0, oo)), (theta, 0, 2*pi))
print('\nGaussian normalization  I² = ∫∫ e^(−r²/2) r dr dθ:')
print(' ', I_sq, '= 2π?', simplify(I_sq - 2*pi) == 0)
print('  Therefore I = √(2π) =', sqrt(2*pi))
```

**What do you see?** SymPy confirms both Jacobian determinants symbolically — the polar determinant simplifies to $r$ and the spherical to $\rho^2\sin\phi$. The Gaussian normalization integral evaluates to exactly $2\pi$, confirming $I = \sqrt{2\pi}$.

### Section 2 — Geometric Visualization of the Jacobian as Area Scaling

Show how a uniform grid in $(r,\theta)$-space maps to a non-uniform grid in $(x,y)$-space, with each cell's area proportional to $r\,\Delta r\,\Delta\theta$. This makes the $r$ Jacobian factor geometrically obvious.

```python
# ============================================================
# Section 2 — Jacobian as area scaling: polar grid visualization
# ============================================================
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

nr = 6;  ntheta = 12
r_vals     = np.linspace(0.1, 2.0, nr + 1)
theta_vals = np.linspace(0, 2*np.pi, ntheta + 1)

# --- Left: uniform grid in (r, θ) space ---
ax = axes[0]
for rv in r_vals:
    ax.axhline(rv, color=TS_GRAY, lw=0.8, alpha=0.6)
for tv in theta_vals:
    ax.axvline(tv, color=TS_GRAY, lw=0.8, alpha=0.6)

# Shade two cells: one near origin (small r), one far (large r)
for i_r, color in [(0, TS_AMBER), (4, TS_BLUE)]:
    r0, r1 = r_vals[i_r], r_vals[i_r+1]
    t0, t1 = theta_vals[2], theta_vals[3]
    rect = plt.Polygon([(t0,r0),(t1,r0),(t1,r1),(t0,r1)],
                       facecolor=color, alpha=0.4, edgecolor=color, lw=1.5)
    ax.add_patch(rect)
    area = (r1-r0)*(t1-t0)
    ax.text((t0+t1)/2, (r0+r1)/2, f'{area:.3f}',
            ha='center', va='center', fontsize=8, color='black')

ax.set_xlim(-0.1, 2*np.pi+0.1)
ax.set_ylim(-0.1, 2.2)
ax.set_xlabel(r'$\theta$')
ax.set_ylabel(r'$r$')
ax.set_title('$(r, \\theta)$ space — uniform grid\ncell area = $\\Delta r\\,\\Delta\\theta$ (same for all cells)')

# --- Right: same cells mapped to (x,y) space ---
ax2 = axes[1]
# Draw polar grid in Cartesian
for rv in r_vals:
    circ = plt.Circle((0,0), rv, fill=False, color=TS_GRAY, lw=0.8, alpha=0.5)
    ax2.add_patch(circ)
for tv in theta_vals:
    ax2.plot([0, 2.1*np.cos(tv)], [0, 2.1*np.sin(tv)],
             color=TS_GRAY, lw=0.8, alpha=0.5)

# Shade same two cells in Cartesian
for i_r, color in [(0, TS_AMBER), (4, TS_BLUE)]:
    r0, r1 = r_vals[i_r], r_vals[i_r+1]
    t0, t1 = theta_vals[2], theta_vals[3]
    t_arc = np.linspace(t0, t1, 30)
    outer = np.column_stack([r1*np.cos(t_arc), r1*np.sin(t_arc)])
    inner = np.column_stack([r0*np.cos(t_arc[::-1]), r0*np.sin(t_arc[::-1])])
    patch_pts = np.vstack([outer, inner])
    poly = plt.Polygon(patch_pts, facecolor=color, alpha=0.4,
                       edgecolor=color, lw=1.5)
    ax2.add_patch(poly)
    r_mid = (r0+r1)/2
    t_mid = (t0+t1)/2
    approx_area = r_mid*(r1-r0)*(t1-t0)
    ax2.text(r_mid*np.cos(t_mid)*0.85, r_mid*np.sin(t_mid)*0.85,
             f'≈{approx_area:.3f}',
             ha='center', va='center', fontsize=8, color='black')

ax2.set_xlim(-2.3, 2.3); ax2.set_ylim(-2.3, 2.3)
ax2.set_aspect('equal')
ax2.set_xlabel('x'); ax2.set_ylabel('y')
ax2.set_title('$(x, y)$ space — cells have different areas\narea $\\approx r\\,\\Delta r\\,\\Delta\\theta$ (larger at large $r$)')

plt.suptitle('Jacobian factor $r$: equal $(r,\\theta)$ cells map to unequal $(x,y)$ areas',
             fontsize=11, y=1.01)
plt.tight_layout()
plt.show()
```

**What do you see?** In $(r,\theta)$-space (left), the two highlighted cells have identical area $\Delta r\,\Delta\theta$. In $(x,y)$-space (right), the same cells have very different areas — the cell near the origin (amber) is tiny, while the cell far from the origin (blue) is much larger. The Jacobian factor $r$ exactly captures this discrepancy: the area in Cartesian space is $r\,\Delta r\,\Delta\theta$, not $\Delta r\,\Delta\theta$. Without the factor of $r$, the integral would give the wrong answer.

### Section 3 — Statistical Bridge: Gaussian Normalization — Numerical and Analytical

Verify the Gaussian normalization constant numerically, visualize how the polar integral accumulates, and demonstrate that the integral of the standard normal density is exactly 1.

```python
# ============================================================
# Section 3 — Gaussian normalization: polar trick visualization
# ============================================================
from scipy import integrate as sci_int

# --- Step 1: show the 1D integral has no elementary antiderivative ---
# But numerically: I = ∫_{-R}^{R} e^{-x²/2} dx → √(2π) as R→∞
R_vals = np.logspace(-1, 2, 200)
I_vals = np.array([sci_int.quad(lambda x: np.exp(-x**2/2), -R, R)[0] for R in R_vals])
sqrt2pi = np.sqrt(2 * np.pi)

# --- Step 2: the radial integral I² = 2π accumulated over r ---
r_vals_rad = np.linspace(0, 6, 500)
integrand_r = r_vals_rad * np.exp(-r_vals_rad**2 / 2)  # r·e^(−r²/2)
I_sq_cumulative = 2 * np.pi * np.cumsum(integrand_r) * (r_vals_rad[1] - r_vals_rad[0])

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Left: 1D integral converging to √(2π)
ax = axes[0]
ax.semilogx(R_vals, I_vals, color=TS_BLUE, lw=2, label=r'$\int_{-R}^{R} e^{-x^2/2}\,dx$')
ax.axhline(sqrt2pi, color=TS_AMBER, lw=2, ls='--',
           label=fr'$\sqrt{{2\pi}} \approx {sqrt2pi:.5f}$')
ax.set_xlabel('R (log scale)')
ax.set_ylabel('Integral value')
ax.set_title('1D Gaussian integral converges\nto $\\sqrt{2\\pi}$ as $R \\to \\infty$')
ax.legend(fontsize=9)

# Middle: radial integrand r·e^(−r²/2) and cumulative I²
ax2 = axes[1]
ax2_r = ax2.twinx()
ax2.plot(r_vals_rad, integrand_r, color=TS_BLUE, lw=2,
         label=r'$r\,e^{-r^2/2}$ (integrand)')
ax2_r.plot(r_vals_rad, I_sq_cumulative, color=TS_GREEN, lw=2, ls='--',
           label=r'$2\pi\int_0^r \cdot\,dr$ (cumulative)')
ax2_r.axhline(2*np.pi, color=TS_AMBER, lw=1.5, ls=':',
              label=r'$2\pi$')
ax2.set_xlabel('r')
ax2.set_ylabel(r'$r\,e^{-r^2/2}$', color=TS_BLUE)
ax2_r.set_ylabel(r'Cumulative $I^2$', color=TS_GREEN)
ax2.set_title(r'Polar integral: $I^2 = 2\pi\int_0^\infty r\,e^{-r^2/2}\,dr = 2\pi$')
lines1, labels1 = ax2.get_legend_handles_labels()
lines2, labels2 = ax2_r.get_legend_handles_labels()
ax2.legend(lines1+lines2, labels1+labels2, fontsize=8)

# Right: standard normal density integrates to 1
ax3 = axes[2]
x_plot = np.linspace(-4, 4, 400)
phi = np.exp(-x_plot**2/2) / sqrt2pi
ax3.plot(x_plot, phi, color=TS_BLUE, lw=2.5,
         label=r'$\phi(x) = \frac{1}{\sqrt{2\pi}}e^{-x^2/2}$')
ax3.fill_between(x_plot, phi, alpha=0.15, color=TS_BLUE)
# Shade ±1σ and ±2σ
for lo, hi, col, lab in [(-1,1,TS_GREEN,'±1σ: 68.3%'),
                          (-2,2,TS_AMBER,'±2σ: 95.4%')]:
    mask = (x_plot >= lo) & (x_plot <= hi)
    ax3.fill_between(x_plot[mask], phi[mask], alpha=0.25, color=col, label=lab)

norm_const, _ = sci_int.quad(lambda x: np.exp(-x**2/2)/sqrt2pi, -np.inf, np.inf)
ax3.set_xlabel('x')
ax3.set_ylabel(r'$\phi(x)$')
ax3.set_title(f'Standard normal density\n$\\int_{{-\\infty}}^{{\\infty}}\\phi(x)\\,dx = {norm_const:.8f}$')
ax3.legend(fontsize=9)

plt.suptitle(r'Gaussian normalization: $\int e^{-x^2/2}\,dx = \sqrt{2\pi}$ from the polar trick',
             fontsize=11, y=1.01)
plt.tight_layout()
plt.show()

print(f'∫φ(x)dx = {norm_const:.10f}  (exact: 1.0000000000)')
print(f'√(2π)   = {sqrt2pi:.10f}')
```

**What do you see?**

- **Left**: the 1D Gaussian integral converges to $\sqrt{2\pi}$ from below as $R$ grows. The convergence is rapid — by $R = 5$ the integral is essentially at its limit.
- **Middle**: the radial integrand $r\,e^{-r^2/2}$ rises from zero, peaks at $r=1$, then decays. Its integral over $[0, \infty)$ is exactly 1 (from the substitution $u = r^2/2$), giving $I^2 = 2\pi$.
- **Right**: the standard normal density integrates to 1.0000000000 — confirmed to 10 decimal places. The $\pm 1\sigma$ and $\pm 2\sigma$ regions contain 68.3\% and 95.4\% of the probability mass respectively, the familiar empirical rule.

**The key point**: the single line `I² = 2π` follows from converting a product of two 1D integrals into a 2D polar integral. This is perhaps the most elegant application of the change-of-variables formula in all of probability.
