## 1. The Double Integral over a Rectangle

### 1.1 Definition

Let $f : R \to \mathbb{R}$ be a bounded function on the rectangle $R = [a, b] \times [c, d]$. Partition $[a,b]$ into $m$ subintervals and $[c,d]$ into $n$ subintervals, forming an $m \times n$ grid of sub-rectangles with area $\Delta A_{ij} = \Delta x_i \Delta y_j$. Pick a sample point $(x_{ij}^*, y_{ij}^*)$ in each sub-rectangle. The **Riemann sum** is:

$$
S_{mn} = \sum_{i=1}^m \sum_{j=1}^n f(x_{ij}^*, y_{ij}^*)\,\Delta A_{ij}
$$

**Definition.** The **double integral** of $f$ over $R$ is:

$$
\iint_R f(x, y)\,dA \coloneqq \lim_{\|\Delta\| \to 0} S_{mn}
$$

when this limit exists (i.e., is the same regardless of how the sample points are chosen).

### 1.2 Geometric Interpretation

When $f(x, y) \geq 0$, the double integral $\iint_R f\,dA$ is the **volume** of the solid region above $R$ and below the surface $z = f(x, y)$. Each term $f(x_{ij}^*, y_{ij}^*)\,\Delta A_{ij}$ is the volume of a thin rectangular column of height $f$ and base area $\Delta A$.

When $f$ takes both positive and negative values, the integral is signed volume — positive where $f > 0$, negative where $f < 0$.

### 1.3 Properties

The double integral inherits all linearity properties from the single-variable integral:

$$
\iint_R (af + bg)\,dA = a\iint_R f\,dA + b\iint_R g\,dA
$$
$$
\iint_{R_1 \cup R_2} f\,dA = \iint_{R_1} f\,dA + \iint_{R_2} f\,dA \quad \text{(disjoint } R_1, R_2\text{)}
$$
$$
\iint_R 1\,dA = \text{area}(R)
$$

## 2. Fubini's Theorem

The definition as a limit of Riemann sums is theoretically important but computationally useless. Fubini's theorem is what makes double integrals computable.

**Theorem (Fubini).** Let $f$ be continuous on the rectangle $R = [a, b] \times [c, d]$. Then:

$$
\iint_R f(x, y)\,dA = \int_a^b \left[\int_c^d f(x, y)\,dy\right]dx = \int_c^d \left[\int_a^b f(x, y)\,dx\right]dy
$$

Both iterated integrals give the same result. You can integrate in either order.

**How to read this.** The inner integral $\int_c^d f(x, y)\,dy$ treats $x$ as a constant and integrates over $y$ — this is exactly a partial integral, the integration analogue of the partial derivative. The outer integral then integrates the result over $x$.

**Worked Example 2.1.** Evaluate $\iint_R (x^2 y + 3y^2)\,dA$ over $R = [0, 2] \times [1, 3]$.

Integrate $y$ first (inner), then $x$ (outer):

$$
\int_0^2 \left[\int_1^3 (x^2 y + 3y^2)\,dy\right]dx
= \int_0^2 \left[\frac{x^2 y^2}{2} + y^3\right]_{y=1}^{y=3}\,dx
= \int_0^2 \left[(\tfrac{9x^2}{2} + 27) - (\tfrac{x^2}{2} + 1)\right]dx
$$

$$
= \int_0^2 (4x^2 + 26)\,dx
= \left[\frac{4x^3}{3} + 26x\right]_0^2
= \frac{32}{3} + 52 = \frac{188}{3}
$$

**Worked Example 2.2.** Evaluate $\iint_R e^{x+y}\,dA$ over $R = [0,1] \times [0,1]$.

$$
\int_0^1 \left[\int_0^1 e^{x+y}\,dy\right]dx
= \int_0^1 \left[e^{x+y}\right]_{y=0}^{y=1}\,dx
= \int_0^1 (e^{x+1} - e^x)\,dx
= (e-1)\int_0^1 e^x\,dx
= (e-1)^2
$$

**When $f(x,y) = g(x)h(y)$ on a rectangle, the integral factors:**

$$
\iint_{[a,b]\times[c,d]} g(x)h(y)\,dA = \left(\int_a^b g(x)\,dx\right)\left(\int_c^d h(y)\,dy\right)
$$

This factorization is fundamental for **independent random variables** — a joint density that factors as $f(x,y) = f_X(x)\,f_Y(y)$ has independent components, and all joint integrals factor into products of univariate integrals.

## 3. Double Integrals over Non-Rectangular Regions

Most regions of interest are not rectangles. We handle them by describing the region with variable limits.

### 3.1 Type I and Type II Regions

**Type I** (bounded by functions of $x$): For region $D = \{(x,y) : a \leq x \leq b,\ g_1(x) \leq y \leq g_2(x)\}$:

$$
\iint_D f(x,y)\,dA = \int_a^b \int_{g_1(x)}^{g_2(x)} f(x,y)\,dy\,dx
$$

**Type II** (bounded by functions of $y$): For region $D = \{(x,y) : c \leq y \leq d,\ h_1(y) \leq x \leq h_2(y)\}$:

$$
\iint_D f(x,y)\,dA = \int_c^d \int_{h_1(y)}^{h_2(y)} f(x,y)\,dx\,dy
$$

**Worked Example 3.1.** Evaluate $\iint_D xy\,dA$ where $D$ is the region bounded by $y = x$ and $y = x^2$.

The curves intersect at $(0,0)$ and $(1,1)$. For $0 \leq x \leq 1$: $x^2 \leq y \leq x$.

$$
\int_0^1 \int_{x^2}^{x} xy\,dy\,dx
= \int_0^1 x\left[\frac{y^2}{2}\right]_{x^2}^{x}\,dx
= \int_0^1 x \cdot \frac{x^2 - x^4}{2}\,dx
= \frac{1}{2}\int_0^1 (x^3 - x^5)\,dx
= \frac{1}{2}\left[\frac{x^4}{4} - \frac{x^6}{6}\right]_0^1
= \frac{1}{2}\cdot\frac{1}{12} = \frac{1}{24}
$$

### 3.2 Reversing Integration Order

Sometimes the inner integral $\int f(x,y)\,dy$ has no closed form, but $\int f(x,y)\,dx$ does. Reversing the order of integration — re-describing the same region as Type II instead of Type I — can make an intractable integral computable.

**Worked Example 3.2.** Evaluate $\int_0^1 \int_x^1 e^{y^2}\,dy\,dx$.

$e^{y^2}$ has no elementary antiderivative in $y$, so the inner integral cannot be computed as written.

**Step 1.** Identify the region: $0 \leq x \leq 1$ and $x \leq y \leq 1$, i.e., $\{(x,y) : 0 \leq x \leq y \leq 1\}$.

**Step 2.** Re-describe as Type II: $0 \leq y \leq 1$ and $0 \leq x \leq y$.

$$
\int_0^1 \int_0^y e^{y^2}\,dx\,dy
= \int_0^1 e^{y^2} \cdot y\,dy
= \frac{1}{2}\left[e^{y^2}\right]_0^1
= \frac{e - 1}{2}
$$

## 4. Statistical Bridge — Joint Densities and Marginals `[PD]`

### 4.1 Joint Probability Densities

A **joint probability density function** $f(x, y)$ for two continuous random variables $(X, Y)$ satisfies:

$$
f(x, y) \geq 0 \qquad \text{and} \qquad \iint_{\mathbb{R}^2} f(x, y)\,dA = 1
$$

The probability that $(X, Y)$ falls in a region $D$ is:

$$
P((X,Y) \in D) = \iint_D f(x, y)\,dA
$$

### 4.2 Marginal Densities via Iterated Integration

The **marginal density** of $X$ alone is obtained by integrating out $Y$:

$$
f_X(x) = \int_{-\infty}^{\infty} f(x, y)\,dy
$$

This is exactly the inner integral in Fubini's theorem. We integrate over all possible values of $Y$ to recover the distribution of $X$ alone.

Similarly, $f_Y(y) = \int_{-\infty}^{\infty} f(x, y)\,dx$.

### 4.3 Worked Example — Uniform Joint Density

Let $(X, Y)$ have joint density $f(x, y) = 6x$ on the region $D = \{(x,y) : 0 \leq x \leq 1,\ 0 \leq y \leq x\}$ (a triangle).

**Verify normalization:**

$$
\int_0^1 \int_0^x 6x\,dy\,dx = \int_0^1 6x \cdot x\,dx = \int_0^1 6x^2\,dx = \left[2x^3\right]_0^1 = 2 \neq 1
$$

Adjusting: use $f(x,y) = 3x$ instead. Then $\int_0^1 3x^2\,dx = 1$. ✓

**Marginal density of $X$:**

$$
f_X(x) = \int_0^x 3x\,dy = 3x \cdot x = 3x^2, \qquad 0 \leq x \leq 1
$$

**Marginal density of $Y$:**

$$
f_Y(y) = \int_y^1 3x\,dx = \frac{3}{2}\left[x^2\right]_y^1 = \frac{3}{2}(1 - y^2), \qquad 0 \leq y \leq 1
$$

**Note the region bounds**: when integrating out $x$ to get $f_Y(y)$, the lower limit is $y$ (not $0$) because the constraint is $y \leq x$. Setting up integration limits correctly over non-rectangular regions is the key skill.

### 4.4 Conditional Density

The conditional density of $Y$ given $X = x$ is:

$$
f_{Y|X}(y \mid x) = \frac{f(x, y)}{f_X(x)}
$$

For the example above: $f_{Y|X}(y \mid x) = \frac{3x}{3x^2} = \frac{1}{x}$ for $0 \leq y \leq x$ — a uniform distribution on $[0, x]$. This is the definition of $Y \mid X = x \sim \text{Uniform}(0, x)$, which matches the triangular construction of the joint density.

### Section 1 — Symbolic Verification of Worked Examples

Use SymPy to verify the iterated integrals from Sections 2 and 3, and confirm the order-reversal trick on Example 3.2.

```python
# ============================================================
# Section 1 — Symbolic verification
# ============================================================
x, y = symbols('x y', real=True)

# --- Example 2.1: ∫∫ (x²y + 3y²) dA, R = [0,2]×[1,3] ---
f21 = x**2 * y + 3*y**2
inner_21 = integrate(f21, (y, 1, 3))
result_21 = integrate(inner_21, (x, 0, 2))
print('Example 2.1  ∫∫ (x²y + 3y²) dA over [0,2]×[1,3]')
print('  Result:', result_21, '= 188/3?', result_21 == sp.Rational(188, 3))

# Verify same result in reverse order
inner_21r = integrate(f21, (x, 0, 2))
result_21r = integrate(inner_21r, (y, 1, 3))
print('  Reversed order:', result_21r, '  Match:', result_21 == result_21r)

# --- Example 2.2: ∫∫ e^(x+y) dA, R = [0,1]×[0,1] ---
f22 = exp(x + y)
result_22 = integrate(integrate(f22, (y, 0, 1)), (x, 0, 1))
print('\nExample 2.2  ∫∫ e^(x+y) dA over [0,1]×[0,1]')
print('  Result:', simplify(result_22))
print('  Equals (e-1)²?', simplify(result_22 - (sp.E-1)**2) == 0)

# --- Example 3.1: ∫∫ xy dA, D between y=x and y=x² ---
f31 = x * y
result_31 = integrate(integrate(f31, (y, x**2, x)), (x, 0, 1))
print('\nExample 3.1  ∫∫ xy dA between y=x and y=x²')
print('  Result:', result_31, '= 1/24?', result_31 == sp.Rational(1, 24))

# --- Example 3.2: order reversal for ∫∫ e^(y²) ---
# Original order: ∫₀¹ ∫ₓ¹ e^(y²) dy dx  — inner integral has no closed form
# Reversed: ∫₀¹ ∫₀ʸ e^(y²) dx dy
f32 = exp(y**2)
result_32 = integrate(integrate(f32, (x, 0, y)), (y, 0, 1))
print('\nExample 3.2  Order reversal for ∫∫ e^(y²)')
print('  Result (reversed order):', simplify(result_32))
print('  Equals (e-1)/2?', simplify(result_32 - (sp.E - 1)/2) == 0)
```

**What do you see?** All results match the hand calculations. Note especially Example 3.2 — SymPy cannot evaluate the original order (inner integral of $e^{y^2}$ over $y$ has no closed form), but the reversed order produces $(e-1)/2$ immediately. Order reversal is not just a trick — it is often the only way to compute an integral in closed form.

### Section 2 — Riemann Sum Convergence

Visualize how the double Riemann sum converges to the exact double integral as the grid is refined. This grounds the definition and shows that the computable iterated integral formula is the limit of a geometric construction.

```python
# ============================================================
# Section 2 — Riemann sum convergence visualization
# ============================================================
# f(x,y) = x²y + 3y²  on R = [0,2]×[1,3]
# Exact value: 188/3 ≈ 62.667

def f_ex(x, y):
    return x**2 * y + 3*y**2

exact = 188 / 3

def riemann_sum(n_x, n_y, corner='lower-left'):
    """Compute Riemann sum on [0,2]×[1,3] with n_x × n_y grid."""
    xs = np.linspace(0, 2, n_x + 1)
    ys = np.linspace(1, 3, n_y + 1)
    dx = 2 / n_x;  dy = 2 / n_y
    total = 0.0
    for i in range(n_x):
        for j in range(n_y):
            if corner == 'lower-left':
                xij, yij = xs[i], ys[j]
            else:
                xij = (xs[i] + xs[i+1]) / 2
                yij = (ys[j] + ys[j+1]) / 2
            total += f_ex(xij, yij) * dx * dy
    return total

ns = [1, 2, 4, 8, 16, 32, 64, 128]
rs_ll  = [riemann_sum(n, n, 'lower-left') for n in ns]
rs_mid = [riemann_sum(n, n, 'midpoint')   for n in ns]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Left: convergence plot
ax = axes[0]
ax.semilogx(ns, rs_ll,  'o-', color=TS_BLUE,  lw=2, label='Lower-left corners')
ax.semilogx(ns, rs_mid, 's-', color=TS_AMBER, lw=2, label='Midpoints')
ax.axhline(exact, color=TS_GREEN, lw=2, ls='--',
           label=f'Exact = 188/3 ≈ {exact:.4f}')
ax.set_xlabel('Grid size $n \\times n$ (log scale)')
ax.set_ylabel('Riemann sum value')
ax.set_title('Convergence of double Riemann sum')
ax.legend(fontsize=9)

# Right: 3D visualization of the surface and columns for n=4
ax3d = fig.add_subplot(122, projection='3d')
xg = np.linspace(0, 2, 60)
yg = np.linspace(1, 3, 60)
Xm, Ym = np.meshgrid(xg, yg)
Zm = f_ex(Xm, Ym)
ax3d.plot_surface(Xm, Ym, Zm, alpha=0.35, color=TS_BLUE, linewidth=0)

n_col = 4
xs4 = np.linspace(0, 2, n_col + 1)
ys4 = np.linspace(1, 3, n_col + 1)
dx4 = 2/n_col;  dy4 = 2/n_col
for i in range(n_col):
    for j in range(n_col):
        xi, yj = xs4[i], ys4[j]
        h = f_ex(xi, yj)
        verts = [
            [xi, yj, 0], [xi+dx4, yj, 0],
            [xi+dx4, yj+dy4, 0], [xi, yj+dy4, 0],
            [xi, yj, h], [xi+dx4, yj, h],
            [xi+dx4, yj+dy4, h], [xi, yj+dy4, h],
        ]
        from mpl_toolkits.mplot3d.art3d import Poly3DCollection
        top = [[verts[4], verts[5], verts[6], verts[7]]]
        ax3d.add_collection3d(Poly3DCollection(top, alpha=0.25,
                              facecolor=TS_AMBER, edgecolor=TS_GRAY, linewidth=0.4))

ax3d.set_xlabel('x'); ax3d.set_ylabel('y'); ax3d.set_zlabel('f(x,y)')
ax3d.set_title(f'Riemann columns (n=4, S={riemann_sum(4,4):.2f})')
ax3d.view_init(elev=25, azim=-50)

plt.suptitle(r'Double Riemann sum $\to$ exact integral as $n \to \infty$',
             fontsize=11, y=1.01)
plt.tight_layout()
plt.show()

print(f'n=128 midpoint error: {abs(rs_mid[-1] - exact):.2e}')
```

**What do you see?** Both Riemann sums converge to the exact value $188/3 \approx 62.667$ as the grid is refined. The midpoint rule converges much faster (error $O(n^{-2})$) than the lower-left corner rule. The 3D panel shows the columns — each column is a rectangular box of base area $\Delta x\,\Delta y$ and height $f(x_{ij}^*, y_{ij}^*)$. The sum of all column volumes is the Riemann sum; in the limit it becomes the volume under the surface.

### Section 3 — Statistical Bridge: Joint Density, Marginals, and Conditional

Visualize a joint density $f(x,y)$ over a non-rectangular region, compute marginals by integration, and verify numerically via Monte Carlo sampling.

```python
# ============================================================
# Section 3 — Joint density, marginals, Monte Carlo verification
# ============================================================
# f(x,y) = 3x  on {0 ≤ y ≤ x ≤ 1}  (normalized, from Section 4.3)

# --- Marginal densities (exact) ---
x_sym, y_sym = sp.symbols('x y', positive=True)
f_joint = 3 * x_sym

fX = sp.integrate(f_joint, (y_sym, 0, x_sym))
fY = sp.integrate(f_joint, (x_sym, y_sym, 1))
print('Marginal f_X(x) =', fX, '(should be 3x²)')
print('Marginal f_Y(y) =', fY, '(should be 3/2·(1−y²))')
print('∫f_X dx =', sp.integrate(fX, (x_sym, 0, 1)))
print('∫f_Y dy =', sp.integrate(fY, (y_sym, 0, 1)))

# --- Monte Carlo verification ---
rng = np.random.default_rng(42)
N = 200_000

# Sample from f(x,y)=3x on {0≤y≤x≤1} via rejection sampling
# Proposal: uniform on [0,1]×[0,1], density = 1
# Accept if y <= x (enforces the triangular region)
u1 = rng.uniform(0, 1, N*3)
u2 = rng.uniform(0, 1, N*3)
# Weight by 3x, normalize by max weight = 3
accept = (u2 <= u1) & (rng.uniform(0, 1, N*3) <= u1)  # acceptance prob ∝ 3x
x_samp = u1[accept][:N]
y_samp = u2[accept & (u2 <= u1)][:N]   # keep y ≤ x

# Simpler: direct inverse CDF sampling
# F_X(x)=x³, so X = U^(1/3);  Y|X ~ Uniform(0,X)
X_mc = rng.uniform(0, 1, N)**(1/3)
Y_mc = rng.uniform(0, 1, N) * X_mc

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Left: joint density surface
ax = axes[0]
xg = np.linspace(0, 1, 200)
yg = np.linspace(0, 1, 200)
Xg, Yg = np.meshgrid(xg, yg)
Fg = np.where(Yg <= Xg, 3*Xg, np.nan)
cf = ax.contourf(Xg, Yg, Fg, levels=15, cmap='Blues', alpha=0.75)
plt.colorbar(cf, ax=ax, label='$f(x,y) = 3x$')
ax.plot([0,1],[0,1], color=TS_AMBER, lw=2, ls='--', label='$y=x$ boundary')
ax.set_xlabel('x'); ax.set_ylabel('y')
ax.set_title('Joint density $f(x,y)=3x$\non $\\{0 \\leq y \\leq x \\leq 1\\}$')
ax.legend(fontsize=9); ax.set_aspect('equal')

# Middle: marginal of X
ax2 = axes[1]
xp = np.linspace(0, 1, 200)
ax2.plot(xp, 3*xp**2, color=TS_BLUE, lw=2.5, label='Exact $f_X(x)=3x^2$')
ax2.hist(X_mc, bins=40, density=True, color=TS_AMBER, alpha=0.55,
         edgecolor='white', label='MC samples')
ax2.set_xlabel('x'); ax2.set_ylabel('density')
ax2.set_title('Marginal of $X$')
ax2.legend(fontsize=9)

# Right: marginal of Y
ax3 = axes[2]
yp = np.linspace(0, 1, 200)
ax3.plot(yp, 1.5*(1 - yp**2), color=TS_GREEN, lw=2.5,
         label=r'Exact $f_Y(y)=\frac{3}{2}(1-y^2)$')
ax3.hist(Y_mc, bins=40, density=True, color=TS_RED, alpha=0.55,
         edgecolor='white', label='MC samples')
ax3.set_xlabel('y'); ax3.set_ylabel('density')
ax3.set_title('Marginal of $Y$')
ax3.legend(fontsize=9)

plt.suptitle('Joint density → marginals via iterated integration', fontsize=11, y=1.01)
plt.tight_layout()
plt.show()

print(f'MC estimate E[X] = {X_mc.mean():.4f}  (exact: {sp.integrate(x_sym*fX,(x_sym,0,1))})')
print(f'MC estimate E[Y] = {Y_mc.mean():.4f}  (exact: {sp.integrate(y_sym*fY,(y_sym,0,1))})')
```

**What do you see?**

- **Joint density**: $f(x,y) = 3x$ is zero above the line $y=x$ (outside the triangular support). The density increases with $x$ — higher $x$ values carry more probability mass.
- **Marginals**: the exact curves $f_X(x) = 3x^2$ and $f_Y(y) = \frac{3}{2}(1-y^2)$ match the MC histograms closely. Each marginal was obtained by integrating the joint density over the other variable — one iterated integral per marginal.
- **Monte Carlo as numerical integration**: the histogram bars are an empirical approximation to the marginal — a Monte Carlo estimate of the integral. The sample mean $E[X]$ printed at the bottom should be close to the exact value $\int_0^1 x \cdot 3x^2\,dx = 3/4$.

| Double integral | $\iint_R f\,dA = \lim_{\|\Delta\|\to 0}\sum_{ij} f(x_{ij}^*, y_{ij}^*)\Delta A_{ij}$ |
| Fubini's theorem | $\iint_R f\,dA = \int_a^b\int_c^d f\,dy\,dx = \int_c^d\int_a^b f\,dx\,dy$ |
| Factoring property | $\iint f(x)g(y)\,dA = \left(\int f\,dx\right)\left(\int g\,dy\right)$ |
| Type I region | $\int_a^b\int_{g_1(x)}^{g_2(x)} f\,dy\,dx$ |
| Type II region | $\int_c^d\int_{h_1(y)}^{h_2(y)} f\,dx\,dy$ |
| Joint density | $\iint_{\mathbb{R}^2} f(x,y)\,dA = 1$; $P((X,Y)\in D) = \iint_D f\,dA$ |
| Marginal density | $f_X(x) = \int_{-\infty}^{\infty} f(x,y)\,dy$ |
| Conditional density | $f_{Y|X}(y|x) = f(x,y)/f_X(x)$ |
